# Demostración de la fase aleatoria en BSB/DSB

En BSB/DSB, si el receptor no conoce la fase de la portadora, se asume que cualquier fase dentro de un ciclo completo es igualmente probable.

Un ciclo completo de fase es:

$$
0 \leq \phi < 2\pi
$$

Como todas las fases son igualmente probables, la densidad debe ser constante:

$$
p_\Phi(\phi)=C
$$

La densidad debe cumplir:

$$
\int_0^{2\pi} p_\Phi(\phi)\,d\phi=1
$$

Sustituyendo:

$$
\int_0^{2\pi} C\,d\phi=1
$$

$$
C[\phi]_0^{2\pi}=1
$$

$$
C(2\pi)=1
$$

$$
C=\frac{1}{2\pi}
$$

Por tanto:

$$
p_\Phi(\phi)=\frac{1}{2\pi}
$$

para:

$$
0 \leq \phi < 2\pi
$$

---

# Aplicación en BSB/DSB

La señal transmitida es:

$$
s(t)=m(t)\cos(\omega_0 t)
$$

El receptor demodula con una portadora local con error de fase:

$$
\cos(\omega_0 t+\phi)
$$

Entonces:

$$
z(t)=m(t)\cos(\omega_0 t)\cos(\omega_0 t+\phi)
$$

Usando:

$$
\cos A\cos B=
\frac{1}{2}\cos(A-B)+\frac{1}{2}\cos(A+B)
$$

con:

$$
A=\omega_0 t
$$

y:

$$
B=\omega_0 t+\phi
$$

queda:

$$
z(t)=
\frac{1}{2}m(t)\cos(-\phi)
+
\frac{1}{2}m(t)\cos(2\omega_0 t+\phi)
$$

Como:

$$
\cos(-\phi)=\cos(\phi)
$$

entonces:

$$
z(t)=
\frac{1}{2}m(t)\cos(\phi)
+
\frac{1}{2}m(t)\cos(2\omega_0 t+\phi)
$$

Después del filtro pasa bajas se elimina la componente de alta frecuencia:

$$
\frac{1}{2}m(t)\cos(2\omega_0 t+\phi)
$$

y queda:

$$
z_{LPF}(t)=\frac{1}{2}m(t)\cos(\phi)
$$

La ganancia de amplitud depende de:

$$
\cos(\phi)
$$

La ganancia de potencia depende de:

$$
\cos^2(\phi)
$$

Como:

$$
p_\Phi(\phi)=\frac{1}{2\pi}
$$

la ganancia promedio de potencia es:

$$
E[\cos^2(\Phi)]
=
\int_0^{2\pi}
\cos^2(\phi)
\frac{1}{2\pi}
\,d\phi
$$

Usando:

$$
\cos^2(\phi)=\frac{1+\cos(2\phi)}{2}
$$

entonces:

$$
E[\cos^2(\Phi)]
=
\frac{1}{2\pi}
\int_0^{2\pi}
\frac{1+\cos(2\phi)}{2}
\,d\phi
$$

$$
E[\cos^2(\Phi)]
=
\frac{1}{4\pi}
\left[
\int_0^{2\pi}1\,d\phi
+
\int_0^{2\pi}\cos(2\phi)\,d\phi
\right]
$$

La primera integral es:

$$
\int_0^{2\pi}1\,d\phi=2\pi
$$

La segunda integral es:

$$
\int_0^{2\pi}\cos(2\phi)\,d\phi=0
$$

Por tanto:

$$
E[\cos^2(\Phi)]
=
\frac{1}{4\pi}
(2\pi)
$$

$$
E[\cos^2(\Phi)]
=
\frac{1}{2}
$$

En dB:

$$
10\log_{10}\left(\frac{1}{2}\right)
=
-3.01\,\text{dB}
$$

# Conclusión

Si la fase es completamente aleatoria en BSB/DSB, entonces:

$$
p_\Phi(\phi)=\frac{1}{2\pi}
$$

y la ganancia promedio de potencia es:

$$
G_P=\frac{1}{2}
$$

lo que equivale a una pérdida de aproximadamente:

$$
-3\,\text{dB}
$$

# Demostración para SSB con fase cero y fase aleatoria

En SSB, la señal modulada se construye usando la transformada de Hilbert del mensaje.

Sea el mensaje:

$$
m(t)
$$

y su transformada de Hilbert:

$$
\hat{m}(t)
$$

Para SSB superior, USB, la señal transmitida es:

$$
s_{USB}(t)=m(t)\cos(\omega_0 t)-\hat{m}(t)\sin(\omega_0 t)
$$

Para SSB inferior, LSB, la señal transmitida es:

$$
s_{LSB}(t)=m(t)\cos(\omega_0 t)+\hat{m}(t)\sin(\omega_0 t)
$$

---

# 1. Demodulación SSB con fase perfecta

Primero analizamos USB.

La señal recibida es:

$$
y(t)=s_{USB}(t)
$$

Por tanto:

$$
y(t)=m(t)\cos(\omega_0 t)-\hat{m}(t)\sin(\omega_0 t)
$$

El receptor multiplica por una portadora local:

$$
\cos(\omega_0 t+\phi)
$$

Para sincronización perfecta:

$$
\phi=0
$$

entonces:

$$
\cos(\omega_0 t+\phi)=\cos(\omega_0 t)
$$

La señal demodulada antes del filtro pasa bajas es:

$$
z(t)=y(t)\cos(\omega_0 t)
$$

Sustituyendo:

$$
z(t)=
\left[
m(t)\cos(\omega_0 t)-\hat{m}(t)\sin(\omega_0 t)
\right]
\cos(\omega_0 t)
$$

Distribuyendo:

$$
z(t)=
m(t)\cos^2(\omega_0 t)
-
\hat{m}(t)\sin(\omega_0 t)\cos(\omega_0 t)
$$

Usamos las identidades:

$$
\cos^2(\omega_0 t)=
\frac{1}{2}
+
\frac{1}{2}\cos(2\omega_0 t)
$$

$$
\sin(\omega_0 t)\cos(\omega_0 t)=
\frac{1}{2}\sin(2\omega_0 t)
$$

Entonces:

$$
z(t)=
m(t)
\left[
\frac{1}{2}
+
\frac{1}{2}\cos(2\omega_0 t)
\right]
-
\hat{m}(t)
\frac{1}{2}\sin(2\omega_0 t)
$$

Después del filtro pasa bajas, se eliminan las componentes en:

$$
2\omega_0
$$

Por tanto queda:

$$
z_{LPF}(t)=\frac{1}{2}m(t)
$$

Si se aplica una ganancia de recuperación igual a 2:

$$
\hat{m}(t)=2z_{LPF}(t)
$$

entonces:

$$
\hat{m}(t)=m(t)
$$

Por tanto, para fase perfecta:

$$
\boxed{\phi=0}
$$

$$
\boxed{z_{LPF}(t)=\frac{1}{2}m(t)}
$$

y con ganancia de 2:

$$
\boxed{\hat{m}(t)=m(t)}
$$

---

# 2. Demodulación SSB con error de fase

Ahora la portadora local tiene error de fase:

$$
\cos(\omega_0 t+\phi)
$$

Usando la identidad:

$$
\cos(\omega_0 t+\phi)
=
\cos(\omega_0 t)\cos(\phi)
-
\sin(\omega_0 t)\sin(\phi)
$$

La señal demodulada es:

$$
z(t)=s_{USB}(t)\cos(\omega_0 t+\phi)
$$

Sustituyendo:

$$
z(t)=
\left[
m(t)\cos(\omega_0 t)-\hat{m}(t)\sin(\omega_0 t)
\right]
\left[
\cos(\omega_0 t)\cos(\phi)
-
\sin(\omega_0 t)\sin(\phi)
\right]
$$

Distribuyendo:

$$
z(t)=
m(t)\cos^2(\omega_0 t)\cos(\phi)
-
m(t)\cos(\omega_0 t)\sin(\omega_0 t)\sin(\phi)
-
\hat{m}(t)\sin(\omega_0 t)\cos(\omega_0 t)\cos(\phi)
+
\hat{m}(t)\sin^2(\omega_0 t)\sin(\phi)
$$

Después del filtro pasa bajas:

$$
\cos^2(\omega_0 t)
\longrightarrow
\frac{1}{2}
$$

$$
\sin^2(\omega_0 t)
\longrightarrow
\frac{1}{2}
$$

$$
\sin(\omega_0 t)\cos(\omega_0 t)
\longrightarrow
0
$$

Entonces:

$$
z_{LPF}(t)=
\frac{1}{2}m(t)\cos(\phi)
+
\frac{1}{2}\hat{m}(t)\sin(\phi)
$$

Para USB:

$$
\boxed{
z_{LPF,USB}(t)=
\frac{1}{2}
\left[
m(t)\cos(\phi)
+
\hat{m}(t)\sin(\phi)
\right]
}
$$

Para LSB el signo cambia:

$$
\boxed{
z_{LPF,LSB}(t)=
\frac{1}{2}
\left[
m(t)\cos(\phi)
-
\hat{m}(t)\sin(\phi)
\right]
}
$$

---

# 3. Fase aleatoria en SSB

Si el receptor no conoce la fase, se modela como una variable aleatoria uniforme en un ciclo completo:

$$
0\leq \phi < 2\pi
$$

Como todos los valores de fase son igualmente probables:

$$
p_\Phi(\phi)=C
$$

La densidad debe cumplir:

$$
\int_0^{2\pi}p_\Phi(\phi)\,d\phi=1
$$

Sustituyendo:

$$
\int_0^{2\pi}C\,d\phi=1
$$

$$
C(2\pi)=1
$$

$$
C=\frac{1}{2\pi}
$$

Por tanto:

$$
\boxed{
p_\Phi(\phi)=\frac{1}{2\pi}
}
$$

---

# 4. Ganancia promedio con fase aleatoria

Para USB, la salida filtrada es:

$$
z_{LPF,USB}(t)=
\frac{1}{2}
\left[
m(t)\cos(\phi)
+
\hat{m}(t)\sin(\phi)
\right]
$$

La parte del mensaje original directamente recuperada es:

$$
\frac{1}{2}m(t)\cos(\phi)
$$

Por tanto, la ganancia de amplitud sobre el mensaje original es:

$$
G_A(\phi)=\cos(\phi)
$$

La ganancia de potencia asociada a esa componente es:

$$
G_P(\phi)=\cos^2(\phi)
$$

Como:

$$
p_\Phi(\phi)=\frac{1}{2\pi}
$$

entonces:

$$
E[\cos^2(\Phi)]
=
\int_0^{2\pi}
\cos^2(\phi)
\frac{1}{2\pi}
d\phi
$$

Usando:

$$
\cos^2(\phi)=\frac{1+\cos(2\phi)}{2}
$$

queda:

$$
E[\cos^2(\Phi)]
=
\frac{1}{2\pi}
\int_0^{2\pi}
\frac{1+\cos(2\phi)}{2}
d\phi
$$

$$
E[\cos^2(\Phi)]
=
\frac{1}{4\pi}
\left[
\int_0^{2\pi}1\,d\phi
+
\int_0^{2\pi}\cos(2\phi)\,d\phi
\right]
$$

La primera integral vale:

$$
\int_0^{2\pi}1\,d\phi=2\pi
$$

La segunda integral vale:

$$
\int_0^{2\pi}\cos(2\phi)\,d\phi=0
$$

Entonces:

$$
E[\cos^2(\Phi)]
=
\frac{1}{4\pi}
(2\pi)
$$

$$
E[\cos^2(\Phi)]
=
\frac{1}{2}
$$

En decibelios:

$$
10\log_{10}\left(\frac{1}{2}\right)
=
-3.01\,\text{dB}
$$

---

# 5. Diferencia importante entre DSB y SSB

En DSB, el error de fase produce simplemente:

$$
z_{LPF}(t)=\frac{1}{2}m(t)\cos(\phi)
$$

Por tanto, si:

$$
\phi=\frac{\pi}{2}
$$

la señal recuperada desaparece.

Pero en SSB ocurre algo diferente:

$$
z_{LPF,USB}(t)=
\frac{1}{2}
\left[
m(t)\cos(\phi)
+
\hat{m}(t)\sin(\phi)
\right]
$$

Es decir, cuando hay error de fase, no solo cambia la amplitud, sino que aparece una componente de la transformada de Hilbert del mensaje.

Por eso, en SSB el error de fase produce una mezcla entre:

$$
m(t)
$$

y:

$$
\hat{m}(t)
$$

Esto genera distorsión de fase en la señal recuperada.

---

# Conclusión

Para SSB con fase perfecta:

$$
\phi=0
$$

la salida es:

$$
z_{LPF}(t)=\frac{1}{2}m(t)
$$

y con ganancia de 2:

$$
\hat{m}(t)=m(t)
$$

Para SSB con fase aleatoria:

$$
p_\Phi(\phi)=\frac{1}{2\pi}
$$

la salida USB es:

$$
z_{LPF,USB}(t)=
\frac{1}{2}
\left[
m(t)\cos(\phi)
+
\hat{m}(t)\sin(\phi)
\right]
$$

y la salida LSB es:

$$
z_{LPF,LSB}(t)=
\frac{1}{2}
\left[
m(t)\cos(\phi)
-
\hat{m}(t)\sin(\phi)
\right]
$$

La componente útil directamente alineada con \(m(t)\) tiene ganancia promedio de potencia:

$$
\boxed{
G_P=\frac{1}{2}
}
$$

equivalente a:

$$
\boxed{
-3.01\,\text{dB}
}
$$

Sin embargo, en SSB esa pérdida aparece acompañada de una mezcla con la transformada de Hilbert del mensaje, lo cual se interpreta como distorsión de fase.

# Modelo de FM de banda ancha en canal gaussiano aditivo

En modulación FM, la información del mensaje no se coloca directamente en la amplitud de la portadora, sino en su frecuencia instantánea.

La señal FM se escribe como:

$$
s_{FM}(t)=A_c\cos\left(\omega_c t+k_f\int_{-\infty}^{t}m(\tau)d\tau\right)
$$

donde:

$$
A_c
$$

es la amplitud de la portadora,

$$
\omega_c=2\pi f_c
$$

es la frecuencia angular de la portadora,

$$
m(t)
$$

es la señal mensaje,

y

$$
k_f
$$

es la sensibilidad de frecuencia del modulador.

---

# Frecuencia instantánea

La fase instantánea de la señal FM es:

$$
\theta(t)=\omega_c t+k_f\int_{-\infty}^{t}m(\tau)d\tau
$$

La frecuencia angular instantánea se obtiene derivando la fase:

$$
\omega_i(t)=\frac{d\theta(t)}{dt}
$$

Entonces:

$$
\omega_i(t)=\omega_c+k_f m(t)
$$

En Hz:

$$
f_i(t)=f_c+\frac{k_f}{2\pi}m(t)
$$

Esto muestra que el mensaje controla la desviación instantánea de frecuencia.

---

# Desviación máxima de frecuencia

Si el mensaje tiene amplitud máxima:

$$
|m(t)|_{\max}=A_m
$$

entonces la desviación máxima de frecuencia es:

$$
\Delta f=\frac{k_f A_m}{2\pi}
$$

---

# Índice de modulación FM

Para una señal mensaje senoidal:

$$
m(t)=A_m\cos(\omega_m t)
$$

la señal FM queda:

$$
s_{FM}(t)=A_c\cos\left(\omega_c t+\beta\sin(\omega_m t)\right)
$$

donde:

$$
\beta=\frac{\Delta f}{f_m}
$$

es el índice de modulación FM.

---

# FM de banda angosta y banda ancha

Si:

$$
\beta \ll 1
$$

la señal es FM de banda angosta.

Pero si:

$$
\beta>1
$$

la señal se considera FM de banda ancha.

En FM de banda ancha aparecen muchas bandas laterales alrededor de la portadora. Por eso el ancho de banda teórico es infinito, pero en la práctica se usa la regla de Carson:

$$
B_T\approx 2(\Delta f+f_m)
$$

Como:

$$
\Delta f=\beta f_m
$$

entonces:

$$
B_T\approx 2f_m(\beta+1)
$$

---

# Canal gaussiano aditivo

En un canal AWGN, la señal recibida se modela como:

$$
r(t)=s_{FM}(t)+n(t)
$$

donde:

$$
n(t)\sim \mathcal{N}(0,\sigma_n^2)
$$

es ruido gaussiano aditivo de media cero y varianza:

$$
\sigma_n^2
$$

El ruido afecta la señal recibida, pero en FM la información está principalmente en las variaciones de frecuencia, no en las variaciones de amplitud.

---

# Modelo con ruido en fase y cuadratura

Una señal paso banda con ruido puede escribirse como:

$$
r(t)=A_c\cos(\theta(t))+n(t)
$$

donde:

$$
\theta(t)=\omega_c t+k_f\int_{-\infty}^{t}m(\tau)d\tau
$$

El ruido paso banda puede descomponerse como:

$$
n(t)=n_I(t)\cos(\omega_c t)-n_Q(t)\sin(\omega_c t)
$$

Por tanto:

$$
r(t)=A_c\cos(\theta(t))+n_I(t)\cos(\omega_c t)-n_Q(t)\sin(\omega_c t)
$$

donde:

$$
n_I(t),n_Q(t)\sim \mathcal{N}(0,\sigma_n^2)
$$

---

# Demodulación FM

El demodulador FM busca recuperar la frecuencia instantánea. Como:

$$
\omega_i(t)=\frac{d\theta(t)}{dt}
$$

y:

$$
\omega_i(t)=\omega_c+k_fm(t)
$$

entonces el mensaje se obtiene como:

$$
m(t)=\frac{1}{k_f}\left[\omega_i(t)-\omega_c\right]
$$

Por eso un demodulador FM ideal realiza una operación equivalente a estimar la fase instantánea y derivarla.

---

# Índice de demodulación

En el proceso de demodulación, puede definirse una ganancia o índice de demodulación como:

$$
K_d=\frac{1}{k_f}
$$

De esta manera, la salida del demodulador puede escribirse como:

$$
\hat{m}(t)=K_d\left[\omega_i(t)-\omega_c\right]
$$

Si:

$$
K_d=\frac{1}{k_f}
$$

entonces:

$$
\hat{m}(t)=m(t)
$$

En otras palabras, el índice de demodulación permite compensar la sensibilidad del modulador FM.

---

# FM con ruido después de demodulación

Cuando hay ruido, la fase instantánea recibida puede escribirse aproximadamente como:

$$
\theta_r(t)=\theta(t)+\phi_n(t)
$$

donde:

$$
\phi_n(t)
$$

representa la perturbación de fase causada por el ruido.

Entonces:

$$
\omega_r(t)=\frac{d\theta_r(t)}{dt}
$$

$$
\omega_r(t)=\frac{d}{dt}\left[\theta(t)+\phi_n(t)\right]
$$

$$
\omega_r(t)=\omega_c+k_fm(t)+\frac{d\phi_n(t)}{dt}
$$

Al demodular:

$$
\hat{m}(t)=K_d\left[\omega_r(t)-\omega_c\right]
$$

Sustituyendo:

$$
\hat{m}(t)=K_d\left[k_fm(t)+\frac{d\phi_n(t)}{dt}\right]
$$

Si:

$$
K_d=\frac{1}{k_f}
$$

queda:

$$
\hat{m}(t)=m(t)+\frac{1}{k_f}\frac{d\phi_n(t)}{dt}
$$

Esto muestra que el ruido en FM aparece principalmente como ruido de fase derivado.

---

# Efecto del índice de modulación

El índice de modulación FM es:

$$
\beta=\frac{\Delta f}{f_m}
$$

Si aumenta:

$$
\beta
$$

entonces aumenta la desviación de frecuencia:

$$
\Delta f
$$

La señal ocupa más ancho de banda:

$$
B_T\approx 2f_m(\beta+1)
$$

pero también mejora la robustez frente al ruido, porque la información queda más separada en frecuencia.

En FM de banda ancha:

$$
\beta>1
$$

la señal presenta mejor comportamiento frente al ruido que AM, aunque necesita más ancho de banda.

---

# Conclusión

El modelo de FM de banda ancha en canal gaussiano aditivo es:

$$
r(t)=A_c\cos\left(\omega_c t+k_f\int_{-\infty}^{t}m(\tau)d\tau\right)+n(t)
$$

con:

$$
n(t)\sim \mathcal{N}(0,\sigma_n^2)
$$

El índice de modulación es:

$$
\beta=\frac{\Delta f}{f_m}
$$

y para FM de banda ancha:

$$
\beta>1
$$

El demodulador estima la frecuencia instantánea:

$$
\omega_i(t)=\frac{d\theta(t)}{dt}
$$

y recupera el mensaje mediante:

$$
\hat{m}(t)=K_d\left[\omega_i(t)-\omega_c\right]
$$

Si:

$$
K_d=\frac{1}{k_f}
$$

entonces idealmente:

$$
\hat{m}(t)=m(t)
$$

En presencia de ruido:

$$
\hat{m}(t)=m(t)+\frac{1}{k_f}\frac{d\phi_n(t)}{dt}
$$

lo cual muestra que el ruido en FM afecta principalmente la fase instantánea y, después de la derivación, aparece como ruido en la señal demodulada.

# Demodulador FM por bloques

El demodulador FM por discriminación de frecuencia puede representarse con la siguiente cadena:

$$
\text{Limitador de nivel}
\rightarrow
\text{Derivador}
\rightarrow
\text{Detector de amplitud}
\rightarrow
\text{Filtro pasa bajos}
$$

---

# 1. Limitador de nivel

La señal FM recibida puede tener variaciones de amplitud debido al ruido del canal:

$$
r(t)=s_{FM}(t)+n(t)
$$

El limitador de nivel elimina o reduce esas variaciones de amplitud y deja una señal de amplitud aproximadamente constante:

$$
r_L(t)=A_L\cos(\theta(t))
$$

## ¿Para qué sirve?

Sirve para eliminar ruido de amplitud, porque en FM la información no está en la amplitud, sino en la frecuencia instantánea.

---

# 2. Derivador

El derivador calcula la derivada de la señal limitada:

$$
r_D(t)=\frac{d}{dt}\left[A_L\cos(\theta(t))\right]
$$

Aplicando regla de la cadena:

$$
r_D(t)=-A_L\sin(\theta(t))\frac{d\theta(t)}{dt}
$$

Como:

$$
\frac{d\theta(t)}{dt}=\omega_i(t)
$$

entonces:

$$
r_D(t)=-A_L\omega_i(t)\sin(\theta(t))
$$

La frecuencia instantánea en FM es:

$$
\omega_i(t)=\omega_c+k_fm(t)
$$

por tanto:

$$
r_D(t)=-A_L[\omega_c+k_fm(t)]\sin(\theta(t))
$$

## ¿Para qué sirve?

Sirve para convertir las variaciones de frecuencia de la señal FM en variaciones de amplitud.

---

# 3. Detector de amplitud

Después del derivador, la amplitud de la señal es:

$$
A(t)=A_L[\omega_c+k_fm(t)]
$$

El detector de amplitud extrae esa envolvente:

$$
E(t)=A_L[\omega_c+k_fm(t)]
$$

Separando términos:

$$
E(t)=A_L\omega_c+A_Lk_fm(t)
$$

## ¿Para qué sirve?

Sirve para extraer la amplitud variable generada por el derivador. Esa amplitud contiene el mensaje original.

---

# 4. Filtro pasa bajos

El filtro pasa bajos elimina componentes de alta frecuencia y conserva la parte lenta asociada al mensaje.

A partir de:

$$
E(t)=A_L\omega_c+A_Lk_fm(t)
$$

se elimina la componente constante:

$$
A_L\omega_c
$$

y queda:

$$
E_{AC}(t)=A_Lk_fm(t)
$$

Luego se aplica una ganancia:

$$
K_d=\frac{1}{A_Lk_f}
$$

Entonces:

$$
\hat{m}(t)=K_dE_{AC}(t)
$$

$$
\hat{m}(t)=m(t)
$$

## ¿Para qué sirve?

Sirve para dejar solo el mensaje recuperado y eliminar componentes no deseadas.

---

# Resumen

El proceso completo es:

$$
r(t)
\rightarrow
r_L(t)
\rightarrow
r_D(t)
\rightarrow
E(t)
\rightarrow
\hat{m}(t)
$$

donde:

$$
r_L(t)=A_L\cos(\theta(t))
$$

$$
r_D(t)=-A_L[\omega_c+k_fm(t)]\sin(\theta(t))
$$

$$
E(t)=A_L\omega_c+A_Lk_fm(t)
$$

$$
\hat{m}(t)=m(t)
$$

Por eso el demodulador FM funciona convirtiendo primero las variaciones de frecuencia en variaciones de amplitud, y luego detectando esa amplitud para recuperar el mensaje.

# Demodulación de FM de banda ancha usando limitador, derivador, detector de amplitud y pasa bajos

La señal FM de banda ancha se modela como:

$$
s_{FM}(t)=A_c\cos(\theta(t))
$$

donde:

$$
\theta(t)=\omega_c t+k_f\int m(\tau)d\tau
$$

La frecuencia instantánea es:

$$
\omega_i(t)=\frac{d\theta(t)}{dt}
$$

Por tanto:

$$
\omega_i(t)=\omega_c+k_fm(t)
$$

En FM de banda ancha se cumple:

$$
\beta=\frac{\Delta f}{f_m}>1
$$

---

# Bloques del demodulador FM

El demodulador se implementa con:

$$
\text{Limitador de nivel}
\rightarrow
\text{Derivador}
\rightarrow
\text{Detector de amplitud}
\rightarrow
\text{Filtro pasa bajos}
$$

---

# 1. Limitador de nivel

La señal recibida en un canal con ruido es:

$$
r(t)=s_{FM}(t)+n(t)
$$

El limitador elimina variaciones de amplitud causadas por el ruido:

$$
r_L(t)=A_L\cos(\theta(t))
$$

Sirve porque en FM la información no está en la amplitud, sino en la frecuencia instantánea.

---

# 2. Derivador

Se deriva la señal limitada:

$$
r_D(t)=\frac{d}{dt}\left[A_L\cos(\theta(t))\right]
$$

Aplicando regla de la cadena:

$$
r_D(t)=-A_L\sin(\theta(t))\frac{d\theta(t)}{dt}
$$

Como:

$$
\frac{d\theta(t)}{dt}=\omega_i(t)
$$

entonces:

$$
r_D(t)=-A_L\omega_i(t)\sin(\theta(t))
$$

Sustituyendo:

$$
\omega_i(t)=\omega_c+k_fm(t)
$$

queda:

$$
r_D(t)=-A_L[\omega_c+k_fm(t)]\sin(\theta(t))
$$

El derivador convierte las variaciones de frecuencia en variaciones de amplitud.

---

# 3. Detector de amplitud

La amplitud de la señal derivada es:

$$
A_D(t)=A_L[\omega_c+k_fm(t)]
$$

El detector de amplitud extrae esa envolvente:

$$
E(t)=A_L[\omega_c+k_fm(t)]
$$

Nota: en la práctica el detector responde a la envolvente absoluta,
$$
E(t)=A_L\left|\omega_c+k_fm(t)\right|,
$$
pero con desvíos moderados y una portadora alta el término dentro del valor absoluto es positivo y se conserva la señal de mensaje.

Separando:

$$
E(t)=A_L\omega_c+A_Lk_fm(t)
$$

---

# 4. Filtro pasa bajos

El filtro pasa bajos deja la componente lenta asociada al mensaje.

Luego se elimina la componente DC:

$$
A_L\omega_c
$$

y queda:

$$
E_{AC}(t)=A_Lk_fm(t)
$$

Finalmente se aplica una ganancia:

$$
K_d=\frac{1}{A_Lk_f}
$$

Por tanto:

$$
\hat{m}(t)=K_dE_{AC}(t)
$$

$$
\hat{m}(t)=m(t)
$$

---

# Conclusión

El demodulador FM por bloques recupera el mensaje porque el derivador transforma la variación de frecuencia en variación de amplitud.

Después, el detector de amplitud extrae esa variación y el filtro pasa bajos deja solamente el mensaje.

El proceso completo es:

$$
r(t)
\rightarrow
r_L(t)
\rightarrow
r_D(t)
\rightarrow
E(t)
\rightarrow
\hat{m}(t)
$$

---

# Cálculo teórico de la relación señal a ruido de salida

Ahora se analiza el mismo demodulador FM, pero incluyendo el efecto del ruido en el canal.

La señal recibida en un canal gaussiano aditivo es:

$$
r(t)=s_{FM}(t)+n(t)
$$

Para una señal paso banda, el ruido puede descomponerse en dos componentes:

$$
n(t)=n_I(t)\cos(\omega_c t)-n_Q(t)\sin(\omega_c t)
$$

donde:

$$
n_I(t)
$$

es la componente de ruido en fase, y:

$$
n_Q(t)
$$

es la componente de ruido en cuadratura.

En FM, después del limitador de nivel, las variaciones de amplitud se reducen. Por eso, el ruido que más afecta la demodulación es el ruido de cuadratura, porque produce perturbaciones en la fase instantánea.

Para alta relación señal a ruido, la perturbación de fase puede aproximarse como:

$$
\phi_n(t)\approx \frac{n_Q(t)}{A_c}
$$

Por tanto, la señal después del limitador se puede escribir como:

$$
r_L(t)=A_L\cos\left[\theta(t)+\phi_n(t)\right]
$$

---

# 1. Salida del derivador con ruido

El derivador actúa sobre:

$$
r_L(t)=A_L\cos\left[\theta(t)+\phi_n(t)\right]
$$

Entonces:

$$
r_D(t)=\frac{d}{dt}
\left\{
A_L\cos\left[\theta(t)+\phi_n(t)\right]
\right\}
$$

Aplicando la regla de la cadena:

$$
r_D(t)=
-A_L
\sin\left[\theta(t)+\phi_n(t)\right]
\frac{d}{dt}
\left[
\theta(t)+\phi_n(t)
\right]
$$

Separando la derivada:

$$
r_D(t)=
-A_L
\sin\left[\theta(t)+\phi_n(t)\right]
\left[
\frac{d\theta(t)}{dt}
+
\frac{d\phi_n(t)}{dt}
\right]
$$

Como:

$$
\frac{d\theta(t)}{dt}
=
\omega_c+k_fm(t)
$$

entonces:

$$
r_D(t)=
-A_L
\sin\left[\theta(t)+\phi_n(t)\right]
\left[
\omega_c+k_fm(t)+\frac{d\phi_n(t)}{dt}
\right]
$$

---

# 2. Salida del detector de amplitud

El detector de amplitud extrae la envolvente de la señal derivada.

Por tanto:

$$
E(t)=
A_L
\left[
\omega_c+k_fm(t)+\frac{d\phi_n(t)}{dt}
\right]
$$

Distribuyendo:

$$
E(t)=
A_L\omega_c
+
A_Lk_fm(t)
+
A_L\frac{d\phi_n(t)}{dt}
$$

En esta expresión aparecen tres términos:

Componente continua:

$$
A_L\omega_c
$$

Componente útil del mensaje:

$$
A_Lk_fm(t)
$$

Componente de ruido:

$$
A_L\frac{d\phi_n(t)}{dt}
$$

---

# 3. Salida del filtro pasa bajos y normalización

El filtro pasa bajos deja pasar la banda del mensaje y elimina componentes no deseadas.

Además, se elimina la componente DC:

$$
A_L\omega_c
$$

Por tanto, antes de normalizar queda:

$$
y_o(t)=A_Lk_fm(t)+A_L\frac{d\phi_n(t)}{dt}
$$

Ahora se aplica la ganancia del detector:

$$
K_d=\frac{1}{A_Lk_f}
$$

Entonces:

$$
\hat{m}(t)=K_d y_o(t)
$$

Sustituyendo:

$$
\hat{m}(t)=
\frac{1}{A_Lk_f}
\left[
A_Lk_fm(t)+A_L\frac{d\phi_n(t)}{dt}
\right]
$$

Simplificando:

$$
\hat{m}(t)=
m(t)
+
\frac{1}{k_f}
\frac{d\phi_n(t)}{dt}
$$

Por tanto, la salida del detector FM se puede escribir como:

$$
\hat{m}(t)=m(t)+n_o(t)
$$

donde el ruido de salida es:

$$
n_o(t)=
\frac{1}{k_f}
\frac{d\phi_n(t)}{dt}
$$

Como:

$$
\phi_n(t)\approx \frac{n_Q(t)}{A_c}
$$

entonces:

$$
n_o(t)=
\frac{1}{k_fA_c}
\frac{dn_Q(t)}{dt}
$$

---

# 4. Densidad espectral del ruido de salida

Derivar una señal en el tiempo equivale a multiplicar su transformada por:

$$
j2\pi f
$$

Por tanto:

$$
N_o(f)=
\frac{j2\pi f}{k_fA_c}
N_Q(f)
$$

La densidad espectral de potencia del ruido de salida es:

$$
S_{n_o}(f)=
\left|
\frac{j2\pi f}{k_fA_c}
\right|^2
S_{n_Q}(f)
$$

Entonces:

$$
S_{n_o}(f)=
\frac{(2\pi f)^2}{k_f^2A_c^2}
S_{n_Q}(f)
$$

Si se usa la convención de frecuencia angular, también puede escribirse como:

$$
S_{n_o}(\omega)=
\frac{\omega^2}{k_f^2A_c^2}
S_{n_Q}(\omega)
$$

Esto muestra que el ruido a la salida del detector FM no es plano, sino que crece con la frecuencia:

$$
S_{n_o}(f)\propto f^2
$$

Por eso en FM el ruido demodulado afecta más las frecuencias altas del mensaje.

---

# 5. Potencia de ruido de salida

Si el mensaje está limitado a un ancho de banda:

$$
B_m
$$

y se usa la forma práctica en Hz, la potencia de ruido de salida se obtiene integrando la PSD en la banda del mensaje:

$$
N_o=
\int_0^{B_m}
S_{n_o}(f)\,df
$$

Para el análisis clásico del detector FM se obtiene:

$$
N_o=
\frac{N_0B_m^3}{3k_f^2A_c^2}
$$

Este resultado aparece porque:

$$
\int_0^{B_m}f^2\,df
=
\frac{B_m^3}{3}
$$

---

# 6. Potencia de señal de salida

Después de la normalización, la componente útil recuperada es:

$$
m(t)
$$

Por tanto, la potencia de señal de salida es:

$$
S_o=P_m
$$

Si el mensaje es senoidal:

$$
m(t)=A_m\cos(2\pi f_m t)
$$

entonces su potencia promedio es:

$$
P_m=\frac{A_m^2}{2}
$$

Por tanto:

$$
S_o=\frac{A_m^2}{2}
$$

---

# 7. Relación señal a ruido de salida

La relación señal a ruido de salida se define como:

$$
\left(\frac{S}{N}\right)_o
=
\frac{S_o}{N_o}
$$

Sustituyendo:

$$
\left(\frac{S}{N}\right)_o
=
\frac{
\frac{A_m^2}{2}
}{
\frac{N_0B_m^3}{3k_f^2A_c^2}
}
$$

Resolviendo:

$$
\left(\frac{S}{N}\right)_o
=
\frac{3k_f^2A_c^2A_m^2}{2N_0B_m^3}
$$

Como la desviación máxima de frecuencia es:

$$
\Delta f=k_fA_m
$$

entonces:

$$
k_f^2A_m^2=\Delta f^2
$$

Por tanto:

$$
\left(\frac{S}{N}\right)_o
=
\frac{3A_c^2\Delta f^2}{2N_0B_m^3}
$$

La potencia de la portadora es:

$$
C=\frac{A_c^2}{2}
$$

por lo tanto:

$$
A_c^2=2C
$$

Sustituyendo:

$$
\left(\frac{S}{N}\right)_o
=
\frac{3(2C)\Delta f^2}{2N_0B_m^3}
$$

Finalmente:

$$
\left(\frac{S}{N}\right)_o
=
\frac{3C\Delta f^2}{N_0B_m^3}
$$

---

# 8. Relación con el índice de modulación

El índice de modulación FM de banda ancha se define como:

$$
\beta=\frac{\Delta f}{B_m}
$$

De ahí:

$$
\Delta f=\beta B_m
$$

Sustituyendo:

$$
\left(\frac{S}{N}\right)_o
=
\frac{3C(\beta B_m)^2}{N_0B_m^3}
$$

$$
\left(\frac{S}{N}\right)_o
=
\frac{3C\beta^2B_m^2}{N_0B_m^3}
$$

Simplificando:

$$
\left(\frac{S}{N}\right)_o
=
3\beta^2
\frac{C}{N_0B_m}
$$

Como la relación portadora a ruido de entrada es:

$$
\left(\frac{C}{N}\right)_i
=
\frac{C}{N_0B_m}
$$

entonces:

$$
\left(\frac{S}{N}\right)_o
=
3\beta^2
\left(\frac{C}{N}\right)_i
$$

---

# 9. Ganancia de proceso

La ganancia de proceso se define como:

$$
G_p=
\frac{
\left(\frac{S}{N}\right)_o
}{
\left(\frac{C}{N}\right)_i
}
$$

Sustituyendo:

$$
G_p=
\frac{
3\beta^2
\left(\frac{C}{N}\right)_i
}{
\left(\frac{C}{N}\right)_i
}
$$

Por tanto:

$$
G_p=3\beta^2
$$

En decibelios:

$$
G_{p,dB}=10\log_{10}(3\beta^2)
$$

También puede escribirse como:

$$
G_{p,dB}=10\log_{10}(3)+20\log_{10}(\beta)
$$

Como:

$$
10\log_{10}(3)\approx 4.77
$$

entonces:

$$
G_{p,dB}=4.77+20\log_{10}(\beta)
$$

---

# Conclusión final

El demodulador FM de banda ancha usando los bloques:

$$
\text{Limitador de nivel}
\rightarrow
\text{Derivador}
\rightarrow
\text{Detector de amplitud}
\rightarrow
\text{Filtro pasa bajos}
$$

permite recuperar el mensaje porque el derivador transforma las variaciones de frecuencia en variaciones de amplitud.

La salida normalizada del detector es:

$$
\hat{m}(t)=m(t)+n_o(t)
$$

donde:

$$
n_o(t)=
\frac{1}{k_fA_c}
\frac{dn_Q(t)}{dt}
$$

La relación señal a ruido de salida es:

$$
\left(\frac{S}{N}\right)_o
=
3\beta^2
\left(\frac{C}{N}\right)_i
$$

y la ganancia de proceso es:

$$
G_p=3\beta^2
$$

Como en FM de banda ancha:

$$
\beta>1
$$

entonces la ganancia de proceso puede ser considerablemente mayor que uno. Esto explica por qué FM de banda ancha mejora la relación señal a ruido a costa de ocupar mayor ancho de banda.

```python
import torch
import matplotlib.pyplot as plt

# ============================================================
# 1. PARÁMETROS GENERALES
# ============================================================

torch.manual_seed(10)

fs = 50000          # Frecuencia de muestreo [Hz]
T = 0.2             # Duración [s]
N = int(fs * T)

t = torch.arange(N) / fs

fm = 50.0           # Frecuencia del mensaje [Hz]
fc = 3000.0         # Frecuencia portadora [Hz]

Am = 1.0            # Amplitud del mensaje
Ac = 1.0            # Amplitud de la portadora

SNR_dB = 15.0       # SNR del canal

# Dos casos para comparar:
# beta pequeño: menor excursión, menor ancho de banda, menor fidelidad
# beta grande: mayor excursión, mayor ancho de banda, mayor fidelidad
betas = [1.0, 5.0]

# Mensaje original
m = Am * torch.cos(2 * torch.pi * fm * t)

# ============================================================
# 2. FUNCIONES AUXILIARES
# ============================================================

def analytic_signal(x):
    """
    Calcula la señal analítica usando FFT.
    Equivalente a usar transformada de Hilbert.
    """
    X = torch.fft.fft(x)
    H = torch.zeros_like(X)

    n = x.numel()

    if n % 2 == 0:
        H[0] = 1
        H[n // 2] = 1
        H[1:n // 2] = 2
    else:
        H[0] = 1
        H[1:(n + 1) // 2] = 2

    return torch.fft.ifft(X * H)


def lpf_fft(x, cutoff, fs):
    """
    Filtro pasa bajos ideal usando FFT.
    """
    X = torch.fft.fft(x)
    f = torch.fft.fftfreq(x.numel(), d=1/fs)

    H = torch.abs(f) <= cutoff

    y = torch.fft.ifft(X * H)

    return torch.real(y)


def add_awgn(x, SNR_dB):
    """
    Agrega ruido gaussiano aditivo con una SNR dada.
    """
    Ps = torch.mean(x**2)
    SNR = 10**(SNR_dB / 10)
    Pn = Ps / SNR

    n = torch.sqrt(Pn) * torch.randn_like(x)

    return x + n


def spectrum_db(x, fs):
    """
    Espectro normalizado en dB.
    """
    window = torch.hann_window(x.numel())
    X = torch.fft.fftshift(torch.fft.fft(x * window))
    f = torch.fft.fftshift(torch.fft.fftfreq(x.numel(), d=1/fs))

    mag = torch.abs(X)
    mag = mag / torch.max(mag)

    mag_db = 20 * torch.log10(mag + 1e-12)

    return f, mag_db


def fm_modulate(beta):
    """
    Genera una señal FM:
    
    s_FM(t) = Ac cos(2 pi fc t + beta sin(2 pi fm t))
    
    Para mensaje cosenoidal:
    
    beta = Delta_f / fm
    """
    delta_f = beta * fm
    kf = delta_f / Am

    theta = 2 * torch.pi * fc * t + beta * torch.sin(2 * torch.pi * fm * t)

    s_fm = Ac * torch.cos(theta)

    f_inst = fc + delta_f * torch.cos(2 * torch.pi * fm * t)

    return s_fm, f_inst, delta_f, kf


def fm_demod_blocks(r, kf):
    """
    Demodulador FM usando exactamente los bloques:
    
    1. Limitador de nivel
    2. Derivador
    3. Detector de amplitud
    4. Filtro pasa bajos
    """

    # ========================================================
    # BLOQUE 1: LIMITADOR DE NIVEL
    # ========================================================
    # Se obtiene la señal analítica para conservar fase
    z = analytic_signal(r)

    # Se normaliza la amplitud y se conserva la fase
    z_limitada = z / (torch.abs(z) + 1e-12)

    # Señal limitada real
    r_L = torch.real(z_limitada)

    # ========================================================
    # BLOQUE 2: DERIVADOR
    # ========================================================
    r_D = torch.zeros_like(r_L)
    r_D[1:] = (r_L[1:] - r_L[:-1]) * fs

    # ========================================================
    # BLOQUE 3: DETECTOR DE AMPLITUD
    # ========================================================
    envolvente = torch.abs(analytic_signal(r_D))

    # ========================================================
    # BLOQUE 4: FILTRO PASA BAJOS
    # ========================================================
    cutoff = 2.5 * fm
    envolvente_lpf = lpf_fft(envolvente, cutoff, fs)

    # Se elimina la componente DC
    envolvente_ac = envolvente_lpf - torch.mean(envolvente_lpf)

    # Ganancia del demodulador
    # Después del derivador y detector:
    # E_AC(t) ≈ 2 pi kf m(t)
    m_rec = envolvente_ac / (2 * torch.pi * kf)

    return m_rec, r_L, r_D, envolvente, envolvente_lpf


# ============================================================
# 3. SIMULACIÓN DE LOS DOS CASOS
# ============================================================

resultados = {}

for beta in betas:

    s_fm, f_inst, delta_f, kf = fm_modulate(beta)

    r = add_awgn(s_fm, SNR_dB)

    m_rec, r_L, r_D, envolvente, envolvente_lpf = fm_demod_blocks(r, kf)

    BT_carson = 2 * (delta_f + fm)

    Gp = 3 * beta**2
    Gp_dB = 10 * torch.log10(torch.tensor(Gp))

    # Cálculo experimental de SNR de salida
    ini = int(0.02 * fs)
    fin = int(0.18 * fs)

    error = m_rec[ini:fin] - m[ini:fin]

    S_out = torch.mean(m[ini:fin]**2)
    N_out = torch.mean(error**2)

    SNR_out_dB = 10 * torch.log10(S_out / N_out)

    resultados[beta] = {
        "s_fm": s_fm,
        "r": r,
        "m_rec": m_rec,
        "f_inst": f_inst,
        "delta_f": delta_f,
        "kf": kf,
        "BT_carson": BT_carson,
        "Gp": Gp,
        "Gp_dB": Gp_dB.item(),
        "SNR_out_dB": SNR_out_dB.item()
    }


# ============================================================
# 4. IMPRIMIR RESULTADOS TEÓRICOS
# ============================================================

print("============================================================")
print("COMPARACIÓN ENTRE DOS EXCURSIONES DE FRECUENCIA")
print("============================================================")

for beta in betas:
    data = resultados[beta]

    print(f"\nCaso beta = {beta}")
    print(f"Delta f = {data['delta_f']:.2f} Hz")
    print(f"Ancho de banda por Carson = {data['BT_carson']:.2f} Hz")
    print(f"Ganancia de proceso Gp = 3 beta^2 = {data['Gp']:.2f}")
    print(f"Ganancia de proceso en dB = {data['Gp_dB']:.2f} dB")
    print(f"SNR de salida experimental = {data['SNR_out_dB']:.2f} dB")


# ============================================================
# 5. GRÁFICA 1: EXCURSIÓN DE FRECUENCIA
# ============================================================

plt.figure(figsize=(11, 5))

for beta in betas:
    plt.plot(
        t,
        resultados[beta]["f_inst"] - fc,
        label=fr"$\beta={beta}$, $\Delta f={resultados[beta]['delta_f']:.0f}$ Hz"
    )

plt.xlim(0, 0.06)
plt.title("Excursión de frecuencia: mayor beta implica mayor desplazamiento alrededor de fc")
plt.xlabel("Tiempo [s]")
plt.ylabel(r"$f_i(t)-f_c$ [Hz]")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()


# ============================================================
# 6. GRÁFICA 2: ESPECTRO DE LAS SEÑALES FM
# ============================================================

plt.figure(figsize=(11, 5))

for beta in betas:
    f, S_db = spectrum_db(resultados[beta]["s_fm"], fs)
    plt.plot(f, S_db, label=fr"$\beta={beta}$")

plt.xlim(fc - 800, fc + 800)
plt.ylim(-80, 5)
plt.title("Espectro de FM: mayor excursión implica mayor ancho de banda")
plt.xlabel("Frecuencia [Hz]")
plt.ylabel("Magnitud normalizada [dB]")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()


# ============================================================
# 7. GRÁFICA 3: MENSAJE RECUPERADO
# ============================================================

plt.figure(figsize=(11, 7))

plt.subplot(2, 1, 1)
plt.plot(t, m, label="Mensaje original")
plt.plot(t, resultados[1.0]["m_rec"], "--", label="Mensaje recuperado")
plt.xlim(0, 0.08)
plt.title(r"Demodulación FM con menor excursión: $\beta=1$")
plt.ylabel("Amplitud")
plt.grid(True)
plt.legend()

plt.subplot(2, 1, 2)
plt.plot(t, m, label="Mensaje original")
plt.plot(t, resultados[5.0]["m_rec"], "--", label="Mensaje recuperado")
plt.xlim(0, 0.08)
plt.title(r"Demodulación FM con mayor excursión: $\beta=5$")
plt.xlabel("Tiempo [s]")
plt.ylabel("Amplitud")
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()


# ============================================================
# 8. GRÁFICA 4: ANCHO DE BANDA Y GANANCIA DE PROCESO
# ============================================================

BT_vals = [resultados[b]["BT_carson"] for b in betas]
Gp_vals = [resultados[b]["Gp_dB"] for b in betas]

plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.bar([str(b) for b in betas], BT_vals)
plt.title("Ancho de banda por Carson")
plt.xlabel(r"$\beta$")
plt.ylabel(r"$B_T$ [Hz]")
plt.grid(True, axis="y")

plt.subplot(1, 2, 2)
plt.bar([str(b) for b in betas], Gp_vals)
plt.title("Ganancia de proceso")
plt.xlabel(r"$\beta$")
plt.ylabel(r"$G_p$ [dB]")
plt.grid(True, axis="y")

plt.tight_layout()
plt.show()
```

# Explicación de las gráficas de FM de banda ancha

En la simulación se comparan dos casos de FM:

$$
\beta=1
$$

y

$$
\beta=5
$$

El objetivo es visualizar que al aumentar el índice de modulación FM aumenta la excursión de frecuencia, aumenta el ancho de banda ocupado y mejora la fidelidad frente al ruido.

La relación principal es:

$$
\beta=\frac{\Delta f}{f_m}
$$

donde:

$$
\Delta f
$$

es la excursión máxima de frecuencia, y:

$$
f_m
$$

es la frecuencia del mensaje.

---

# 1. Gráfica de excursión de frecuencia

La primera gráfica muestra:

$$
f_i(t)-f_c
$$

es decir, cuánto se separa la frecuencia instantánea de la portadora.

La frecuencia instantánea de una señal FM es:

$$
f_i(t)=f_c+\Delta f \cos(2\pi f_m t)
$$

Por tanto:

$$
f_i(t)-f_c=\Delta f \cos(2\pi f_m t)
$$

Cuando:

$$
\beta=1
$$

la excursión es menor:

$$
\Delta f=\beta f_m=1f_m
$$

Cuando:

$$
\beta=5
$$

la excursión es mayor:

$$
\Delta f=5f_m
$$

Por eso en la gráfica la curva de:

$$
\beta=5
$$

tiene mayor amplitud vertical.

## Interpretación física

Esta gráfica muestra que una FM con mayor índice de modulación mueve más su frecuencia instantánea alrededor de la portadora.

En palabras simples:

$$
\beta \uparrow
\Rightarrow
\Delta f \uparrow
$$

Esto significa que el mensaje queda representado mediante cambios de frecuencia más grandes.

---

# 2. Gráfica del espectro de la señal FM

La segunda gráfica muestra el espectro de la señal FM alrededor de la portadora:

$$
f_c
$$

Para FM, la señal no ocupa solamente una frecuencia, sino varias componentes laterales alrededor de la portadora.

Cuando aumenta:

$$
\beta
$$

aparecen más bandas laterales significativas.

Por eso, para:

$$
\beta=5
$$

el espectro es más ancho que para:

$$
\beta=1
$$

El ancho de banda práctico se aproxima con la regla de Carson:

$$
B_T\approx 2(\Delta f+f_m)
$$

Como:

$$
\Delta f=\beta f_m
$$

entonces:

$$
B_T\approx 2f_m(\beta+1)
$$

Por tanto, si aumenta:

$$
\beta
$$

también aumenta el ancho de banda:

$$
B_T
$$

## Interpretación física

Esta gráfica muestra el costo de aumentar la fidelidad en FM.

Mayor excursión de frecuencia significa que la señal ocupa más espacio en frecuencia.

En forma resumida:

$$
\beta \uparrow
\Rightarrow
\Delta f \uparrow
\Rightarrow
B_T \uparrow
$$

---

# 3. Gráfica del mensaje recuperado

La tercera gráfica compara el mensaje original con el mensaje recuperado después de demodular.

El demodulador usado es:

$$
\text{Limitador}
\rightarrow
\text{Derivador}
\rightarrow
\text{Detector de amplitud}
\rightarrow
\text{Pasa bajos}
$$

En el caso de menor excursión:

$$
\beta=1
$$

la señal recuperada es más sensible al ruido, porque los cambios de frecuencia son pequeños y el ruido puede ocultarlos con más facilidad.

En el caso de mayor excursión:

$$
\beta=5
$$

los cambios de frecuencia son más grandes, entonces el detector FM puede distinguir mejor el mensaje frente al ruido.

## Interpretación física

Esta gráfica muestra que con mayor índice de modulación la señal recuperada se parece más al mensaje original.

Esto representa mayor fidelidad.

La idea es:

$$
\beta \uparrow
\Rightarrow
G_p \uparrow
\Rightarrow
\text{mejor recuperación del mensaje}
$$

---

# 4. Gráfica de ancho de banda y ganancia de proceso

La última gráfica compara dos cantidades importantes:

## Ancho de banda por Carson

El ancho de banda práctico es:

$$
B_T\approx 2f_m(\beta+1)
$$

Por eso, cuando:

$$
\beta=5
$$

el ancho de banda es mayor que cuando:

$$
\beta=1
$$

Esto confirma que FM de banda ancha necesita más ancho de banda.

---

## Ganancia de proceso

La ganancia de proceso en FM de banda ancha es:

$$
G_p=3\beta^2
$$

En decibelios:

$$
G_{p,dB}=10\log_{10}(3\beta^2)
$$

Por tanto, al aumentar:

$$
\beta
$$

la ganancia de proceso aumenta cuadráticamente.

Si:

$$
\beta=1
$$

entonces:

$$
G_p=3(1)^2=3
$$

Si:

$$
\beta=5
$$

entonces:

$$
G_p=3(5)^2=75
$$

Esto muestra que el caso con:

$$
\beta=5
$$

tiene mucha mayor ganancia de proceso.

## Interpretación física

La ganancia de proceso indica cuánto mejora la relación señal a ruido después de demodular FM.

Entonces:

$$
\beta \uparrow
\Rightarrow
G_p \uparrow
$$

pero también:

$$
\beta \uparrow
\Rightarrow
B_T \uparrow
$$

---

# Conclusión general

Las gráficas muestran el principio fundamental de FM de banda ancha:

$$
\boxed{
\text{FM intercambia ancho de banda por fidelidad}
}
$$

Al aumentar el índice de modulación:

$$
\beta
$$

aumenta la excursión de frecuencia:

$$
\Delta f
$$

aumenta el ancho de banda ocupado:

$$
B_T
$$

y aumenta la ganancia de proceso:

$$
G_p
$$

Por tanto:

$$
\beta \uparrow
\Rightarrow
\Delta f \uparrow
\Rightarrow
B_T \uparrow
\Rightarrow
G_p \uparrow
\Rightarrow
\text{mayor fidelidad}
$$

Esto significa que una señal FM de banda ancha puede ser más resistente al ruido, pero necesita ocupar más ancho de banda en el canal.
