## Comparación entre serie de Taylor y expansión de Bessel

En el análisis espectral de una señal PM aparece el término no lineal:

$$
e^{jk_p x(t)}
$$

La forma de tratar este término depende de la condición de modulación.

| Método | Cuándo usarlo | Resultado principal |
|---|---|---|
| **Serie de Taylor** | Se usa cuando la desviación de fase es pequeña, es decir, cuando se cumple $|k_px(t)|\ll 1$. Esta es la condición de PM de banda angosta. | Permite aproximar $e^{jk_px(t)}\approx 1+jk_px(t)$. Con esta aproximación, el espectro de la señal PM queda relacionado directamente con el espectro de la señal moduladora $X(f)$. |
| **Serie de Fourier / Bessel** | Se usa cuando la señal moduladora es periódica. Para el caso particular de una moduladora senoidal, la expansión se realiza usando funciones de Bessel. Este análisis es necesario en PM de banda ancha. | Permite obtener las bandas laterales exactas ubicadas en $f_c\pm nf_m$, cuyas amplitudes están determinadas por las funciones de Bessel $J_n(\beta)$. |

En banda angosta, el término exponencial se aproxima mediante Taylor:

$$
e^{jk_p x(t)}
\approx
1+jk_p x(t)
$$

Esta aproximación solo es válida si:

$$
|k_px(t)|\ll 1
$$

Por tanto, el espectro contiene principalmente la portadora y dos bandas laterales principales.

En cambio, en banda ancha no se puede hacer esta aproximación. Si la señal moduladora es senoidal:

$$
x(t)=A_m\cos(\omega_m t)
$$

entonces:

$$
e^{jk_p x(t)}
=
e^{j\beta\cos(\omega_m t)}
$$

donde:

$$
\beta=k_pA_m
$$

En este caso se usa la expansión en funciones de Bessel:

$$
e^{j\beta\cos(\omega_m t)}
=
\sum_{n=-\infty}^{\infty}
j^nJ_n(\beta)e^{jn\omega_m t}
$$

Esta expresión muestra que la señal PM contiene componentes espectrales en:

$$
f=f_c+nf_m
$$

con:

$$
n=0,\pm1,\pm2,\pm3,\ldots
$$

La amplitud de cada componente está determinada por $J_n(\beta)$. Por eso, cuando $\beta$ aumenta, más bandas laterales tienen amplitud significativa y el espectro se ensancha.

# Ancho de banda en FM de banda ancha

La modulación en frecuencia se define como:

$$
s_{FM}(t)=A_c\cos\left(2\pi f_c t+2\pi k_f\int_{-\infty}^{t}x(\tau)d\tau\right)
$$

donde:

- $A_c$ es la amplitud de la portadora.
- $f_c$ es la frecuencia de la portadora.
- $x(t)$ es la señal moduladora.
- $k_f$ es la sensibilidad de frecuencia, en Hz por unidad de amplitud.
- $k_f x(t)$ representa la desviación instantánea de frecuencia respecto a $f_c$.

La fase instantánea de la señal FM es:

$$
\theta_i(t)=2\pi f_c t+2\pi k_f\int_{-\infty}^{t}x(\tau)d\tau
$$

La frecuencia instantánea se obtiene derivando la fase:

$$
f_i(t)=\frac{1}{2\pi}\frac{d\theta_i(t)}{dt}
$$

Entonces:

$$
f_i(t)=f_c+k_f x(t)
$$

Por tanto, en FM la desviación instantánea de frecuencia es:

$$
\Delta f(t)=k_f x(t)
$$

y la desviación máxima de frecuencia es:

$$
\Delta f = k_f \max|x(t)|
$$

# Por qué el ancho de banda teórico de FM es infinito

Aunque la señal moduladora $x(t)$ tenga ancho de banda limitado, la señal FM es una modulación angular no lineal. La señal FM contiene el término complejo:

$$
e^{j2\pi k_f\int x(t)dt}
$$

Si este término se expande mediante serie de Taylor:

$$
e^{j2\pi k_f\int x(t)dt}
=
1
+j2\pi k_f\int x(t)dt
+\frac{\left(j2\pi k_f\int x(t)dt\right)^2}{2!}
+\frac{\left(j2\pi k_f\int x(t)dt\right)^3}{3!}
+\cdots
$$

La serie tiene infinitos términos. Cada potencia genera nuevas componentes espectrales. Por esta razón, desde el punto de vista matemático ideal, el espectro de una señal FM puede tener infinitas bandas laterales.

Por tanto:

$$
B_{teórico}=\infty
$$

Sin embargo, en la práctica las bandas laterales lejanas tienen amplitudes muy pequeñas. Por eso se define un ancho de banda práctico, considerando solamente las componentes espectrales significativas.

# Caso senoidal y relación con funciones de Bessel

Si la señal moduladora es:

$$
x(t)=A_m\cos(2\pi f_m t)
$$

entonces:

$$
\int x(t)dt
=
\frac{A_m}{2\pi f_m}\sin(2\pi f_m t)
$$

Sustituyendo en la señal FM:

$$
s_{FM}(t)
=
A_c\cos\left(
2\pi f_c t+
2\pi k_f
\frac{A_m}{2\pi f_m}
\sin(2\pi f_m t)
\right)
$$

Simplificando:

$$
s_{FM}(t)
=
A_c\cos\left(
2\pi f_c t+
\beta\sin(2\pi f_m t)
\right)
$$

donde el índice de modulación FM es:

$$
\beta=\frac{k_f A_m}{f_m}
$$

Como:

$$
\Delta f=k_fA_m
$$

también se puede escribir:

$$
\beta=\frac{\Delta f}{f_m}
$$

Para este caso senoidal, el espectro se puede expresar usando funciones de Bessel:

$$
e^{j\beta\sin(2\pi f_m t)}
=
\sum_{n=-\infty}^{\infty}
J_n(\beta)e^{jn2\pi f_m t}
$$

Esto muestra que aparecen componentes en:

$$
f=f_c+n f_m
$$

con:

$$
n=0,\pm1,\pm2,\pm3,\ldots
$$

La serie es infinita, por eso el ancho de banda teórico es infinito. Pero en la práctica solo se consideran las bandas laterales cuyo valor $J_n(\beta)$ es significativo.

## Verificación experimental de la regla de Carson en FM

En modulación FM una señal senoidal se escribe como:

$$
s_{FM}(t) = A_c \cos\left(2\pi f_c t + \beta \sin\left(2\pi f_m t\right)\right)
$$

Aquí:
- $A_c$ es la amplitud de la portadora.
- $f_c$ es la frecuencia de la portadora.
- $f_m$ es la frecuencia de la señal moduladora.
- $\beta$ es el índice de modulación FM.

Para FM se define la desviación máxima de frecuencia como:

$$
\Delta f = \beta f_m
$$

La regla de Carson predice el ancho de banda práctico:

$$
B_T \approx 2(\Delta f + f_m) = 2 f_m (\beta + 1)
$$

Esto significa que, aunque el ancho de banda teórico de FM sea infinito (debido a las infinitas bandas laterales), en la práctica se puede usar un intervalo finito alrededor de $f_c$ donde se concentra la mayor parte de la potencia.

En este experimento usamos la forma canónica de FM con índice $\beta$, por lo que no necesitamos definir $k_f$ explícitamente: basta con construir la señal como $s_{FM}(t)=A_c\cos(2\pi f_c t + \beta \sin(2\pi f_m t))$.

El experimento siguiente implementa estos pasos:

1. Generar la señal FM para varios valores de $\beta$.
2. Calcular $\Delta f$, $B_T$, $f_c - B_T/2$ y $f_c + B_T/2$.
3. Calcular la FFT de la señal FM.
4. Calcular la potencia espectral $P(f)=|S(f)|^2$.
5. Calcular la potencia total en frecuencias positivas.
6. Calcular la potencia dentro de la banda de Carson.
7. Calcular el porcentaje de potencia dentro de Carson.
8. Mostrar una tabla de resultados y graficar cada espectro con los límites de Carson.

A continuación se ejecuta el experimento en Python para verificar la regla de Carson con datos numéricos.


```python
import numpy as np
import matplotlib.pyplot as plt

# ============================================================
# PARÁMETROS GENERALES
# ============================================================

fs = 200000          # Frecuencia de muestreo [Hz]
T = 0.1              # Duración de la simulación [s]
t = np.arange(0, T, 1/fs)

Ac = 1               # Amplitud de la portadora
fc = 20000           # Frecuencia portadora [Hz]
fm = 1000            # Frecuencia de la moduladora [Hz]

betas = [0.2, 1, 3, 5, 8]

# ============================================================
# FUNCIONES AUXILIARES
# ============================================================

def calcular_espectro(signal, fs):
    """Calcula la FFT y devuelve frecuencias y espectro complejo."""
    N = len(signal)
    S = np.fft.fft(signal)
    f = np.fft.fftfreq(N, d=1/fs)
    return f, S


def potencia_positiva(signal, fs):
    """Devuelve frecuencias positivas y potencia espectral para una señal real."""
    f, S = calcular_espectro(signal, fs)
    mask_pos = f >= 0
    f_pos = f[mask_pos]
    P_pos = np.abs(S[mask_pos])**2
    return f_pos, P_pos


def porcentaje_en_carson(signal, fs, fc, BT):
    """Calcula el porcentaje de potencia espectral positiva dentro de la banda de Carson."""
    f_pos, P_pos = potencia_positiva(signal, fs)
    f_low = fc - BT/2
    f_high = fc + BT/2
    mask_carson = (f_pos >= f_low) & (f_pos <= f_high)
    P_total = np.sum(P_pos)
    P_carson = np.sum(P_pos[mask_carson])
    porcentaje = 100 * P_carson / P_total if P_total > 0 else 0.0
    return porcentaje, f_pos, P_pos

# ============================================================
# EXPERIMENTO: VERIFICAR LA REGLA DE CARSON
# ============================================================

resultados = []

plt.figure(figsize=(12, 16))

for i, beta in enumerate(betas):
    Delta_f = beta * fm
    BT = 2 * (Delta_f + fm)
    f_low = fc - BT / 2
    f_high = fc + BT / 2

    s_fm = Ac * np.cos(2 * np.pi * fc * t + beta * np.sin(2 * np.pi * fm * t))
    porcentaje, f_pos, P_pos = porcentaje_en_carson(s_fm, fs, fc, BT)

    resultados.append({
        'beta': beta,
        'Delta_f': Delta_f,
        'B_T': BT,
        'f_low': f_low,
        'f_high': f_high,
        'porcentaje': porcentaje,
    })

    plt.subplot(len(betas), 1, i + 1)
    plt.plot(f_pos, P_pos, color='tab:blue', linewidth=0.9)
    plt.yscale('log')
    plt.axvline(fc, color='black', linestyle='-', linewidth=1.0, label='$f_c$')
    plt.axvline(f_low, color='red', linestyle='--', linewidth=1.0, label='$f_c - B_T/2$')
    plt.axvline(f_high, color='green', linestyle='--', linewidth=1.0, label='$f_c + B_T/2$')
    plt.title(f"Espectro FM para $\\beta={beta}$, $B_T \\approx {BT:.0f}$ Hz")
    plt.xlabel('Frecuencia [Hz]')
    plt.ylabel('$P(f)=|S(f)|^2$')
    plt.xlim(fc - 15000, fc + 15000)
    plt.grid(True, which='both', linestyle=':', alpha=0.5)
    plt.legend(loc='upper right')

plt.tight_layout()
plt.show()

# ============================================================
# TABLA DE RESULTADOS
# ============================================================

print('Resultados de la regla de Carson en FM')
print('---------------------------------------------------------------')
print(f"{'beta':>5} {'Delta_f [Hz]':>12} {'B_T [Hz]':>12} {'f_low [Hz]':>12} {'f_high [Hz]':>12} {'% dentro':>10}")
print('---------------------------------------------------------------')
for r in resultados:
    print(
        f"{r['beta']:5.1f} {r['Delta_f']:12.1f} {r['B_T']:12.1f} {r['f_low']:12.1f} {r['f_high']:12.1f} {r['porcentaje']:10.2f}"
    )
print('---------------------------------------------------------------')
```

# Interpretación experimental de la regla de Carson

En la simulación se generó una señal FM senoidal:

$$
s_{FM}(t)=A_c\cos\left(2\pi f_c t+\beta\sin(2\pi f_m t)\right)
$$

Para cada valor de $\beta$, se calculó:

$$
\Delta f=\beta f_m
$$

y luego se aplicó la regla de Carson:

$$
B_T\approx 2(\Delta f+f_m)
$$

El intervalo estimado por Carson es:

$$
f_c-\frac{B_T}{2}
\leq f \leq
f_c+\frac{B_T}{2}
$$

En las gráficas del espectro se dibujan dos líneas verticales que representan esos límites. Experimentalmente se observa que, aunque el espectro tiene más componentes fuera de ese intervalo, la mayor parte de la potencia significativa queda dentro de los límites de Carson.

Esto confirma la diferencia entre ancho de banda teórico y ancho de banda práctico:

- El ancho de banda teórico es infinito, porque la expansión espectral de FM tiene infinitas bandas laterales.
- El ancho de banda práctico es finito, porque las bandas laterales lejanas tienen amplitudes muy pequeñas.
- La regla de Carson aproxima el ancho de banda donde se concentra la mayor parte de la potencia útil de la señal FM.

Por tanto, la regla de Carson no dice que fuera de ese intervalo no exista espectro. Lo que indica es que fuera de ese intervalo la potencia suele ser suficientemente pequeña como para despreciarse en un análisis práctico.

## Conclusiones teóricas y experimentales

- El ancho de banda teórico de FM es infinito porque la expansión espectral de una señal FM senoidal tiene infinitas bandas laterales en $f_c \pm n f_m$.
- En la práctica se usa la regla de Carson para estimar una banda finita que contiene la mayor parte de la potencia útil de la señal.
- La regla de Carson predice:

  $$
  B_T \approx 2(\Delta f + f_m) = 2 f_m (\beta + 1)
  $$

  con $\Delta f = \beta f_m$.
- Experimentalmente, con la FFT se verifica la potencia espectral $P(f)=|S(f)|^2$, se suma la potencia en frecuencias positivas y se calcula qué porcentaje cae dentro del intervalo de Carson.
- Un valor de $0.98$ representa el 98 % de la potencia acumulada dentro de la banda de Carson, pero ese valor debe comprobarse con los datos reales de la simulación para cada $\beta$.
