## Regresión como modelo de dependencia estadística entre variables aleatorias

La regresión se entiende como un modelo funcional que busca describir la relación de dependencia entre un conjunto de variables de entrada y un conjunto de variables de salida. En forma general, se representa como:

$$
\eta = f(\xi)
$$

donde $\xi$ representa la variable o conjunto de variables de entrada, $\eta$ representa la variable o conjunto de variables de salida, y $f(\cdot)$ es la función que modela la dependencia entre ambas.

La idea de “caja negra” significa que inicialmente no conocemos la forma exacta de la función $f$. Solo observamos pares de datos entrada-salida y, a partir de ellos, intentamos aproximar la relación que existe entre las variables.

En problemas reales, esta relación no suele ser perfectamente determinista. Por eso se introduce incertidumbre mediante un término de error o ruido:

$$
\eta = f(\xi) + \varepsilon
$$

donde $\varepsilon$ representa la parte de la salida que el modelo no logra explicar. Este término puede deberse a ruido de medición, variables no consideradas, perturbaciones externas o variabilidad natural del sistema.

Desde el punto de vista estadístico, la regresión busca aproximar el comportamiento promedio de la salida cuando se conoce la entrada. Es decir, el modelo intenta aproximar:

$$
f(x) \approx \mathbb{E}[\eta \mid \xi = x]
$$

Esto significa que la regresión no predice necesariamente el valor exacto de la salida, sino el valor esperado o tendencia promedio de la variable dependiente para una entrada dada.

---

### Organización matricial de los datos

Si se tienen $M$ observaciones, $s$ variables de entrada y $r$ variables de salida, los datos se organizan en dos matrices:

$$
\mathbf{X}=
\begin{bmatrix}
x_{11} & x_{12} & \cdots & x_{1s}\\
x_{21} & x_{22} & \cdots & x_{2s}\\
\vdots & \vdots & \ddots & \vdots\\
x_{M1} & x_{M2} & \cdots & x_{Ms}
\end{bmatrix}
\in \mathbb{R}^{M\times s}
$$

y

$$
\mathbf{Y}=
\begin{bmatrix}
y_{11} & y_{12} & \cdots & y_{1r}\\
y_{21} & y_{22} & \cdots & y_{2r}\\
\vdots & \vdots & \ddots & \vdots\\
y_{M1} & y_{M2} & \cdots & y_{Mr}
\end{bmatrix}
\in \mathbb{R}^{M\times r}
$$

En la matriz $\mathbf{X}$:

- Cada fila representa una observación, muestra o experimento.
- Cada columna representa una variable de entrada o predictor.
- El elemento $x_{ij}$ representa el valor de la variable de entrada $j$ en la observación $i$.

Por tanto, la fila $i$ de $\mathbf{X}$ es:

$$
\mathbf{x}_i =
\begin{bmatrix}
x_{i1} & x_{i2} & \cdots & x_{is}
\end{bmatrix}
\in \mathbb{R}^{1\times s}
$$

Esta fila representa una muestra con $s$ características.

En la matriz $\mathbf{Y}$:

- Cada fila contiene las respuestas asociadas a una observación.
- Cada columna representa una variable de salida.
- El elemento $y_{ij}$ representa el valor de la salida $j$ en la observación $i$.

Por tanto, la fila $i$ de $\mathbf{Y}$ es:

$$
\mathbf{y}_i =
\begin{bmatrix}
y_{i1} & y_{i2} & \cdots & y_{ir}
\end{bmatrix}
\in \mathbb{R}^{1\times r}
$$

Esta fila representa las $r$ respuestas observadas para la muestra $i$.

---

### Forma funcional del modelo

De manera global, el modelo de regresión puede escribirse como una transformación entre matrices de datos:

$$
f:\mathbb{R}^{M\times s}\rightarrow\mathbb{R}^{M\times r}
$$

Sin embargo, desde el punto de vista de cada observación individual, el modelo actúa como una función:

$$
f:\mathbb{R}^{s}\rightarrow\mathbb{R}^{r}
$$

es decir, toma un vector de entrada con $s$ características y produce un vector de salida con $r$ respuestas.

Para la observación $i$, esto se expresa como:

$$
\mathbf{y}_i = f(\mathbf{x}_i) + \boldsymbol{\varepsilon}_i
$$

donde:

$$
\boldsymbol{\varepsilon}_i =
\begin{bmatrix}
\varepsilon_{i1} & \varepsilon_{i2} & \cdots & \varepsilon_{ir}
\end{bmatrix}
\in \mathbb{R}^{1\times r}
$$

es el vector de errores asociado a la observación $i$.

De forma matricial, el modelo completo puede escribirse como:

$$
\mathbf{Y}=f(\mathbf{X})+\mathbf{E}
$$

donde:

$$
\mathbf{E}=
\begin{bmatrix}
\varepsilon_{11} & \varepsilon_{12} & \cdots & \varepsilon_{1r}\\
\varepsilon_{21} & \varepsilon_{22} & \cdots & \varepsilon_{2r}\\
\vdots & \vdots & \ddots & \vdots\\
\varepsilon_{M1} & \varepsilon_{M2} & \cdots & \varepsilon_{Mr}
\end{bmatrix}
\in \mathbb{R}^{M\times r}
$$

representa todos los errores o residuales del modelo.

---

### Interpretación geométrica

Cada observación $\mathbf{x}_i$ puede interpretarse como un punto en un espacio de dimensión $s$:

$$
\mathbf{x}_i \in \mathbb{R}^{s}
$$

Por ejemplo:

- Si $s=1$, los datos de entrada viven sobre una recta.
- Si $s=2$, los datos viven en un plano.
- Si $s=3$, los datos viven en un espacio tridimensional.
- Si $s>3$, los datos viven en un espacio de alta dimensión.

La regresión intenta encontrar una regla que asocie cada punto de entrada con una salida correspondiente.

Cuando $r=1$, cada entrada se asocia con un único valor escalar de salida. Cuando $r>1$, cada entrada se asocia con un vector de salidas.

---

### Interpretación estadística

Como las variables se consideran aleatorias, la relación entre entrada y salida no es exacta. Por eso se introduce el error $\varepsilon$.

El modelo ideal intenta aproximar:

$$
f(x) \approx \mathbb{E}[\eta \mid \xi=x]
$$

Esto significa que, para una entrada fija $x$, la regresión intenta predecir el valor promedio esperado de la salida.

Por tanto, la regresión no predice necesariamente el resultado exacto de cada observación, sino la tendencia estadística principal que relaciona las entradas con las salidas.

---

### Idea central

En resumen, la regresión busca aprender una función de dependencia estadística entre variables aleatorias usando datos observados. Su objetivo no es encontrar una igualdad exacta, sino una aproximación que capture la tendencia principal de los datos bajo incertidumbre:

$$
\text{entrada} \longrightarrow \text{modelo de dependencia} \longrightarrow \text{salida estimada}
$$

o, de forma matemática:

$$
\mathbf{X} \longrightarrow f(\mathbf{X}) \approx \mathbf{Y}
$$

## Suposiciones sobre modelos de regresión

En un problema de regresión, los datos de entrada y salida se organizan normalmente en matrices. Si se tienen $M$ observaciones, $s$ variables de entrada y $r$ variables de salida, se define:

$$
\mathbf{X} =
\begin{bmatrix}
x_{11} & x_{12} & \cdots & x_{1s} \\
x_{21} & x_{22} & \cdots & x_{2s} \\
\vdots & \vdots & \ddots & \vdots \\
x_{M1} & x_{M2} & \cdots & x_{Ms}
\end{bmatrix}
\in \mathbb{R}^{M \times s}
$$

y

$$
\mathbf{Y} =
\begin{bmatrix}
y_{11} & y_{12} & \cdots & y_{1r} \\
y_{21} & y_{22} & \cdots & y_{2r} \\
\vdots & \vdots & \ddots & \vdots \\
y_{M1} & y_{M2} & \cdots & y_{Mr}
\end{bmatrix}
\in \mathbb{R}^{M \times r}
$$

donde:

- $M$ es el número de observaciones o muestras.
- $s$ es el número de variables de entrada o predictores.
- $r$ es el número de variables de salida o respuestas.
- $x_{ij}$ representa el valor de la variable de entrada $j$ en la observación $i$.
- $y_{ij}$ representa el valor de la variable de salida $j$ en la observación $i$.

Cada fila de $\mathbf{X}$ representa una observación completa:

$$
\mathbf{x}_i =
\begin{bmatrix}
x_{i1} & x_{i2} & \cdots & x_{is}
\end{bmatrix}
\in \mathbb{R}^{1\times s}
$$

y cada fila de $\mathbf{Y}$ representa la salida asociada a esa observación:

$$
\mathbf{y}_i =
\begin{bmatrix}
y_{i1} & y_{i2} & \cdots & y_{ir}
\end{bmatrix}
\in \mathbb{R}^{1\times r}
$$

Por tanto, cada observación puede entenderse como un par:

$$
(\mathbf{x}_i,\mathbf{y}_i)
$$

donde $\mathbf{x}_i$ contiene la información disponible para predecir y $\mathbf{y}_i$ contiene la respuesta que se desea explicar.

---

### Modelo simple o SISO

Un modelo SISO aparece cuando:

$$
s=1,\qquad r=1
$$

SISO significa *Single Input Single Output*, es decir, una sola entrada y una sola salida.

En este caso, la matriz de entrada y la matriz de salida toman la forma:

$$
\mathbf{X} =
\begin{bmatrix}
x_{11}\\
x_{21}\\
\vdots\\
x_{M1}
\end{bmatrix}
\in \mathbb{R}^{M\times 1},
\qquad
\mathbf{Y} =
\begin{bmatrix}
y_{11}\\
y_{21}\\
\vdots\\
y_{M1}
\end{bmatrix}
\in \mathbb{R}^{M\times 1}
$$

Cada observación tiene una única variable de entrada y una única variable de salida:

$$
x_i \longrightarrow y_i
$$

Un ejemplo sería estudiar la relación entre la temperatura exterior y el rendimiento de una máquina:

$$
\text{rendimiento}=f(\text{temperatura})
$$

Geométricamente, los datos se representan como puntos en el plano:

$$
(x_i,y_i)\in \mathbb{R}^2
$$

En este caso, si se usa un modelo lineal, la regresión busca ajustar una recta a la nube de puntos.

---

### Modelo múltiple tipo MISO

Un modelo MISO aparece cuando:

$$
s\geq 2,\qquad r=1
$$

MISO significa *Multiple Input Single Output*, es decir, múltiples entradas y una sola salida.

La matriz de entrada tiene varias columnas, pero la salida tiene una sola columna:

$$
\mathbf{X} =
\begin{bmatrix}
x_{11} & x_{12} & \cdots & x_{1s} \\
x_{21} & x_{22} & \cdots & x_{2s} \\
\vdots & \vdots & \ddots & \vdots \\
x_{M1} & x_{M2} & \cdots & x_{Ms}
\end{bmatrix}
\in \mathbb{R}^{M\times s},
\qquad
\mathbf{Y} =
\begin{bmatrix}
y_{11}\\
y_{21}\\
\vdots\\
y_{M1}
\end{bmatrix}
\in \mathbb{R}^{M\times 1}
$$

Cada observación contiene varias variables de entrada:

$$
\mathbf{x}_i =
\begin{bmatrix}
x_{i1} & x_{i2} & \cdots & x_{is}
\end{bmatrix}
$$

y una única respuesta:

$$
y_i
$$

Por ejemplo, el rendimiento de un proceso industrial puede depender de varias variables:

$$
\text{rendimiento}
=
f(\text{temperatura},\text{presión},\text{velocidad},\text{material})
$$

Si $s=4$, cada observación vive en un espacio de cuatro variables de entrada:

$$
\mathbf{x}_i\in\mathbb{R}^{4}
$$

Geométricamente, si el modelo es lineal, ya no se ajusta una recta sino un hiperplano. Para $s=2$, el hiperplano es un plano en tres dimensiones; para $s>2$, es un hiperplano en un espacio de mayor dimensión.

---

### Modelo múltiple tipo MIMO

Un modelo MIMO aparece cuando:

$$
s\geq 2,\qquad r\geq 2
$$

MIMO significa *Multiple Input Multiple Output*, es decir, múltiples entradas y múltiples salidas.

En este caso:

$$
\mathbf{X}\in\mathbb{R}^{M\times s},
\qquad
\mathbf{Y}\in\mathbb{R}^{M\times r}
$$

Cada observación tiene un vector de entrada y un vector de salida:

$$
\mathbf{x}_i \in \mathbb{R}^{1\times s},
\qquad
\mathbf{y}_i \in \mathbb{R}^{1\times r}
$$

Un ejemplo sería estudiar cómo varias condiciones de fabricación afectan varios parámetros de calidad:

$$
(\text{resistencia},\text{durabilidad})
=
f(\text{composición},\text{temperatura},\text{tiempo},\text{presión},\text{humedad})
$$

Si $s=5$ y $r=2$, entonces:

$$
f:\mathbb{R}^{5}\rightarrow\mathbb{R}^{2}
$$

Esto significa que el modelo transforma un vector de cinco entradas en un vector de dos salidas.

En términos geométricos, el modelo aprende una transformación entre un espacio de entrada de dimensión $s$ y un espacio de salida de dimensión $r$.

---

### Observaciones independientes e idénticamente distribuidas

Una suposición frecuente en regresión es que las observaciones son independientes e idénticamente distribuidas, abreviado como i.i.d.

Esto significa dos cosas:

1. Las observaciones son independientes.
2. Todas las observaciones provienen de la misma distribución probabilística.

Si tenemos observaciones:

$$
(\mathbf{x}_1,\mathbf{y}_1),
(\mathbf{x}_2,\mathbf{y}_2),
\dots,
(\mathbf{x}_M,\mathbf{y}_M)
$$

la independencia significa que conocer una observación no cambia la probabilidad de observar otra.

De forma conceptual:

$$
P\left((\mathbf{x}_1,\mathbf{y}_1),\dots,(\mathbf{x}_M,\mathbf{y}_M)\right)
=
\prod_{i=1}^{M} P(\mathbf{x}_i,\mathbf{y}_i)
$$

La condición de idénticamente distribuidas significa que todas las observaciones se generan bajo el mismo mecanismo estadístico.

Por ejemplo, si se mide el consumo energético de una máquina bajo las mismas condiciones de operación, se espera que las observaciones sigan una misma distribución. En cambio, si algunas mediciones se hacen con la máquina nueva y otras con la máquina desgastada, puede que ya no provengan de la misma distribución.

Esta suposición es importante porque muchos métodos de regresión se entrenan suponiendo que los datos futuros se comportarán estadísticamente de manera similar a los datos usados para entrenar.

---

### Homocedasticidad

Otra suposición importante es la homocedasticidad.

La homocedasticidad significa que los errores o residuales del modelo tienen varianza constante.

Si el modelo se escribe como:

$$
y_i = f(\mathbf{x}_i)+\varepsilon_i
$$

entonces la homocedasticidad se expresa como:

$$
\operatorname{Var}(\varepsilon_i)=\sigma^2
\qquad \text{para todo } i
$$

Esto significa que la dispersión del error es aproximadamente la misma para todas las observaciones.

El residual se define como:

$$
e_i = y_i-\hat{y}_i
$$

donde:

- $y_i$ es el valor real observado.
- $\hat{y}_i$ es el valor predicho por el modelo.
- $e_i$ es el error cometido en la observación $i$.

En un modelo homocedástico, los residuales deberían dispersarse de forma relativamente uniforme alrededor de cero.

Visualmente, si se grafican los residuales contra las predicciones, se espera algo como una nube con ancho aproximadamente constante.

En cambio, si la varianza de los residuales cambia según la región del espacio de entrada, aparece heterocedasticidad:

$$
\operatorname{Var}(\varepsilon_i)\neq\sigma^2
$$

En ese caso, algunas zonas del modelo presentan errores pequeños y otras zonas errores grandes.

Esto puede afectar la confiabilidad del modelo, especialmente cuando se interpretan intervalos de confianza, pruebas estadísticas o incertidumbre de predicción.

---

### Idea central de las suposiciones

Las suposiciones sobre modelos de regresión permiten definir el tipo de problema que se está resolviendo y las condiciones bajo las cuales el modelo puede interpretarse correctamente.

En resumen:

- $s$ indica cuántas variables de entrada tiene cada observación.
- $r$ indica cuántas variables de salida se desean predecir.
- $M$ indica cuántas observaciones están disponibles.
- SISO corresponde a una entrada y una salida.
- MISO corresponde a varias entradas y una salida.
- MIMO corresponde a varias entradas y varias salidas.
- La suposición i.i.d. indica que las observaciones son independientes y siguen la misma distribución.
- La homocedasticidad indica que los errores del modelo tienen varianza constante.

Estas ideas son necesarias antes de estudiar regresión lineal, mínimos cuadrados y evaluación de modelos, porque permiten entender qué tipo de relación se está modelando y qué condiciones estadísticas se están suponiendo.

```python
# ============================================================
# EJEMPLO INDEPENDIENTE: HOMOCEDASTICIDAD Y HETEROCEDASTICIDAD
# ============================================================

import numpy as np
import matplotlib.pyplot as plt


# ============================================================
# 1. FIJAR SEMILLA ALEATORIA
# ============================================================

# Esto permite que los resultados sean reproducibles.
# Cada vez que ejecutes el código, obtendrás la misma nube de puntos.
np.random.seed(42)


# ============================================================
# 2. GENERAR VARIABLE DE ENTRADA
# ============================================================

# Se generan 200 valores entre 0 y 10.
# Esta variable representa una entrada genérica x.
x = np.linspace(0, 10, 200)


# ============================================================
# 3. DEFINIR MODELO REAL
# ============================================================

# Suponemos una relación lineal verdadera:
# y = 2 + 3x + ruido
y_teorica = 2 + 3*x


# ============================================================
# 4. CASO HOMOCEDÁSTICO
# ============================================================

# En este caso, el ruido tiene varianza constante.
# La desviación estándar del ruido es siempre la misma.
ruido_homocedastico = np.random.normal(
    loc=0,
    scale=3,
    size=len(x)
)

# Salida observada con ruido homocedástico
y_homo = y_teorica + ruido_homocedastico

# En este ejemplo, asumimos que el modelo predice la tendencia verdadera.
y_pred_homo = y_teorica

# Residuales
residuales_homo = y_homo - y_pred_homo


# ============================================================
# 5. CASO HETEROCEDÁSTICO
# ============================================================

# En este caso, el ruido aumenta cuando x aumenta.
# Por eso la varianza del error no es constante.
ruido_heterocedastico = np.random.normal(
    loc=0,
    scale=0.5 + 0.5*x,
    size=len(x)
)

# Salida observada con ruido heterocedástico
y_hetero = y_teorica + ruido_heterocedastico

# El modelo predice la misma tendencia verdadera
y_pred_hetero = y_teorica

# Residuales
residuales_hetero = y_hetero - y_pred_hetero


# ============================================================
# 6. GRÁFICA DE RESIDUALES HOMOCEDÁSTICOS
# ============================================================

plt.figure(figsize=(8, 5))

plt.scatter(y_pred_homo, residuales_homo)

plt.axhline(y=0, linestyle='--')

plt.xlabel("Valores predichos $\\hat{y}$")
plt.ylabel("Residuales $e_i = y_i - \\hat{y}_i$")
plt.title("Caso homocedástico: varianza aproximadamente constante")

plt.grid(True)
plt.show()


# ============================================================
# 7. GRÁFICA DE RESIDUALES HETEROCEDÁSTICOS
# ============================================================

plt.figure(figsize=(8, 5))

plt.scatter(y_pred_hetero, residuales_hetero)

plt.axhline(y=0, linestyle='--')

plt.xlabel("Valores predichos $\\hat{y}$")
plt.ylabel("Residuales $e_i = y_i - \\hat{y}_i$")
plt.title("Caso heterocedástico: varianza no constante")

plt.grid(True)
plt.show()
```

## Aspectos básicos para construir una dependencia estadística

Antes de construir un modelo de regresión, no basta con asumir que existe una relación entre las variables. Primero es necesario analizar si realmente hay dependencia estadística, qué variables aportan información útil y qué tan bueno es el modelo para explicar o predecir la salida.

En regresión, normalmente se parte de un conjunto de variables de entrada:

$$
\mathbf{X} =
\begin{bmatrix}
X_1 & X_2 & \cdots & X_s
\end{bmatrix}
$$

y una variable o conjunto de variables de salida:

$$
\mathbf{Y}
$$

El objetivo es construir una función:

$$
\hat{\mathbf{Y}} = f(\mathbf{X})
$$

que aproxime lo mejor posible la salida real $\mathbf{Y}$.

Sin embargo, antes de ajustar el modelo aparecen varias preguntas fundamentales.

---

### 1. ¿Qué medidas de incertidumbre se pueden utilizar para evaluar la relación entre variables?

Para estudiar si existe interacción entre dos variables aleatorias, se pueden utilizar medidas estadísticas que cuantifican variabilidad, asociación o dependencia.

Algunas de las más importantes son:

#### Varianza

La varianza mide qué tanto se dispersa una variable respecto a su media:

$$
\operatorname{Var}(X)
=
\mathbb{E}\left[(X-\mathbb{E}[X])^2\right]
$$

Si la varianza es grande, los valores de la variable están muy dispersos. Si la varianza es pequeña, los valores están más concentrados alrededor de la media.

En regresión, la varianza permite entender qué tanta variabilidad existe en las variables antes de intentar modelarlas.

---

#### Covarianza

La covarianza mide si dos variables cambian juntas:

$$
\operatorname{Cov}(X,Y)
=
\mathbb{E}\left[(X-\mathbb{E}[X])(Y-\mathbb{E}[Y])\right]
$$

Si la covarianza es positiva, cuando $X$ aumenta, $Y$ tiende a aumentar.

Si la covarianza es negativa, cuando $X$ aumenta, $Y$ tiende a disminuir.

Si la covarianza es cercana a cero, no hay una relación lineal clara entre ambas variables.

La covarianza permite detectar dependencia lineal, pero su magnitud depende de las unidades de las variables.

---

#### Correlación

Para eliminar el efecto de las unidades, se utiliza la correlación:

$$
\rho_{XY}
=
\frac{\operatorname{Cov}(X,Y)}
{\sigma_X\sigma_Y}
$$

donde $\sigma_X$ y $\sigma_Y$ son las desviaciones estándar de $X$ y $Y$.

La correlación toma valores entre $-1$ y $1$:

$$
-1 \leq \rho_{XY} \leq 1
$$

- Si $\rho_{XY}\approx 1$, existe una relación lineal positiva fuerte.
- Si $\rho_{XY}\approx -1$, existe una relación lineal negativa fuerte.
- Si $\rho_{XY}\approx 0$, no existe una relación lineal evidente.

En el contexto de regresión, la correlación ayuda a identificar qué variables de entrada están más asociadas con la salida.

---

### 2. ¿Todas las variables contribuyen al modelo?

No necesariamente. Un conjunto de datos puede contener muchas variables de entrada, pero no todas aportan información útil para predecir la salida.

Supongamos un modelo lineal múltiple:

$$
\hat{Y}
=
\beta_0
+
\beta_1X_1
+
\beta_2X_2
+
\cdots
+
\beta_sX_s
$$

Cada coeficiente $\beta_j$ mide la contribución de la variable $X_j$ al modelo, manteniendo las demás variables constantes.

Si un coeficiente es grande en magnitud, la variable puede tener una influencia importante sobre la predicción.

Si un coeficiente es cercano a cero, la variable puede aportar poca información.

Sin embargo, la interpretación de los coeficientes depende de la escala de las variables. Por eso muchas veces se estandarizan los datos antes de comparar contribuciones.

---

### 3. ¿Cómo evaluar la contribución de cada variable al modelo?

La contribución de cada variable puede evaluarse de diferentes formas:

#### Análisis de correlación

Se calcula la correlación entre cada variable de entrada y la salida:

$$
\rho_{X_jY}
$$

Esto permite observar qué variables tienen mayor relación lineal individual con la salida.

---

#### Magnitud de los coeficientes

En un modelo lineal, se analizan los coeficientes:

$$
\beta_1,\beta_2,\dots,\beta_s
$$

Si las variables están en escalas comparables, la magnitud de cada coeficiente puede indicar qué variable tiene mayor peso en la predicción.

---

#### Error al eliminar una variable

También se puede entrenar el modelo con todas las variables y luego quitar una variable para observar si el error aumenta.

Si al eliminar $X_j$ el error del modelo aumenta mucho, significa que esa variable era importante.

---

#### Regularización

Métodos como Lasso pueden reducir algunos coeficientes exactamente a cero:

$$
\beta_j = 0
$$

Cuando esto ocurre, el modelo está indicando que la variable asociada puede no ser necesaria para la predicción.

### 4. ¿Cómo evaluar la efectividad de diferentes modelos?

Para comparar modelos de regresión se utilizan métricas de ajuste, error y capacidad explicativa.

La idea fundamental es comparar los valores reales:

$$
\mathbf{y}
=
\begin{bmatrix}
y_1\\
y_2\\
\vdots\\
y_M
\end{bmatrix}
$$

con los valores predichos por el modelo:

$$
\hat{\mathbf{y}}
=
\begin{bmatrix}
\hat{y}_1\\
\hat{y}_2\\
\vdots\\
\hat{y}_M
\end{bmatrix}
$$

El error o residual de cada observación se define como:

$$
e_i = y_i-\hat{y}_i
$$

y el vector completo de errores es:

$$
\mathbf{e}
=
\mathbf{y}
-
\hat{\mathbf{y}}
=
\begin{bmatrix}
y_1-\hat{y}_1\\
y_2-\hat{y}_2\\
\vdots\\
y_M-\hat{y}_M
\end{bmatrix}
$$

Por tanto, evaluar el ajuste de un modelo equivale a medir qué tan grande es el vector de error $\mathbf{e}$.

De forma general, se puede usar una métrica de ajuste o distancia:

$$
d(\mathbf{x},\mathbf{y})
$$

donde $\mathbf{x}$ y $\mathbf{y}$ son dos vectores que se desean comparar.

En regresión, esta métrica suele aplicarse entre el vector real $\mathbf{y}$ y el vector estimado $\hat{\mathbf{y}}$.

---

#### Distancia $L_1$

Una de las métricas de ajuste más usadas es la distancia $L_1$:

$$
\| \mathbf{x} - \mathbf{y} \|_1
=
\sum_{i=1}^{n} |x_i - y_i|
$$

Aplicada al problema de regresión:

$$
\| \mathbf{y} - \hat{\mathbf{y}} \|_1
=
\sum_{i=1}^{M} |y_i - \hat{y}_i|
$$

Esta distancia suma los errores absolutos entre los valores reales y los valores predichos.

Su interpretación es directa: mide la cantidad total de error cometido por el modelo, sin permitir que los errores positivos y negativos se cancelen.

La distancia $L_1$ es menos sensible a valores atípicos que la distancia $L_2$, porque los errores no se elevan al cuadrado.

Por esta razón, puede ser conveniente en problemas donde existen datos ruidosos, mediciones extremas o valores atípicos.

---

#### Distancia $L_2$

Otra métrica ampliamente usada es la distancia $L_2$:

$$
\| \mathbf{x} - \mathbf{y} \|_2
=
\left(
\sum_{i=1}^{n}
(x_i-y_i)^2
\right)^{1/2}
$$

Aplicada al problema de regresión:

$$
\| \mathbf{y} - \hat{\mathbf{y}} \|_2
=
\left(
\sum_{i=1}^{M}
(y_i-\hat{y}_i)^2
\right)^{1/2}
$$

Esta distancia corresponde a la distancia euclidiana entre el vector de valores reales y el vector de valores predichos.

Geométricamente, si $\mathbf{y}$ y $\hat{\mathbf{y}}$ se interpretan como puntos en un espacio de dimensión $M$, la distancia $L_2$ mide la distancia en línea recta entre ellos.

La distancia $L_2$ penaliza más fuertemente los errores grandes, porque cada diferencia se eleva al cuadrado. Por ejemplo:

$$
2^2=4
$$

pero:

$$
10^2=100
$$

Esto significa que un error grande influye mucho más en una métrica basada en $L_2$ que en una métrica basada en $L_1$.

Por esta razón, las métricas basadas en $L_2$ son útiles cuando se quiere castigar fuertemente las predicciones muy alejadas del valor real.

---

#### Error cuadrático medio

Una de las métricas más usadas en regresión es el error cuadrático medio, conocido como MSE:

$$
MSE
=
\frac{1}{M}
\sum_{i=1}^{M}
(y_i-\hat{y}_i)^2
$$

El MSE mide el promedio de los errores al cuadrado.

Esta métrica está directamente relacionada con la distancia $L_2$ al cuadrado, porque usa términos de la forma:

$$
(y_i-\hat{y}_i)^2
$$

Por tanto, el MSE penaliza fuertemente errores grandes.

Si el MSE es pequeño, significa que, en promedio, las predicciones están cerca de los valores reales.

---

#### Raíz del error cuadrático medio

La raíz del error cuadrático medio, RMSE, se define como:

$$
RMSE
=
\sqrt{MSE}
$$

Es decir:

$$
RMSE
=
\sqrt{
\frac{1}{M}
\sum_{i=1}^{M}
(y_i-\hat{y}_i)^2
}
$$

El RMSE tiene las mismas unidades que la variable de salida, por lo que suele ser más interpretable que el MSE.

Por ejemplo, si la variable $Y$ representa ventas, entonces el RMSE también se interpreta en unidades de ventas.

---

#### Error absoluto medio

El error absoluto medio, MAE, se define como:

$$
MAE
=
\frac{1}{M}
\sum_{i=1}^{M}
|y_i-\hat{y}_i|
$$

El MAE mide el promedio de los errores absolutos.

Esta métrica está relacionada con la distancia $L_1$, porque usa términos de la forma:

$$
|y_i-\hat{y}_i|
$$

El MAE es menos sensible a errores extremos que el MSE, ya que no eleva los errores al cuadrado.

Por esta razón, puede ser útil cuando se quiere una medida de error más robusta frente a valores atípicos.

---

#### Coeficiente de determinación

El coeficiente de determinación, $R^2$, se define como:

$$
R^2
=
1-
\frac{
\sum_{i=1}^{M}(y_i-\hat{y}_i)^2
}{
\sum_{i=1}^{M}(y_i-\bar{y})^2
}
$$

donde:

$$
\bar{y}
=
\frac{1}{M}
\sum_{i=1}^{M}
y_i
$$

es el valor promedio de la variable de salida.

El numerador:

$$
\sum_{i=1}^{M}(y_i-\hat{y}_i)^2
$$

representa la suma de errores cuadráticos del modelo.

El denominador:

$$
\sum_{i=1}^{M}(y_i-\bar{y})^2
$$

representa la variabilidad total de los datos respecto a su media.

Por tanto, $R^2$ mide qué proporción de la variabilidad de $Y$ logra explicar el modelo.

Si $R^2$ es cercano a 1, el modelo explica gran parte de la variabilidad de los datos.

Si $R^2$ es cercano a 0, el modelo no mejora mucho respecto a predecir simplemente la media de $Y$.

Si $R^2$ es negativo, el modelo puede estar funcionando peor que usar la media como predicción.

Un modelo no solo debe predecir bien, también puede ser importante que sea interpretable.

En un modelo lineal, la interpretación es directa:

$$
\hat{Y}
=
\beta_0
+
\beta_1X_1
+
\cdots
+
\beta_sX_s
$$

Cada coeficiente indica cómo cambia la predicción cuando cambia una variable, manteniendo las demás constantes.

Por ejemplo, si $\beta_1>0$, entonces al aumentar $X_1$, la salida predicha tiende a aumentar.

Si $\beta_1<0$, entonces al aumentar $X_1$, la salida predicha tiende a disminuir.

Por eso la regresión lineal es útil no solo para predecir, sino también para explicar relaciones entre variables.

---

### Idea central

Para construir una dependencia estadística mediante regresión se deben responder cuatro preguntas:

1. ¿Existe relación estadística entre las variables?
2. ¿Todas las variables aportan información útil?
3. ¿Cómo se mide la contribución de cada variable?
4. ¿Qué tan efectivo e interpretable es el modelo?

Estas preguntas permiten pasar de una simple visualización de datos a la construcción de un modelo estadístico formal.

```python
import numpy as np
import matplotlib.pyplot as plt

# Valores para comparar distancias L1 y L2
x_vals = np.linspace(-3, 3, 100)

y_vals_L1 = np.abs(x_vals)   # L1: distancia absoluta
y_vals_L2 = x_vals**2        # L2: distancia cuadrática

plt.figure(figsize=(10, 6))

plt.plot(x_vals, y_vals_L1, color="red", label="Distancia L1 (|x|)")
plt.plot(x_vals, y_vals_L2, color="blue", label="Distancia L2 (x^2)")

plt.axhline(y=1, color="black", linestyle="-")

plt.legend(loc="upper right")
plt.title("Comparación de distancias L1 y L2")
plt.xlabel("Valores de X")
plt.ylabel("Distancia")

plt.show()
```

## Estimación de los parámetros del modelo de regresión lineal simple

En esta sección se presenta una primera interpretación directa de los parámetros de un modelo de regresión lineal simple.

El modelo lineal simple se escribe como:

$$
\hat{y} = \beta_0 + \beta_1 x
$$

donde:

- $\hat{y}$ es el valor predicho por el modelo.
- $x$ es la variable de entrada.
- $\beta_0$ es el intercepto de la recta.
- $\beta_1$ es la pendiente de la recta.

Si los datos observados contienen ruido, se puede escribir:

$$
y_i = \beta_0 + \beta_1 x_i + \varepsilon_i
$$

donde $\varepsilon_i$ representa el error, ruido o perturbación de la observación $i$.

En el código se genera artificialmente una recta verdadera:

$$
y = -1 + 2.5x
$$

Luego se agrega ruido aleatorio a cada punto:

$$
y_i = -1 + 2.5x_i + \varepsilon_i
$$

Por esta razón, los puntos observados no caen exactamente sobre la recta verdadera, sino alrededor de ella.

---

### Cálculo de pendiente e intercepto usando pares de puntos

Una recta queda completamente determinada por dos puntos. Si se tienen dos puntos:

$$
(x_i,y_i)
\qquad \text{y} \qquad
(x_j,y_j)
$$

la pendiente de la recta que pasa por ellos es:

$$
m_{ij}
=
\frac{y_j-y_i}{x_j-x_i}
$$

Una vez conocida la pendiente, el intercepto se obtiene despejando de la ecuación de la recta:

$$
y_i = b_{ij} + m_{ij}x_i
$$

Por tanto:

$$
b_{ij}
=
y_i-m_{ij}x_i
$$

Cada par de puntos permite construir una recta diferente:

$$
\hat{y}_{ij}
=
b_{ij}
+
m_{ij}x
$$

En la gráfica, las rectas punteadas corresponden a las rectas obtenidas a partir de diferentes pares de puntos seleccionados aleatoriamente.

---

### Promedio de parámetros

Si se seleccionan $K$ pares de puntos, se obtienen $K$ pendientes:

$$
m_1,m_2,\dots,m_K
$$

y $K$ interceptos:

$$
b_1,b_2,\dots,b_K
$$

El código calcula la pendiente promedio:

$$
\bar{m}
=
\frac{1}{K}
\sum_{k=1}^{K}
m_k
$$

y el intercepto promedio:

$$
\bar{b}
=
\frac{1}{K}
\sum_{k=1}^{K}
b_k
$$

Con estos valores se construye una recta promedio:

$$
\hat{y}
=
\bar{b}
+
\bar{m}x
$$

Esta recta aparece en color rojo en la gráfica.

---

### Interpretación de la gráfica

En la gráfica se observan varios elementos:

- La recta verde representa la recta verdadera usada para generar los datos.
- Los puntos azules representan las muestras observadas con ruido.
- Las rectas punteadas representan rectas calculadas usando diferentes pares de puntos.
- Los puntos resaltados con borde negro son los pares usados para calcular cada recta.
- La recta roja representa la recta obtenida al promediar las pendientes e interceptos de las rectas punteadas.

La idea principal es mostrar que, si los datos siguen aproximadamente una relación lineal, es posible estimar una recta a partir de ellos. Sin embargo, estimar la recta usando solo pares de puntos es un método inestable, porque depende mucho de qué puntos se seleccionen y de cuánto ruido tengan.

Por esta razón, este enfoque sirve como interpretación inicial, pero no como método óptimo.

El método formal de regresión lineal buscará estimar los parámetros usando todos los puntos al mismo tiempo, minimizando el error total del modelo.

```python
import numpy as np
import matplotlib.pyplot as plt

# Semilla para reproducibilidad
np.random.seed(42)

# Generar una muestra de puntos a lo largo de una recta verdadera con ruido
n_puntos = 50
x = np.linspace(0, 10, n_puntos)

pendiente_verdadera = 2.5
intercepto_verdadero = -1

ruido = np.random.normal(0, 1, n_puntos)
y = pendiente_verdadera * x + intercepto_verdadero + ruido

# Seleccionar algunos pares de puntos para estimar pendiente e intercepto
n_pares = 5
pares_idx = np.random.choice(range(n_puntos), (n_pares, 2), replace=False)

# Calcular pendientes e interceptos para los pares seleccionados
pendientes = []
interceptos = []

for idx1, idx2 in pares_idx:
    pendiente = (y[idx2] - y[idx1]) / (x[idx2] - x[idx1])
    intercepto = y[idx1] - pendiente * x[idx1]

    pendientes.append(pendiente)
    interceptos.append(intercepto)

# Calcular pendiente e intercepto promedio
pendiente_promedio = np.mean(pendientes)
intercepto_promedio = np.mean(interceptos)

# Crear figura
plt.figure(figsize=(8, 6))

# Graficar la recta verdadera
plt.plot(
    x,
    pendiente_verdadera * x + intercepto_verdadero,
    label="Recta Verdadera",
    color="green"
)

# Graficar los puntos con ruido
plt.scatter(
    x,
    y,
    label="Muestras con ruido",
    color="blue",
    s=20
)

# Colores para las rectas de los pares de puntos
colores = plt.cm.rainbow(np.linspace(0, 1, n_pares))

# Graficar las rectas estimadas por pares de puntos
for (idx1, idx2), pendiente, intercepto, color in zip(
    pares_idx,
    pendientes,
    interceptos,
    colores
):
    plt.plot(
        x,
        pendiente * x + intercepto,
        linestyle="--",
        color=color,
        alpha=0.6
    )

    plt.scatter(
        [x[idx1], x[idx2]],
        [y[idx1], y[idx2]],
        color=color,
        edgecolor="black",
        zorder=5
    )

# Graficar la recta con parámetros promedio
plt.plot(
    x,
    pendiente_promedio * x + intercepto_promedio,
    label="Recta con Parámetros Promedio",
    color="red"
)

# Etiquetas y leyenda
plt.title("Regresión Lineal a partir de parámetros promedio")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()

plt.show()
```

### Segundo enfoque: Cálculo empírico de los parámetros de una regresión lineal:

En este segundo enfoque se interpreta la regresión lineal como un proceso de búsqueda empírica de los mejores parámetros de una recta.

El modelo lineal simple se escribe como:

$$
\hat{y} = \theta_0 + \theta_1 x
$$

donde:

- $\hat{y}$ es el valor predicho por el modelo.
- $x$ es la variable de entrada.
- $\theta_0$ es el intercepto de la recta.
- $\theta_1$ es la pendiente de la recta.

El objetivo es encontrar los valores de $\theta_0$ y $\theta_1$ que hagan que la recta predicha esté lo más cerca posible de los datos observados.

Si se tienen $M$ observaciones:

$$
(x_1,y_1),(x_2,y_2),\dots,(x_M,y_M)
$$

entonces para cada observación el modelo predice:

$$
\hat{y}_i = \theta_0 + \theta_1 x_i
$$

El error de predicción para cada punto es:

$$
e_i = y_i - \hat{y}_i
$$

es decir:

$$
e_i = y_i - (\theta_0 + \theta_1 x_i)
$$

Para medir el error total del modelo se usa el error cuadrático medio, MSE:

$$
MSE(\theta_0,\theta_1)
=
\frac{1}{M}
\sum_{i=1}^{M}
\left(
y_i - \hat{y}_i
\right)^2
$$

Sustituyendo la expresión de la predicción:

$$
MSE(\theta_0,\theta_1)
=
\frac{1}{M}
\sum_{i=1}^{M}
\left(
y_i - (\theta_0+\theta_1x_i)
\right)^2
$$

Esta función mide qué tan buena es una recta específica. Cada combinación de valores $(\theta_0,\theta_1)$ produce una recta diferente y, por tanto, un error diferente.

---

#### Interpretación geométrica

El parámetro $\theta_0$ desplaza la recta hacia arriba o hacia abajo. Por eso se llama intercepto.

El parámetro $\theta_1$ controla la inclinación de la recta. Si $\theta_1$ aumenta, la recta se vuelve más inclinada positivamente. Si $\theta_1$ disminuye, la recta puede volverse más plana o incluso tener pendiente negativa.

Por tanto, ajustar una regresión lineal puede verse geométricamente como mover y rotar una recta hasta encontrar una posición en la que la distancia entre la recta y los puntos sea mínima.

La idea de “rotar la varita” hace referencia precisamente a modificar la pendiente $\theta_1$ y el intercepto $\theta_0$ hasta que la recta se ajuste mejor a los datos.

---

#### Interpretación computacional

En el código se generan datos artificiales a partir de una recta verdadera:

$$
y = 2 + 3x
$$

Luego se agrega ruido aleatorio:

$$
y_i = 2 + 3x_i + \varepsilon_i
$$

donde $\varepsilon_i$ representa el ruido de cada observación.

Después, mediante controles interactivos, el usuario modifica manualmente:

$$
\theta_0
$$

y

$$
\theta_1
$$

para construir una recta candidata:

$$
\hat{y} = \theta_0 + \theta_1 x
$$

El código calcula el error cuadrático medio entre la recta propuesta por el usuario y los datos con ruido:

$$
MSE =
\frac{1}{M}
\sum_{i=1}^{M}
(y_i-\hat{y}_i)^2
$$

La recta que produzca un MSE más bajo será una mejor aproximación a los datos.

---

#### Diferencia con el primer enfoque

En el primer enfoque se tomaban varios pares de puntos, se calculaban varias rectas y luego se promediaban sus parámetros.

En este segundo enfoque, en cambio, se plantea directamente la idea de buscar la recta que minimice una función de error.

Este enfoque todavía es manual o heurístico, porque el usuario mueve los parámetros con controles interactivos. Sin embargo, introduce la idea central de la optimización:

$$
\min_{\theta_0,\theta_1}
MSE(\theta_0,\theta_1)
$$

Es decir, se busca encontrar los parámetros que hacen mínimo el error del modelo.

Más adelante, este procedimiento se formaliza mediante métodos como mínimos cuadrados ordinarios, donde los mejores parámetros se calculan matemáticamente usando todos los datos al mismo tiempo.

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import FloatSlider, Checkbox, interact
from sklearn.metrics import mean_squared_error

np.random.seed(42)

# Función para generar datos de una línea recta y añadir ruido
def generate_data(theta0, theta1, sigma):
    n_points = 100
    np.random.seed(42)

    X = 10 * np.random.rand(n_points, 1)
    y_original = theta1 * X + theta0

    noise = sigma * np.random.randn(n_points, 1)
    y_noisy = y_original + noise

    return X, y_original, y_noisy


class LinearDataVisualizer:
    def __init__(self, theta0, theta1, sigma):
        self.theta0 = theta0
        self.theta1 = theta1
        self.sigma = sigma

    def fit_user_line(self, theta0_user, theta1_user, show_original):
        X, y_original, y_noisy = generate_data(
            self.theta0,
            self.theta1,
            self.sigma
        )

        y_user = theta1_user * X + theta0_user
        mse = mean_squared_error(y_noisy, y_user)

        # Ordenar X para que las líneas se vean continuas
        idx = np.argsort(X[:, 0])
        X_sorted = X[idx]
        y_original_sorted = y_original[idx]
        y_user_sorted = y_user[idx]

        plt.figure(figsize=(10, 6))

        if show_original:
            plt.plot(
                X_sorted,
                y_original_sorted,
                color="green",
                label="Línea original"
            )

        plt.scatter(
            X,
            y_noisy,
            color="red",
            s=15,
            label="Datos con ruido"
        )

        plt.plot(
            X_sorted,
            y_user_sorted,
            "b--",
            label=f"Línea ajustada (Error = {mse:.4f})"
        )

        plt.xlabel("x", fontsize=14)
        plt.ylabel("y", fontsize=14)
        plt.title("Ajuste de la línea por el usuario")
        plt.ylim((0, 35))
        plt.legend()
        plt.show()


visualizer = LinearDataVisualizer(theta0=2, theta1=3, sigma=0.5)

interact(
    visualizer.fit_user_line,
    theta0_user=FloatSlider(
        min=-10,
        max=10,
        step=0.1,
        value=1,
        description="Intercepto"
    ),
    theta1_user=FloatSlider(
        min=-10,
        max=10,
        step=0.1,
        value=2,
        description="Pendiente"
    ),
    show_original=Checkbox(
        value=False,
        description="Mostrar línea original"
    )
);

```
