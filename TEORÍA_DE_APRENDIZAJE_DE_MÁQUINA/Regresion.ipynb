## Regresión como modelo de dependencia estadística entre variables aleatorias

La regresión se entiende como un modelo funcional que busca describir la relación de dependencia entre un conjunto de variables de entrada y un conjunto de variables de salida. En forma general, se representa como:

$$
\eta = f(\xi)
$$

donde $\xi$ representa la variable o conjunto de variables de entrada, $\eta$ representa la variable o conjunto de variables de salida, y $f(\cdot)$ es la función que modela la dependencia entre ambas.

La idea de “caja negra” significa que inicialmente no conocemos la forma exacta de la función $f$. Solo observamos pares de datos entrada-salida y, a partir de ellos, intentamos aproximar la relación que existe entre las variables.

En problemas reales, esta relación no suele ser perfectamente determinista. Por eso se introduce incertidumbre mediante un término de error o ruido:

$$
\eta = f(\xi) + \varepsilon
$$

donde $\varepsilon$ representa la parte de la salida que el modelo no logra explicar. Este término puede deberse a ruido de medición, variables no consideradas, perturbaciones externas o variabilidad natural del sistema.

Desde el punto de vista estadístico, la regresión busca aproximar el comportamiento promedio de la salida cuando se conoce la entrada. Es decir, el modelo intenta aproximar:

$$
f(x) \approx \mathbb{E}[\eta \mid \xi = x]
$$

Esto significa que la regresión no predice necesariamente el valor exacto de la salida, sino el valor esperado o tendencia promedio de la variable dependiente para una entrada dada.

---

### Organización matricial de los datos

Si se tienen $M$ observaciones, $s$ variables de entrada y $r$ variables de salida, los datos se organizan en dos matrices:

$$
\mathbf{X}=
\begin{bmatrix}
x_{11} & x_{12} & \cdots & x_{1s}\\
x_{21} & x_{22} & \cdots & x_{2s}\\
\vdots & \vdots & \ddots & \vdots\\
x_{M1} & x_{M2} & \cdots & x_{Ms}
\end{bmatrix}
\in \mathbb{R}^{M\times s}
$$

y

$$
\mathbf{Y}=
\begin{bmatrix}
y_{11} & y_{12} & \cdots & y_{1r}\\
y_{21} & y_{22} & \cdots & y_{2r}\\
\vdots & \vdots & \ddots & \vdots\\
y_{M1} & y_{M2} & \cdots & y_{Mr}
\end{bmatrix}
\in \mathbb{R}^{M\times r}
$$

En la matriz $\mathbf{X}$:

- Cada fila representa una observación, muestra o experimento.
- Cada columna representa una variable de entrada o predictor.
- El elemento $x_{ij}$ representa el valor de la variable de entrada $j$ en la observación $i$.

Por tanto, la fila $i$ de $\mathbf{X}$ es:

$$
\mathbf{x}_i =
\begin{bmatrix}
x_{i1} & x_{i2} & \cdots & x_{is}
\end{bmatrix}
\in \mathbb{R}^{1\times s}
$$

Esta fila representa una muestra con $s$ características.

En la matriz $\mathbf{Y}$:

- Cada fila contiene las respuestas asociadas a una observación.
- Cada columna representa una variable de salida.
- El elemento $y_{ij}$ representa el valor de la salida $j$ en la observación $i$.

Por tanto, la fila $i$ de $\mathbf{Y}$ es:

$$
\mathbf{y}_i =
\begin{bmatrix}
y_{i1} & y_{i2} & \cdots & y_{ir}
\end{bmatrix}
\in \mathbb{R}^{1\times r}
$$

Esta fila representa las $r$ respuestas observadas para la muestra $i$.

---

### Forma funcional del modelo

De manera global, el modelo de regresión puede escribirse como una transformación entre matrices de datos:

$$
f:\mathbb{R}^{M\times s}\rightarrow\mathbb{R}^{M\times r}
$$

Sin embargo, desde el punto de vista de cada observación individual, el modelo actúa como una función:

$$
f:\mathbb{R}^{s}\rightarrow\mathbb{R}^{r}
$$

es decir, toma un vector de entrada con $s$ características y produce un vector de salida con $r$ respuestas.

Para la observación $i$, esto se expresa como:

$$
\mathbf{y}_i = f(\mathbf{x}_i) + \boldsymbol{\varepsilon}_i
$$

donde:

$$
\boldsymbol{\varepsilon}_i =
\begin{bmatrix}
\varepsilon_{i1} & \varepsilon_{i2} & \cdots & \varepsilon_{ir}
\end{bmatrix}
\in \mathbb{R}^{1\times r}
$$

es el vector de errores asociado a la observación $i$.

De forma matricial, el modelo completo puede escribirse como:

$$
\mathbf{Y}=f(\mathbf{X})+\mathbf{E}
$$

donde:

$$
\mathbf{E}=
\begin{bmatrix}
\varepsilon_{11} & \varepsilon_{12} & \cdots & \varepsilon_{1r}\\
\varepsilon_{21} & \varepsilon_{22} & \cdots & \varepsilon_{2r}\\
\vdots & \vdots & \ddots & \vdots\\
\varepsilon_{M1} & \varepsilon_{M2} & \cdots & \varepsilon_{Mr}
\end{bmatrix}
\in \mathbb{R}^{M\times r}
$$

representa todos los errores o residuales del modelo.

---

### Interpretación geométrica

Cada observación $\mathbf{x}_i$ puede interpretarse como un punto en un espacio de dimensión $s$:

$$
\mathbf{x}_i \in \mathbb{R}^{s}
$$

Por ejemplo:

- Si $s=1$, los datos de entrada viven sobre una recta.
- Si $s=2$, los datos viven en un plano.
- Si $s=3$, los datos viven en un espacio tridimensional.
- Si $s>3$, los datos viven en un espacio de alta dimensión.

La regresión intenta encontrar una regla que asocie cada punto de entrada con una salida correspondiente.

Cuando $r=1$, cada entrada se asocia con un único valor escalar de salida. Cuando $r>1$, cada entrada se asocia con un vector de salidas.

---

### Interpretación estadística

Como las variables se consideran aleatorias, la relación entre entrada y salida no es exacta. Por eso se introduce el error $\varepsilon$.

El modelo ideal intenta aproximar:

$$
f(x) \approx \mathbb{E}[\eta \mid \xi=x]
$$

Esto significa que, para una entrada fija $x$, la regresión intenta predecir el valor promedio esperado de la salida.

Por tanto, la regresión no predice necesariamente el resultado exacto de cada observación, sino la tendencia estadística principal que relaciona las entradas con las salidas.

---

### Idea central

En resumen, la regresión busca aprender una función de dependencia estadística entre variables aleatorias usando datos observados. Su objetivo no es encontrar una igualdad exacta, sino una aproximación que capture la tendencia principal de los datos bajo incertidumbre:

$$
\text{entrada} \longrightarrow \text{modelo de dependencia} \longrightarrow \text{salida estimada}
$$

o, de forma matemática:

$$
\mathbf{X} \longrightarrow f(\mathbf{X}) \approx \mathbf{Y}
$$

## Suposiciones sobre modelos de regresión

En un problema de regresión, los datos de entrada y salida se organizan normalmente en matrices. Si se tienen $M$ observaciones, $s$ variables de entrada y $r$ variables de salida, se define:

$$
\mathbf{X} =
\begin{bmatrix}
x_{11} & x_{12} & \cdots & x_{1s} \\
x_{21} & x_{22} & \cdots & x_{2s} \\
\vdots & \vdots & \ddots & \vdots \\
x_{M1} & x_{M2} & \cdots & x_{Ms}
\end{bmatrix}
\in \mathbb{R}^{M \times s}
$$

y

$$
\mathbf{Y} =
\begin{bmatrix}
y_{11} & y_{12} & \cdots & y_{1r} \\
y_{21} & y_{22} & \cdots & y_{2r} \\
\vdots & \vdots & \ddots & \vdots \\
y_{M1} & y_{M2} & \cdots & y_{Mr}
\end{bmatrix}
\in \mathbb{R}^{M \times r}
$$

donde:

- $M$ es el número de observaciones o muestras.
- $s$ es el número de variables de entrada o predictores.
- $r$ es el número de variables de salida o respuestas.
- $x_{ij}$ representa el valor de la variable de entrada $j$ en la observación $i$.
- $y_{ij}$ representa el valor de la variable de salida $j$ en la observación $i$.

Cada fila de $\mathbf{X}$ representa una observación completa:

$$
\mathbf{x}_i =
\begin{bmatrix}
x_{i1} & x_{i2} & \cdots & x_{is}
\end{bmatrix}
\in \mathbb{R}^{1\times s}
$$

y cada fila de $\mathbf{Y}$ representa la salida asociada a esa observación:

$$
\mathbf{y}_i =
\begin{bmatrix}
y_{i1} & y_{i2} & \cdots & y_{ir}
\end{bmatrix}
\in \mathbb{R}^{1\times r}
$$

Por tanto, cada observación puede entenderse como un par:

$$
(\mathbf{x}_i,\mathbf{y}_i)
$$

donde $\mathbf{x}_i$ contiene la información disponible para predecir y $\mathbf{y}_i$ contiene la respuesta que se desea explicar.

---

### Modelo simple o SISO

Un modelo SISO aparece cuando:

$$
s=1,\qquad r=1
$$

SISO significa *Single Input Single Output*, es decir, una sola entrada y una sola salida.

En este caso, la matriz de entrada y la matriz de salida toman la forma:

$$
\mathbf{X} =
\begin{bmatrix}
x_{11}\\
x_{21}\\
\vdots\\
x_{M1}
\end{bmatrix}
\in \mathbb{R}^{M\times 1},
\qquad
\mathbf{Y} =
\begin{bmatrix}
y_{11}\\
y_{21}\\
\vdots\\
y_{M1}
\end{bmatrix}
\in \mathbb{R}^{M\times 1}
$$

Cada observación tiene una única variable de entrada y una única variable de salida:

$$
x_i \longrightarrow y_i
$$

Un ejemplo sería estudiar la relación entre la temperatura exterior y el rendimiento de una máquina:

$$
\text{rendimiento}=f(\text{temperatura})
$$

Geométricamente, los datos se representan como puntos en el plano:

$$
(x_i,y_i)\in \mathbb{R}^2
$$

En este caso, si se usa un modelo lineal, la regresión busca ajustar una recta a la nube de puntos.

---

### Modelo múltiple tipo MISO

Un modelo MISO aparece cuando:

$$
s\geq 2,\qquad r=1
$$

MISO significa *Multiple Input Single Output*, es decir, múltiples entradas y una sola salida.

La matriz de entrada tiene varias columnas, pero la salida tiene una sola columna:

$$
\mathbf{X} =
\begin{bmatrix}
x_{11} & x_{12} & \cdots & x_{1s} \\
x_{21} & x_{22} & \cdots & x_{2s} \\
\vdots & \vdots & \ddots & \vdots \\
x_{M1} & x_{M2} & \cdots & x_{Ms}
\end{bmatrix}
\in \mathbb{R}^{M\times s},
\qquad
\mathbf{Y} =
\begin{bmatrix}
y_{11}\\
y_{21}\\
\vdots\\
y_{M1}
\end{bmatrix}
\in \mathbb{R}^{M\times 1}
$$

Cada observación contiene varias variables de entrada:

$$
\mathbf{x}_i =
\begin{bmatrix}
x_{i1} & x_{i2} & \cdots & x_{is}
\end{bmatrix}
$$

y una única respuesta:

$$
y_i
$$

Por ejemplo, el rendimiento de un proceso industrial puede depender de varias variables:

$$
\text{rendimiento}
=
f(\text{temperatura},\text{presión},\text{velocidad},\text{material})
$$

Si $s=4$, cada observación vive en un espacio de cuatro variables de entrada:

$$
\mathbf{x}_i\in\mathbb{R}^{4}
$$

Geométricamente, si el modelo es lineal, ya no se ajusta una recta sino un hiperplano. Para $s=2$, el hiperplano es un plano en tres dimensiones; para $s>2$, es un hiperplano en un espacio de mayor dimensión.

---

### Modelo múltiple tipo MIMO

Un modelo MIMO aparece cuando:

$$
s\geq 2,\qquad r\geq 2
$$

MIMO significa *Multiple Input Multiple Output*, es decir, múltiples entradas y múltiples salidas.

En este caso:

$$
\mathbf{X}\in\mathbb{R}^{M\times s},
\qquad
\mathbf{Y}\in\mathbb{R}^{M\times r}
$$

Cada observación tiene un vector de entrada y un vector de salida:

$$
\mathbf{x}_i \in \mathbb{R}^{1\times s},
\qquad
\mathbf{y}_i \in \mathbb{R}^{1\times r}
$$

Un ejemplo sería estudiar cómo varias condiciones de fabricación afectan varios parámetros de calidad:

$$
(\text{resistencia},\text{durabilidad})
=
f(\text{composición},\text{temperatura},\text{tiempo},\text{presión},\text{humedad})
$$

Si $s=5$ y $r=2$, entonces:

$$
f:\mathbb{R}^{5}\rightarrow\mathbb{R}^{2}
$$

Esto significa que el modelo transforma un vector de cinco entradas en un vector de dos salidas.

En términos geométricos, el modelo aprende una transformación entre un espacio de entrada de dimensión $s$ y un espacio de salida de dimensión $r$.

---

### Observaciones independientes e idénticamente distribuidas

Una suposición frecuente en regresión es que las observaciones son independientes e idénticamente distribuidas, abreviado como i.i.d.

Esto significa dos cosas:

1. Las observaciones son independientes.
2. Todas las observaciones provienen de la misma distribución probabilística.

Si tenemos observaciones:

$$
(\mathbf{x}_1,\mathbf{y}_1),
(\mathbf{x}_2,\mathbf{y}_2),
\dots,
(\mathbf{x}_M,\mathbf{y}_M)
$$

la independencia significa que conocer una observación no cambia la probabilidad de observar otra.

De forma conceptual:

$$
P\left((\mathbf{x}_1,\mathbf{y}_1),\dots,(\mathbf{x}_M,\mathbf{y}_M)\right)
=
\prod_{i=1}^{M} P(\mathbf{x}_i,\mathbf{y}_i)
$$

La condición de idénticamente distribuidas significa que todas las observaciones se generan bajo el mismo mecanismo estadístico.

Por ejemplo, si se mide el consumo energético de una máquina bajo las mismas condiciones de operación, se espera que las observaciones sigan una misma distribución. En cambio, si algunas mediciones se hacen con la máquina nueva y otras con la máquina desgastada, puede que ya no provengan de la misma distribución.

Esta suposición es importante porque muchos métodos de regresión se entrenan suponiendo que los datos futuros se comportarán estadísticamente de manera similar a los datos usados para entrenar.

---

### Homocedasticidad

Otra suposición importante es la homocedasticidad.

La homocedasticidad significa que los errores o residuales del modelo tienen varianza constante.

Si el modelo se escribe como:

$$
y_i = f(\mathbf{x}_i)+\varepsilon_i
$$

entonces la homocedasticidad se expresa como:

$$
\operatorname{Var}(\varepsilon_i)=\sigma^2
\qquad \text{para todo } i
$$

Esto significa que la dispersión del error es aproximadamente la misma para todas las observaciones.

El residual se define como:

$$
e_i = y_i-\hat{y}_i
$$

donde:

- $y_i$ es el valor real observado.
- $\hat{y}_i$ es el valor predicho por el modelo.
- $e_i$ es el error cometido en la observación $i$.

En un modelo homocedástico, los residuales deberían dispersarse de forma relativamente uniforme alrededor de cero.

Visualmente, si se grafican los residuales contra las predicciones, se espera algo como una nube con ancho aproximadamente constante.

En cambio, si la varianza de los residuales cambia según la región del espacio de entrada, aparece heterocedasticidad:

$$
\operatorname{Var}(\varepsilon_i)\neq\sigma^2
$$

En ese caso, algunas zonas del modelo presentan errores pequeños y otras zonas errores grandes.

Esto puede afectar la confiabilidad del modelo, especialmente cuando se interpretan intervalos de confianza, pruebas estadísticas o incertidumbre de predicción.

---

### Idea central de las suposiciones

Las suposiciones sobre modelos de regresión permiten definir el tipo de problema que se está resolviendo y las condiciones bajo las cuales el modelo puede interpretarse correctamente.

En resumen:

- $s$ indica cuántas variables de entrada tiene cada observación.
- $r$ indica cuántas variables de salida se desean predecir.
- $M$ indica cuántas observaciones están disponibles.
- SISO corresponde a una entrada y una salida.
- MISO corresponde a varias entradas y una salida.
- MIMO corresponde a varias entradas y varias salidas.
- La suposición i.i.d. indica que las observaciones son independientes y siguen la misma distribución.
- La homocedasticidad indica que los errores del modelo tienen varianza constante.

Estas ideas son necesarias antes de estudiar regresión lineal, mínimos cuadrados y evaluación de modelos, porque permiten entender qué tipo de relación se está modelando y qué condiciones estadísticas se están suponiendo.

```python
# ============================================================
# EJEMPLO INDEPENDIENTE: HOMOCEDASTICIDAD Y HETEROCEDASTICIDAD
# ============================================================

import numpy as np
import matplotlib.pyplot as plt


# ============================================================
# 1. FIJAR SEMILLA ALEATORIA
# ============================================================

# Esto permite que los resultados sean reproducibles.
# Cada vez que ejecutes el código, obtendrás la misma nube de puntos.
np.random.seed(42)


# ============================================================
# 2. GENERAR VARIABLE DE ENTRADA
# ============================================================

# Se generan 200 valores entre 0 y 10.
# Esta variable representa una entrada genérica x.
x = np.linspace(0, 10, 200)


# ============================================================
# 3. DEFINIR MODELO REAL
# ============================================================

# Suponemos una relación lineal verdadera:
# y = 2 + 3x + ruido
y_teorica = 2 + 3*x


# ============================================================
# 4. CASO HOMOCEDÁSTICO
# ============================================================

# En este caso, el ruido tiene varianza constante.
# La desviación estándar del ruido es siempre la misma.
ruido_homocedastico = np.random.normal(
    loc=0,
    scale=3,
    size=len(x)
)

# Salida observada con ruido homocedástico
y_homo = y_teorica + ruido_homocedastico

# En este ejemplo, asumimos que el modelo predice la tendencia verdadera.
y_pred_homo = y_teorica

# Residuales
residuales_homo = y_homo - y_pred_homo


# ============================================================
# 5. CASO HETEROCEDÁSTICO
# ============================================================

# En este caso, el ruido aumenta cuando x aumenta.
# Por eso la varianza del error no es constante.
ruido_heterocedastico = np.random.normal(
    loc=0,
    scale=0.5 + 0.5*x,
    size=len(x)
)

# Salida observada con ruido heterocedástico
y_hetero = y_teorica + ruido_heterocedastico

# El modelo predice la misma tendencia verdadera
y_pred_hetero = y_teorica

# Residuales
residuales_hetero = y_hetero - y_pred_hetero


# ============================================================
# 6. GRÁFICA DE RESIDUALES HOMOCEDÁSTICOS
# ============================================================

plt.figure(figsize=(8, 5))

plt.scatter(y_pred_homo, residuales_homo)

plt.axhline(y=0, linestyle='--')

plt.xlabel("Valores predichos $\\hat{y}$")
plt.ylabel("Residuales $e_i = y_i - \\hat{y}_i$")
plt.title("Caso homocedástico: varianza aproximadamente constante")

plt.grid(True)
plt.show()


# ============================================================
# 7. GRÁFICA DE RESIDUALES HETEROCEDÁSTICOS
# ============================================================

plt.figure(figsize=(8, 5))

plt.scatter(y_pred_hetero, residuales_hetero)

plt.axhline(y=0, linestyle='--')

plt.xlabel("Valores predichos $\\hat{y}$")
plt.ylabel("Residuales $e_i = y_i - \\hat{y}_i$")
plt.title("Caso heterocedástico: varianza no constante")

plt.grid(True)
plt.show()
```

$$
\mathbf{X} =
\begin{bmatrix}

\end{bmatrix}
$$

y una variable o conjunto de variables de salida:

$$
\mathbf{Y}
$$

El objetivo es construir una función:

$$
\hat{\mathbf{Y}} = f(\mathbf{X})
$$

que aproxime lo mejor posible la salida real $\mathbf{Y}$.

Sin embargo, antes de ajustar el modelo aparecen varias preguntas fundamentales.

---

### 1. ¿Qué medidas de incertidumbre se pueden utilizar para evaluar la relación entre variables?

Para estudiar si existe interacción entre dos variables aleatorias, se pueden utilizar medidas estadísticas que cuantifican variabilidad, asociación o dependencia.

Algunas de las más importantes son:

#### Varianza

La varianza mide qué tanto se dispersa una variable respecto a su media:

$$
\operatorname{Var}(X)
=
\mathbb{E}\left[(X-\mathbb{E}[X])^2\right]
$$

Si la varianza es grande, los valores de la variable están muy dispersos. Si la varianza es pequeña, los valores están más concentrados alrededor de la media.

En regresión, la varianza permite entender qué tanta variabilidad existe en las variables antes de intentar modelarlas.

---

#### Covarianza

La covarianza mide si dos variables cambian juntas:

$$
\operatorname{Cov}(X,Y)
=
\mathbb{E}\left[(X-\mathbb{E}[X])(Y-\mathbb{E}[Y])\right]
$$

Si la covarianza es positiva, cuando $X$ aumenta, $Y$ tiende a aumentar.

Si la covarianza es negativa, cuando $X$ aumenta, $Y$ tiende a disminuir.

Si la covarianza es cercana a cero, no hay una relación lineal clara entre ambas variables.

La covarianza permite detectar dependencia lineal, pero su magnitud depende de las unidades de las variables.

---

#### Correlación

Para eliminar el efecto de las unidades, se utiliza la correlación:

$$
\rho_{XY}
=
\frac{\operatorname{Cov}(X,Y)}
{\sigma_X\sigma_Y}
$$

donde $\sigma_X$ y $\sigma_Y$ son las desviaciones estándar de $X$ y $Y$.

La correlación toma valores entre $-1$ y $1$:

$$
-1 \leq \rho_{XY} \leq 1
$$

- Si $\rho_{XY}\approx 1$, existe una relación lineal positiva fuerte.
- Si $\rho_{XY}\approx -1$, existe una relación lineal negativa fuerte.
- Si $\rho_{XY}\approx 0$, no existe una relación lineal evidente.

En el contexto de regresión, la correlación ayuda a identificar qué variables de entrada están más asociadas con la salida.

---

### 2. ¿Todas las variables contribuyen al modelo?

No necesariamente. Un conjunto de datos puede contener muchas variables de entrada, pero no todas aportan información útil para predecir la salida.

Supongamos un modelo lineal múltiple:

$$
\hat{Y}
=
\beta_0
+
\beta_1X_1
+
\beta_2X_2
+
\cdots
+
\beta_sX_s
$$

Cada coeficiente $\beta_j$ mide la contribución de la variable $X_j$ al modelo, manteniendo las demás variables constantes.

Si un coeficiente es grande en magnitud, la variable puede tener una influencia importante sobre la predicción.

Si un coeficiente es cercano a cero, la variable puede aportar poca información.

Sin embargo, la interpretación de los coeficientes depende de la escala de las variables. Por eso muchas veces se estandarizan los datos antes de comparar contribuciones.

---

### 3. ¿Cómo evaluar la contribución de cada variable al modelo?

#### Análisis de correlación

Se calcula la correlación entre cada variable de entrada y la salida:

$$
\rho_{X_jY}
$$

Esto permite observar qué variables tienen mayor relación lineal individual con la salida.

---

#### Magnitud de los coeficientes

En un modelo lineal, se analizan los coeficientes:

$$

$$

Si las variables están en escalas comparables, la magnitud de cada coeficiente puede indicar qué variable tiene mayor peso en la predicción.

---

También se puede entrenar el modelo con todas las variables y luego quitar una variable para observar si el error aumenta.

Si al eliminar $X_j$ el error del modelo aumenta mucho, significa que esa variable era importante.

---

$$
\beta_j = 0
$$

Cuando esto ocurre, el modelo está indicando que la variable asociada puede no ser necesaria para la predicción.

### 4. ¿Cómo evaluar la efectividad de diferentes modelos?

Para comparar modelos de regresión se utilizan métricas de ajuste, error y capacidad explicativa.

La idea fundamental es comparar los valores reales:

$$
\mathbf{y}
=
\begin{bmatrix}
y_1\\
y_2\\
\vdots\\
y_M
\end{bmatrix}
$$

con los valores predichos por el modelo:

$$
\hat{\mathbf{y}}
=
\begin{bmatrix}
\hat{y}_1\\
\hat{y}_2\\
\vdots\\
\hat{y}_M
\end{bmatrix}
$$

El error o residual de cada observación se define como:

$$
e_i = y_i-\hat{y}_i
$$

y el vector completo de errores es:

$$
\mathbf{e}
=
\mathbf{y}
-
\hat{\mathbf{y}}
=
\begin{bmatrix}
y_1-\hat{y}_1\\
y_2-\hat{y}_2\\
\vdots\\
y_M-\hat{y}_M
\end{bmatrix}
$$

Por tanto, evaluar el ajuste de un modelo equivale a medir qué tan grande es el vector de error $\mathbf{e}$.

De forma general, se puede usar una métrica de ajuste o distancia:

$$
d(\mathbf{x},\mathbf{y})
$$

donde $\mathbf{x}$ y $\mathbf{y}$ son dos vectores que se desean comparar.

En regresión, esta métrica suele aplicarse entre el vector real $\mathbf{y}$ y el vector estimado $\hat{\mathbf{y}}$.

---

#### Distancia $L_1$

Una de las métricas de ajuste más usadas es la distancia $L_1$:

$$
\| \mathbf{x} - \mathbf{y} \|_1
=
\sum_{i=1}^{n} |x_i - y_i|
$$

Aplicada al problema de regresión:

$$
\| \mathbf{y} - \hat{\mathbf{y}} \|_1
=
\sum_{i=1}^{M} |y_i - \hat{y}_i|
$$

Esta distancia suma los errores absolutos entre los valores reales y los valores predichos.

Su interpretación es directa: mide la cantidad total de error cometido por el modelo, sin permitir que los errores positivos y negativos se cancelen.

La distancia $L_1$ es menos sensible a valores atípicos que la distancia $L_2$, porque los errores no se elevan al cuadrado.

Por esta razón, puede ser conveniente en problemas donde existen datos ruidosos, mediciones extremas o valores atípicos.

---

#### Distancia $L_2$

Otra métrica ampliamente usada es la distancia $L_2$:

$$
\| \mathbf{x} - \mathbf{y} \|_2
=
\left(
\sum_{i=1}^{n}
(x_i-y_i)^2
\right)^{1/2}
$$

Aplicada al problema de regresión:

$$
\| \mathbf{y} - \hat{\mathbf{y}} \|_2
=
\left(
\sum_{i=1}^{M}
(y_i-\hat{y}_i)^2
\right)^{1/2}
$$

Esta distancia corresponde a la distancia euclidiana entre el vector de valores reales y el vector de valores predichos.

Geométricamente, si $\mathbf{y}$ y $\hat{\mathbf{y}}$ se interpretan como puntos en un espacio de dimensión $M$, la distancia $L_2$ mide la distancia en línea recta entre ellos.

La distancia $L_2$ penaliza más fuertemente los errores grandes, porque cada diferencia se eleva al cuadrado. Por ejemplo:

$$
2^2=4
$$

pero:

$$
10^2=100
$$

Esto significa que un error grande influye mucho más en una métrica basada en $L_2$ que en una métrica basada en $L_1$.

Por esta razón, las métricas basadas en $L_2$ son útiles cuando se quiere castigar fuertemente las predicciones muy alejadas del valor real.

---

#### Error cuadrático medio

Una de las métricas más usadas en regresión es el error cuadrático medio, conocido como MSE:

$$
MSE
=
\frac{1}{M}
\sum_{i=1}^{M}
(y_i-\hat{y}_i)^2
$$

El MSE mide el promedio de los errores al cuadrado.

Esta métrica está directamente relacionada con la distancia $L_2$ al cuadrado, porque usa términos de la forma:

$$
(y_i-\hat{y}_i)^2
$$

Por tanto, el MSE penaliza fuertemente errores grandes.

Si el MSE es pequeño, significa que, en promedio, las predicciones están cerca de los valores reales.

---

#### Raíz del error cuadrático medio

La raíz del error cuadrático medio, RMSE, se define como:

$$
RMSE
=
\sqrt{MSE}
$$

Es decir:

$$
RMSE
=
\sqrt{
\frac{1}{M}
\sum_{i=1}^{M}
(y_i-\hat{y}_i)^2
}
$$

El RMSE tiene las mismas unidades que la variable de salida, por lo que suele ser más interpretable que el MSE.

Por ejemplo, si la variable $Y$ representa ventas, entonces el RMSE también se interpreta en unidades de ventas.

---

#### Error absoluto medio

El error absoluto medio, MAE, se define como:

$$
MAE
=
\frac{1}{M}
\sum_{i=1}^{M}
|y_i-\hat{y}_i|
$$

El MAE mide el promedio de los errores absolutos.

Esta métrica está relacionada con la distancia $L_1$, porque usa términos de la forma:

$$
|y_i-\hat{y}_i|
$$

El MAE es menos sensible a errores extremos que el MSE, ya que no eleva los errores al cuadrado.

Por esta razón, puede ser útil cuando se quiere una medida de error más robusta frente a valores atípicos.

---

#### Coeficiente de determinación

El coeficiente de determinación, $R^2$, se define como:

$$
R^2
=
1-
\frac{
\sum_{i=1}^{M}(y_i-\hat{y}_i)^2
}{
\sum_{i=1}^{M}(y_i-\bar{y})^2
}
$$

donde:

$$
\bar{y}
=
\frac{1}{M}
\sum_{i=1}^{M}
y_i
$$

es el valor promedio de la variable de salida.

El numerador:

$$
\sum_{i=1}^{M}(y_i-\hat{y}_i)^2
$$

representa la suma de errores cuadráticos del modelo.

El denominador:

$$
\sum_{i=1}^{M}(y_i-\bar{y})^2
$$

representa la variabilidad total de los datos respecto a su media.

Por tanto, $R^2$ mide qué proporción de la variabilidad de $Y$ logra explicar el modelo.

Si $R^2$ es cercano a 1, el modelo explica gran parte de la variabilidad de los datos.

Si $R^2$ es cercano a 0, el modelo no mejora mucho respecto a predecir simplemente la media de $Y$.

Si $R^2$ es negativo, el modelo puede estar funcionando peor que usar la media como predicción.

Por ejemplo, si $\beta_1>0$, entonces al aumentar $X_1$, la salida predicha tiende a aumentar.

Si $\beta_1<0$, entonces al aumentar $X_1$, la salida predicha tiende a disminuir.

Por eso la regresión lineal es útil no solo para predecir, sino también para explicar relaciones entre variables.

---

### Idea central

Para construir una dependencia estadística mediante regresión se deben responder cuatro preguntas:

1. ¿Existe relación estadística entre las variables?
2. ¿Todas las variables aportan información útil?
3. ¿Cómo se mide la contribución de cada variable?
4. ¿Qué tan efectivo e interpretable es el modelo?

Estas preguntas permiten pasar de una simple visualización de datos a la construcción de un modelo estadístico formal.

```python
import numpy as np
import matplotlib.pyplot as plt

# Valores para comparar distancias L1 y L2
x_vals = np.linspace(-3, 3, 100)

y_vals_L1 = np.abs(x_vals)   # L1: distancia absoluta
y_vals_L2 = x_vals**2        # L2: distancia cuadrática

plt.figure(figsize=(10, 6))

plt.plot(x_vals, y_vals_L1, color="red", label="Distancia L1 (|x|)")
plt.plot(x_vals, y_vals_L2, color="blue", label="Distancia L2 (x^2)")

plt.axhline(y=1, color="black", linestyle="-")

plt.legend(loc="upper right")
plt.title("Comparación de distancias L1 y L2")
plt.xlabel("Valores de X")
plt.ylabel("Distancia")

plt.show()
```

## Estimación de los parámetros del modelo de regresión lineal simple

En esta sección se presenta una primera interpretación directa de los parámetros de un modelo de regresión lineal simple.

El modelo lineal simple se escribe como:

$$
\hat{y} = \beta_0 + \beta_1 x
$$

donde:

- $\hat{y}$ es el valor predicho por el modelo.
- $x$ es la variable de entrada.
- $\beta_0$ es el intercepto de la recta.
- $\beta_1$ es la pendiente de la recta.

Si los datos observados contienen ruido, se puede escribir:

$$
y_i = \beta_0 + \beta_1 x_i + \varepsilon_i
$$

donde $\varepsilon_i$ representa el error, ruido o perturbación de la observación $i$.

En el código se genera artificialmente una recta verdadera:

$$
y = -1 + 2.5x
$$

Luego se agrega ruido aleatorio a cada punto:

$$
y_i = -1 + 2.5x_i + \varepsilon_i
$$

Por esta razón, los puntos observados no caen exactamente sobre la recta verdadera, sino alrededor de ella.

---

### Cálculo de pendiente e intercepto usando pares de puntos

Una recta queda completamente determinada por dos puntos. Si se tienen dos puntos:

$$
(x_i,y_i)
\qquad \text{y} \qquad
(x_j,y_j)
$$

la pendiente de la recta que pasa por ellos es:

$$
m_{ij}
=
\frac{y_j-y_i}{x_j-x_i}
$$

Una vez conocida la pendiente, el intercepto se obtiene despejando de la ecuación de la recta:

$$
y_i = b_{ij} + m_{ij}x_i
$$

Por tanto:

$$
b_{ij}
=
y_i-m_{ij}x_i
$$

Cada par de puntos permite construir una recta diferente:

$$
\hat{y}_{ij}
=
b_{ij}
+
m_{ij}x
$$

En la gráfica, las rectas punteadas corresponden a las rectas obtenidas a partir de diferentes pares de puntos seleccionados aleatoriamente.

---

### Promedio de parámetros

Si se seleccionan $K$ pares de puntos, se obtienen $K$ pendientes:

$$
m_1,m_2,\dots,m_K
$$

y $K$ interceptos:

$$
b_1,b_2,\dots,b_K
$$

El código calcula la pendiente promedio:

$$
\bar{m}
=
\frac{1}{K}
\sum_{k=1}^{K}
m_k
$$

y el intercepto promedio:

$$
\bar{b}
=
\frac{1}{K}
\sum_{k=1}^{K}
b_k
$$

Con estos valores se construye una recta promedio:

$$
\hat{y}
=
\bar{b}
+
\bar{m}x
$$

Esta recta aparece en color rojo en la gráfica.

---

### Interpretación de la gráfica

En la gráfica se observan varios elementos:

- La recta verde representa la recta verdadera usada para generar los datos.
- Los puntos azules representan las muestras observadas con ruido.
- Las rectas punteadas representan rectas calculadas usando diferentes pares de puntos.
- Los puntos resaltados con borde negro son los pares usados para calcular cada recta.
- La recta roja representa la recta obtenida al promediar las pendientes e interceptos de las rectas punteadas.

La idea principal es mostrar que, si los datos siguen aproximadamente una relación lineal, es posible estimar una recta a partir de ellos. Sin embargo, estimar la recta usando solo pares de puntos es un método inestable, porque depende mucho de qué puntos se seleccionen y de cuánto ruido tengan.

Por esta razón, este enfoque sirve como interpretación inicial, pero no como método óptimo.

El método formal de regresión lineal buscará estimar los parámetros usando todos los puntos al mismo tiempo, minimizando el error total del modelo.

```python
import numpy as np
import matplotlib.pyplot as plt

# Semilla para reproducibilidad
np.random.seed(42)

# Generar una muestra de puntos a lo largo de una recta verdadera con ruido
n_puntos = 50
x = np.linspace(0, 10, n_puntos)

pendiente_verdadera = 2.5
intercepto_verdadero = -1

ruido = np.random.normal(0, 1, n_puntos)
y = pendiente_verdadera * x + intercepto_verdadero + ruido

# Seleccionar algunos pares de puntos para estimar pendiente e intercepto
n_pares = 5
pares_idx = np.random.choice(range(n_puntos), (n_pares, 2), replace=False)

# Calcular pendientes e interceptos para los pares seleccionados
pendientes = []
interceptos = []

for idx1, idx2 in pares_idx:
    pendiente = (y[idx2] - y[idx1]) / (x[idx2] - x[idx1])
    intercepto = y[idx1] - pendiente * x[idx1]

    pendientes.append(pendiente)
    interceptos.append(intercepto)

# Calcular pendiente e intercepto promedio
pendiente_promedio = np.mean(pendientes)
intercepto_promedio = np.mean(interceptos)

# Crear figura
plt.figure(figsize=(8, 6))

# Graficar la recta verdadera
plt.plot(
    x,
    pendiente_verdadera * x + intercepto_verdadero,
    label="Recta Verdadera",
    color="green"
)

# Graficar los puntos con ruido
plt.scatter(
    x,
    y,
    label="Muestras con ruido",
    color="blue",
    s=20
)

# Colores para las rectas de los pares de puntos
colores = plt.cm.rainbow(np.linspace(0, 1, n_pares))

# Graficar las rectas estimadas por pares de puntos
for (idx1, idx2), pendiente, intercepto, color in zip(
    pares_idx,
    pendientes,
    interceptos,
    colores
):
    plt.plot(
        x,
        pendiente * x + intercepto,
        linestyle="--",
        color=color,
        alpha=0.6
    )

    plt.scatter(
        [x[idx1], x[idx2]],
        [y[idx1], y[idx2]],
        color=color,
        edgecolor="black",
        zorder=5
    )

# Graficar la recta con parámetros promedio
plt.plot(
    x,
    pendiente_promedio * x + intercepto_promedio,
    label="Recta con Parámetros Promedio",
    color="red"
)

# Etiquetas y leyenda
plt.title("Regresión Lineal a partir de parámetros promedio")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()

plt.show()
```

### Segundo enfoque: Cálculo empírico de los parámetros de una regresión lineal:

En este segundo enfoque se interpreta la regresión lineal como un proceso de búsqueda empírica de los mejores parámetros de una recta.

El modelo lineal simple se escribe como:

$$
\hat{y} = \theta_0 + \theta_1 x
$$

donde:

- $\hat{y}$ es el valor predicho por el modelo.
- $x$ es la variable de entrada.
- $\theta_0$ es el intercepto de la recta.
- $\theta_1$ es la pendiente de la recta.

El objetivo es encontrar los valores de $\theta_0$ y $\theta_1$ que hagan que la recta predicha esté lo más cerca posible de los datos observados.

Si se tienen $M$ observaciones:

$$
(x_1,y_1),(x_2,y_2),\dots,(x_M,y_M)
$$

entonces para cada observación el modelo predice:

$$
\hat{y}_i = \theta_0 + \theta_1 x_i
$$

El error de predicción para cada punto es:

$$
e_i = y_i - \hat{y}_i
$$

es decir:

$$
e_i = y_i - (\theta_0 + \theta_1 x_i)
$$

Para medir el error total del modelo se usa el error cuadrático medio, MSE:

$$
MSE(\theta_0,\theta_1)
=
\frac{1}{M}
\sum_{i=1}^{M}
\left(
y_i - \hat{y}_i
\right)^2
$$

Sustituyendo la expresión de la predicción:

$$
MSE(\theta_0,\theta_1)
=
\frac{1}{M}
\sum_{i=1}^{M}
\left(
y_i - (\theta_0+\theta_1x_i)
\right)^2
$$

Esta función mide qué tan buena es una recta específica. Cada combinación de valores $(\theta_0,\theta_1)$ produce una recta diferente y, por tanto, un error diferente.

---

#### Interpretación geométrica

El parámetro $\theta_0$ desplaza la recta hacia arriba o hacia abajo. Por eso se llama intercepto.

El parámetro $\theta_1$ controla la inclinación de la recta. Si $\theta_1$ aumenta, la recta se vuelve más inclinada positivamente. Si $\theta_1$ disminuye, la recta puede volverse más plana o incluso tener pendiente negativa.

Por tanto, ajustar una regresión lineal puede verse geométricamente como mover y rotar una recta hasta encontrar una posición en la que la distancia entre la recta y los puntos sea mínima.

La idea de “rotar la varita” hace referencia precisamente a modificar la pendiente $\theta_1$ y el intercepto $\theta_0$ hasta que la recta se ajuste mejor a los datos.

---

#### Interpretación computacional

En el código se generan datos artificiales a partir de una recta verdadera:

$$
y = 2 + 3x
$$

Luego se agrega ruido aleatorio:

$$
y_i = 2 + 3x_i + \varepsilon_i
$$

donde $\varepsilon_i$ representa el ruido de cada observación.

Después, mediante controles interactivos, el usuario modifica manualmente:

$$
\theta_0
$$

y

$$
\theta_1
$$

para construir una recta candidata:

$$
\hat{y} = \theta_0 + \theta_1 x
$$

El código calcula el error cuadrático medio entre la recta propuesta por el usuario y los datos con ruido:

$$
MSE =
\frac{1}{M}
\sum_{i=1}^{M}
(y_i-\hat{y}_i)^2
$$

La recta que produzca un MSE más bajo será una mejor aproximación a los datos.

---

#### Diferencia con el primer enfoque

En el primer enfoque se tomaban varios pares de puntos, se calculaban varias rectas y luego se promediaban sus parámetros.

En este segundo enfoque, en cambio, se plantea directamente la idea de buscar la recta que minimice una función de error.

Este enfoque todavía es manual o heurístico, porque el usuario mueve los parámetros con controles interactivos. Sin embargo, introduce la idea central de la optimización:

$$
\min_{\theta_0,\theta_1}
MSE(\theta_0,\theta_1)
$$

Es decir, se busca encontrar los parámetros que hacen mínimo el error del modelo.

Más adelante, este procedimiento se formaliza mediante métodos como mínimos cuadrados ordinarios, donde los mejores parámetros se calculan matemáticamente usando todos los datos al mismo tiempo.

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import FloatSlider, Checkbox, interact
from sklearn.metrics import mean_squared_error

np.random.seed(42)

# Función para generar datos de una línea recta y añadir ruido
def generate_data(theta0, theta1, sigma):
    n_points = 100
    np.random.seed(42)

    X = 10 * np.random.rand(n_points, 1)
    y_original = theta1 * X + theta0

    noise = sigma * np.random.randn(n_points, 1)
    y_noisy = y_original + noise

    return X, y_original, y_noisy


class LinearDataVisualizer:
    def __init__(self, theta0, theta1, sigma):
        self.theta0 = theta0
        self.theta1 = theta1
        self.sigma = sigma

    def fit_user_line(self, theta0_user, theta1_user, show_original):
        X, y_original, y_noisy = generate_data(
            self.theta0,
            self.theta1,
            self.sigma
        )

        y_user = theta1_user * X + theta0_user
        mse = mean_squared_error(y_noisy, y_user)

        # Ordenar X para que las líneas se vean continuas
        idx = np.argsort(X[:, 0])
        X_sorted = X[idx]
        y_original_sorted = y_original[idx]
        y_user_sorted = y_user[idx]

        plt.figure(figsize=(10, 6))

        if show_original:
            plt.plot(
                X_sorted,
                y_original_sorted,
                color="green",
                label="Línea original"
            )

        plt.scatter(
            X,
            y_noisy,
            color="red",
            s=15,
            label="Datos con ruido"
        )

        plt.plot(
            X_sorted,
            y_user_sorted,
            "b--",
            label=f"Línea ajustada (Error = {mse:.4f})"
        )

        plt.xlabel("x", fontsize=14)
        plt.ylabel("y", fontsize=14)
        plt.title("Ajuste de la línea por el usuario")
        plt.ylim((0, 35))
        plt.legend()
        plt.show()


visualizer = LinearDataVisualizer(theta0=2, theta1=3, sigma=0.5)

interact(
    visualizer.fit_user_line,
    theta0_user=FloatSlider(
        min=-10,
        max=10,
        step=0.1,
        value=1,
        description="Intercepto"
    ),
    theta1_user=FloatSlider(
        min=-10,
        max=10,
        step=0.1,
        value=2,
        description="Pendiente"
    ),
    show_original=Checkbox(
        value=False,
        description="Mostrar línea original"
    )
);

```

### Tercer enfoque: Estimación óptima por $\ell_2$

En este tercer enfoque se estima la recta de regresión lineal de manera óptima usando el criterio de mínimos cuadrados. A diferencia del enfoque anterior, donde el usuario modificaba manualmente la pendiente y el intercepto, ahora el modelo calcula automáticamente los parámetros que minimizan el error cuadrático total.

El modelo de regresión lineal simple se escribe como:

$$
\hat{y}_i = \theta_0 + \theta_1 x_i
$$

donde:

- $\hat{y}_i$ es el valor predicho para la observación $i$.
- $x_i$ es la variable de entrada.
- $\theta_0$ es el intercepto.
- $\theta_1$ es la pendiente.

Los datos observados con ruido se pueden escribir como:

$$
y_i = \theta_0 + \theta_1 x_i + \varepsilon_i
$$

donde $\varepsilon_i$ representa el ruido, error o perturbación de la observación.

---

#### Error del modelo

Para cada punto, el error o residual se define como:

$$
e_i = y_i - \hat{y}_i
$$

Sustituyendo la predicción del modelo:

$$
e_i = y_i - (\theta_0 + \theta_1x_i)
$$

El objetivo de la regresión lineal por mínimos cuadrados es encontrar los valores de $\theta_0$ y $\theta_1$ que hagan que todos los errores sean pequeños al mismo tiempo.

Para medir el error global se usa la norma $\ell_2$ al cuadrado:

$$
J(\theta_0,\theta_1)
=
\sum_{i=1}^{M}
\left(
y_i - \hat{y}_i
\right)^2
$$

o equivalentemente:

$$
J(\theta_0,\theta_1)
=
\sum_{i=1}^{M}
\left(
y_i - (\theta_0+\theta_1x_i)
\right)^2
$$

Esta función se llama función de costo o función objetivo.

---

#### ¿Por qué se llama estimación por $\ell_2$?

El vector de errores se puede escribir como:

$$
\mathbf{e}
=
\mathbf{y}
-
\hat{\mathbf{y}}
$$

donde:

$$
\mathbf{y}
=
\begin{bmatrix}
y_1\\
y_2\\
\vdots\\
y_M
\end{bmatrix}
,
\qquad
\hat{\mathbf{y}}
=
\begin{bmatrix}
\hat{y}_1\\
\hat{y}_2\\
\vdots\\
\hat{y}_M
\end{bmatrix}
$$

La norma $\ell_2$ del error es:

$$
\|\mathbf{e}\|_2
=
\left(
\sum_{i=1}^{M}
e_i^2
\right)^{1/2}
$$

Por tanto, minimizar el error cuadrático total equivale a minimizar:

$$
\|\mathbf{y}-\hat{\mathbf{y}}\|_2^2
$$

Es decir:

$$
\min_{\theta_0,\theta_1}
\|\mathbf{y}-\hat{\mathbf{y}}\|_2^2
$$

Como el error se eleva al cuadrado, los errores grandes son penalizados con mayor fuerza. Por esta razón, el método $\ell_2$ intenta evitar que existan puntos demasiado alejados de la recta estimada.

---

#### Forma matricial del modelo

Para escribir el modelo en forma matricial, se agrega una columna de unos para representar el intercepto.

Si se tienen $M$ observaciones:

$$
(x_1,y_1),(x_2,y_2),\dots,(x_M,y_M)
$$

la matriz de diseño se define como:

$$
\mathbf{X}_b =
\begin{bmatrix}
1 & x_1\\
1 & x_2\\
\vdots & \vdots\\
1 & x_M
\end{bmatrix}
\in \mathbb{R}^{M\times 2}
$$

La primera columna de unos permite incluir el intercepto $\theta_0$ dentro de la multiplicación matricial.

El vector de parámetros es:

$$
\boldsymbol{\theta}
=
\begin{bmatrix}
\theta_0\\
\theta_1
\end{bmatrix}
\in \mathbb{R}^{2\times 1}
$$

Entonces las predicciones se escriben como:

$$
\hat{\mathbf{y}}
=
\mathbf{X}_b\boldsymbol{\theta}
$$

Revisando las dimensiones:

$$
\mathbf{X}_b \in \mathbb{R}^{M\times 2}
$$

$$
\boldsymbol{\theta}\in \mathbb{R}^{2\times 1}
$$

por tanto:

$$
\hat{\mathbf{y}}
=
\mathbf{X}_b\boldsymbol{\theta}
\in
\mathbb{R}^{M\times 1}
$$

Esto coincide con el tamaño del vector real:

$$
\mathbf{y}\in\mathbb{R}^{M\times 1}
$$

De forma explícita, la multiplicación matricial produce:

$$
\begin{bmatrix}
\hat{y}_1\\
\hat{y}_2\\
\vdots\\
\hat{y}_M
\end{bmatrix}
=
\begin{bmatrix}
1 & x_1\\
1 & x_2\\
\vdots & \vdots\\
1 & x_M
\end{bmatrix}
\begin{bmatrix}
\theta_0\\
\theta_1
\end{bmatrix}
=
\begin{bmatrix}
\theta_0+\theta_1x_1\\
\theta_0+\theta_1x_2\\
\vdots\\
\theta_0+\theta_1x_M
\end{bmatrix}
$$

Así, la forma matricial resume todas las predicciones del modelo en una sola expresión.

---

#### Problema de optimización

El problema de mínimos cuadrados se escribe como:

$$
\min_{\boldsymbol{\theta}}
\|\mathbf{y}-\mathbf{X}_b\boldsymbol{\theta}\|_2^2
$$

Esta expresión significa:

> encontrar el vector de parámetros $\boldsymbol{\theta}$ que hace mínima la distancia euclidiana cuadrática entre los valores reales y los valores predichos.

También se puede escribir como:

$$
\min_{\boldsymbol{\theta}}
(\mathbf{y}-\mathbf{X}_b\boldsymbol{\theta})^T
(\mathbf{y}-\mathbf{X}_b\boldsymbol{\theta})
$$

Esta forma deja claro que se está calculando el producto interno del vector de error consigo mismo:

$$
\|\mathbf{e}\|_2^2 = \mathbf{e}^T\mathbf{e}
$$

Geométricamente, el modelo busca la recta cuya distancia vertical total a los puntos sea mínima en sentido cuadrático.

---

#### Solución óptima

La solución óptima de mínimos cuadrados satisface las ecuaciones normales:

$$
\mathbf{X}_b^T\mathbf{X}_b\boldsymbol{\theta}
=
\mathbf{X}_b^T\mathbf{y}
$$

Si $\mathbf{X}_b^T\mathbf{X}_b$ es invertible, entonces:

$$
\boldsymbol{\theta}
=
(\mathbf{X}_b^T\mathbf{X}_b)^{-1}
\mathbf{X}_b^T\mathbf{y}
$$

Esta es la solución cerrada de la regresión lineal por mínimos cuadrados ordinarios.

La interpretación de esta solución es que se está buscando el vector de parámetros que genera la proyección más cercana de $\mathbf{y}$ sobre el espacio generado por las columnas de $\mathbf{X}_b$.

En la práctica, librerías como `sklearn` no siempre calculan explícitamente la inversa, porque invertir matrices puede ser numéricamente inestable. En su lugar, usan métodos más estables como descomposiciones matriciales.

---

#### ¿Qué hace `LinearRegression` de sklearn?

La clase `LinearRegression` de `sklearn.linear_model` ajusta un modelo lineal de la forma:

$$
\hat{y} = \theta_0 + \theta_1x
$$

Internamente busca los parámetros que minimizan:

$$
\sum_{i=1}^{M}(y_i-\hat{y}_i)^2
$$

Es decir, resuelve un problema de mínimos cuadrados ordinarios.

Cuando se ejecuta:

$$
\texttt{model.fit(X, y\_noisy)}
$$

el método `fit` ajusta el modelo a los datos. Es decir, toma la matriz de entrada $\mathbf{X}$ y el vector de salida observado $\mathbf{y}$, y calcula automáticamente los parámetros que minimizan el error cuadrático total.

En este caso, como el modelo es lineal simple, `sklearn` busca los valores óptimos de:

$$
\theta_0
\qquad \text{y} \qquad
\theta_1
$$

de tal forma que la recta:

$$
\hat{y} = \theta_0+\theta_1x
$$

sea la mejor aproximación a los datos ruidosos en el sentido de mínimos cuadrados.

Después del entrenamiento, los parámetros estimados quedan almacenados en el modelo.

El atributo:

$$
\texttt{model.intercept\_}
$$

contiene el intercepto estimado:

$$
\hat{\theta}_0
$$

El atributo:

$$
\texttt{model.coef\_}
$$

contiene la pendiente estimada:

$$
\hat{\theta}_1
$$

Luego, cuando se ejecuta:

$$
\texttt{y\_pred = model.predict(X)}
$$

el modelo usa los parámetros aprendidos para calcular las predicciones:

$$
\hat{y}_i = \hat{\theta}_0+\hat{\theta}_1x_i
$$

para cada dato de entrada $x_i$.

De forma vectorial:

$$
\hat{\mathbf{y}}
=
\mathbf{X}_b\hat{\boldsymbol{\theta}}
$$

donde:

$$
\hat{\boldsymbol{\theta}}
=
\begin{bmatrix}
\hat{\theta}_0\\
\hat{\theta}_1
\end{bmatrix}
$$

representa el vector de parámetros estimados por el modelo.

---

#### Interpretación de la gráfica

La gráfica obtenida compara tres elementos importantes:

- Los puntos rojos representan los datos observados con ruido.
- La línea verde representa la recta original usada para generar los datos.
- La línea azul punteada representa la recta estimada por el modelo OLS usando `sklearn`.

Los datos fueron generados a partir de una recta verdadera:

$$
y = 2 + 3x + \varepsilon
$$

donde $\varepsilon$ es ruido aleatorio.

Por tanto, la línea verde corresponde a la parte ideal del sistema:

$$
y = 2 + 3x
$$

mientras que los puntos rojos corresponden a las observaciones reales con ruido:

$$
y_i = 2 + 3x_i + \varepsilon_i
$$

La línea azul punteada corresponde a la estimación realizada por mínimos cuadrados ordinarios:

$$
\hat{y}
=
\hat{\theta}_0
+
\hat{\theta}_1x
$$

Si el ruido es pequeño y los datos siguen una relación aproximadamente lineal, la línea azul debería quedar muy cerca de la línea verde.

Esto significa que el método OLS logró recuperar de manera aproximada los parámetros verdaderos que generaron los datos.

---

#### Diferencia entre línea original y estimación OLS

La línea original es conocida porque en este ejemplo los datos fueron generados artificialmente. Es decir, el programador sabe que los parámetros verdaderos son:

$$
\theta_0=2
$$

$$
\theta_1=3
$$

Sin embargo, en un problema real no se conoce la línea original. Solo se observan los datos ruidosos.

Por eso, en un caso real, el objetivo de la regresión es estimar una línea que explique los datos observados lo mejor posible.

La línea OLS no intenta pasar exactamente por todos los puntos. En lugar de eso, busca minimizar la suma total de los errores cuadráticos:

$$
\sum_{i=1}^{M}
(y_i-\hat{y}_i)^2
$$

Por eso, aunque algunos puntos queden por encima de la recta y otros por debajo, la recta estimada es óptima bajo el criterio $\ell_2$.

---

#### Idea central del tercer enfoque

El tercer enfoque formaliza la regresión lineal como un problema de optimización:

$$
\min_{\theta_0,\theta_1}
\sum_{i=1}^{M}
\left(
y_i-(\theta_0+\theta_1x_i)
\right)^2
$$

Este enfoque es más robusto que seleccionar pares de puntos o mover manualmente una recta, porque utiliza todos los datos simultáneamente y encuentra los parámetros que minimizan el error total.

En resumen:

- El primer enfoque estima rectas usando pares de puntos.
- El segundo enfoque mueve manualmente la recta para reducir el error.
- El tercer enfoque calcula automáticamente la recta óptima usando mínimos cuadrados ordinarios.

Por esta razón, el tercer enfoque corresponde al método formal de regresión lineal clásica.

```python
# %% 
# Tercer enfoque: estimación óptima por mínimos cuadrados / OLS

import numpy as np
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Parámetros reales usados para generar los datos
theta0_real = 2
theta1_real = 3
sigma = 0.5

# Generar datos
np.random.seed(42)

n_points = 100
X = 10 * np.random.rand(n_points, 1)

y_original = theta1_real * X + theta0_real
ruido = sigma * np.random.randn(n_points, 1)
y_noisy = y_original + ruido

# Crear y entrenar modelo OLS
model = LinearRegression()
model.fit(X, y_noisy)

# Predicción
y_pred = model.predict(X)

# Parámetros estimados
theta0_estimado = model.intercept_[0]
theta1_estimado = model.coef_[0, 0]

# Métricas
mse = mean_squared_error(y_noisy, y_pred)
r2 = r2_score(y_noisy, y_pred)

print(f"Intercepto real:   {theta0_real:.4f}")
print(f"Intercepto OLS:    {theta0_estimado:.4f}")
print(f"Pendiente real:    {theta1_real:.4f}")
print(f"Pendiente OLS:     {theta1_estimado:.4f}")
print(f"MSE:               {mse:.4f}")
print(f"R²:                {r2:.4f}")

# Ordenar X para graficar líneas sin saltos
idx = np.argsort(X[:, 0])
X_sorted = X[idx]
y_original_sorted = y_original[idx]
y_pred_sorted = y_pred[idx]

# Gráfica
plt.figure(figsize=(10, 6))

plt.scatter(
    X,
    y_noisy,
    color="red",
    s=20,
    label="Datos con ruido"
)

plt.plot(
    X_sorted,
    y_original_sorted,
    color="green",
    linewidth=2,
    label="Línea original"
)

plt.plot(
    X_sorted,
    y_pred_sorted,
    color="blue",
    linestyle="--",
    linewidth=2,
    label="Estimación OLS"
)

plt.xlabel("$x$", fontsize=14)
plt.ylabel("$y$", fontsize=14)
plt.title("Regresión lineal por mínimos cuadrados")
plt.legend()
plt.grid(True)
plt.show()
```

## Evaluación de modelos de regresión lineal

Después de estimar los parámetros de una regresión, el siguiente problema es decidir si el modelo realmente es útil. La recta ajustada puede verse razonable en una gráfica, pero la evaluación formal exige medir el error y verificar si el modelo generaliza a datos que no fueron usados para entrenarlo.

Supongamos que se tienen M observaciones reales:

$$
y_1,y_2,\dots,y_M
$$

Y que el modelo produce las predicciones:

$$
\hat{y}_1,\hat{y}_2,\dots,\hat{y}_M
$$

El error de cada observación se llama residual:

$$
e_i = y_i - \hat{y}_i
$$

En forma vectorial:

$$
\mathbf{e}=\mathbf{y}-\hat{\mathbf{y}}
$$

La evaluación consiste en resumir ese vector de errores con una cantidad interpretable.

---

### Error absoluto medio

$$
MAE = \frac{1}{M}\sum_{i=1}^{M}|y_i-\hat{y}_i|
$$

Esta métrica mide el tamaño promedio del error sin importar el signo. Su ventaja es que se interpreta en las mismas unidades de la variable de salida.

---

### Error cuadrático medio

$$
MSE = \frac{1}{M}\sum_{i=1}^{M}(y_i-\hat{y}_i)^2
$$

Al elevar el residual al cuadrado, los errores grandes pesan mucho más. Por eso el MSE es sensible a valores atípicos.

---

### Raíz del error cuadrático medio

$$
RMSE = \sqrt{MSE}
$$

El RMSE conserva la penalización fuerte sobre errores grandes, pero vuelve a quedar en las unidades originales de y.

---

### Coeficiente de determinación

La variabilidad total de los datos se mide como:

$$
SS_{tot}=\sum_{i=1}^{M}(y_i-\bar{y})^2
$$

La variabilidad no explicada por el modelo se mide como:

$$
SS_{res}=\sum_{i=1}^{M}(y_i-\hat{y}_i)^2
$$

Entonces:

$$
R^2 = 1 - \frac{SS_{res}}{SS_{tot}}
$$

Si R^2 se acerca a 1, el modelo explica gran parte de la variación de y. Si se acerca a 0, el modelo no mejora mucho respecto a usar solamente la media.

---

### Entrenamiento y prueba

Un modelo puede ajustar muy bien los datos con los que fue entrenado y aun así fallar con datos nuevos. Por eso se divide el conjunto de datos en entrenamiento y prueba. La evaluación más importante no es el error de entrenamiento, sino el error de prueba.

```python
# Evaluación de una regresión lineal con entrenamiento y prueba

import numpy as np
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.model_selection import train_test_split

np.random.seed(42)

n_points = 140
X = 10 * np.random.rand(n_points, 1)
y_true = 2 + 3 * X
y = y_true + np.random.normal(0, 2.0, size=(n_points, 1))

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.25,
    random_state=42
)

model = LinearRegression()
model.fit(X_train, y_train)

y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)

def regression_report(y_real, y_pred, name):
    mae = mean_absolute_error(y_real, y_pred)
    mse = mean_squared_error(y_real, y_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_real, y_pred)
    print(f"{name}")
    print(f"  MAE  = {mae:.4f}")
    print(f"  MSE  = {mse:.4f}")
    print(f"  RMSE = {rmse:.4f}")
    print(f"  R²   = {r2:.4f}")
    print()

print(f"Intercepto estimado: {model.intercept_[0]:.4f}")
print(f"Pendiente estimada:  {model.coef_[0, 0]:.4f}\n")
regression_report(y_train, y_train_pred, "Métricas en entrenamiento")
regression_report(y_test, y_test_pred, "Métricas en prueba")

X_plot = np.linspace(0, 10, 200).reshape(-1, 1)
y_plot = model.predict(X_plot)

plt.figure(figsize=(13, 5))

plt.subplot(1, 2, 1)
plt.scatter(X_train, y_train, color="tab:blue", s=24, label="Entrenamiento")
plt.scatter(X_test, y_test, color="tab:red", s=34, label="Prueba")
plt.plot(X_plot, 2 + 3 * X_plot, color="green", linestyle="--", label="Relación real")
plt.plot(X_plot, y_plot, color="black", linewidth=2, label="Modelo estimado")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Ajuste y separación de datos")
plt.legend()
plt.grid(True)

plt.subplot(1, 2, 2)
residuals = y_test - y_test_pred
plt.scatter(y_test_pred, residuals, color="tab:purple", s=34)
plt.axhline(0, color="black", linestyle="--")
plt.xlabel("Predicción en prueba")
plt.ylabel("Residual")
plt.title("Gráfica de residuales")
plt.grid(True)

plt.tight_layout()
plt.show()
```

## Factores que influyen en la calidad de la estimación por OLS

El método OLS estima los parámetros minimizando la suma de errores cuadráticos:

$$
\min_{\boldsymbol{\theta}} \|\mathbf{y}-\mathbf{X}_b\boldsymbol{\theta}\|_2^2
$$

Su solución cerrada es:

$$
\boldsymbol{\theta}=(\mathbf{X}_b^T\mathbf{X}_b)^{-1}\mathbf{X}_b^T\mathbf{y}
$$

Esta expresión muestra que la calidad de la estimación depende tanto de los datos de salida como de la estructura de la matriz de entrada.

---

### Ruido

Si las observaciones se generan como:

$$
y_i = \theta_0 + \theta_1x_i + \varepsilon_i
$$

entonces el término de error representa variación que el modelo no puede explicar. Cuando la varianza del ruido aumenta, los puntos se dispersan más alrededor de la recta y la estimación de los parámetros se vuelve menos precisa.

---

### Valores atípicos

Un valor atípico es una observación que se aleja del patrón general. Como OLS usa errores cuadrados, un residual grande produce una penalización muy grande:

$$
e_i^2 \gg |e_i|
$$

Por eso, un solo punto muy alejado puede arrastrar la recta estimada.

---

### Colinealidad

En regresión con varias entradas, puede ocurrir que una variable sea casi combinación lineal de otra:

$$
X_2 \approx cX_1
$$

En ese caso, las columnas de la matriz de entrada contienen información repetida. Entonces X^T X puede volverse casi singular y la solución de OLS se vuelve inestable.

Una señal práctica de colinealidad es una correlación alta entre predictores:

$$
|\rho_{X_iX_j}| \approx 1
$$

Cuando esto ocurre, el modelo puede seguir prediciendo bien, pero los coeficientes individuales se vuelven difíciles de interpretar.

```python
# Ruido, valores atípicos y colinealidad en OLS

import numpy as np
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression, HuberRegressor

np.random.seed(42)

X_clean = np.linspace(0, 10, 80).reshape(-1, 1)
y_clean = 2 + 3 * X_clean.ravel() + np.random.normal(0, 1.0, X_clean.shape[0])

X_out = np.array([[8.2], [8.8], [9.3]])
y_out = np.array([5, 7, 4])

X_all = np.vstack([X_clean, X_out])
y_all = np.concatenate([y_clean, y_out])

ols_clean = LinearRegression().fit(X_clean, y_clean)
ols_outliers = LinearRegression().fit(X_all, y_all)
huber = HuberRegressor().fit(X_all, y_all)

print("Parámetros estimados")
print(f"OLS sin atípicos:  b0 = {ols_clean.intercept_:.3f}, b1 = {ols_clean.coef_[0]:.3f}")
print(f"OLS con atípicos:  b0 = {ols_outliers.intercept_:.3f}, b1 = {ols_outliers.coef_[0]:.3f}")
print(f"Huber robusto:     b0 = {huber.intercept_:.3f}, b1 = {huber.coef_[0]:.3f}")

X_plot = np.linspace(0, 10, 200).reshape(-1, 1)

x1 = np.random.normal(size=160)
x2 = 0.97 * x1 + 0.04 * np.random.normal(size=160)
x3 = np.random.normal(size=160)
X_multi = np.column_stack([x1, x2, x3])
correlation = np.corrcoef(X_multi, rowvar=False)

plt.figure(figsize=(13, 5))

plt.subplot(1, 2, 1)
plt.scatter(X_clean, y_clean, color="tab:blue", s=24, label="Datos normales")
plt.scatter(X_out, y_out, color="tab:red", s=60, label="Valores atípicos")
plt.plot(X_plot, ols_clean.predict(X_plot), color="green", label="OLS sin atípicos")
plt.plot(X_plot, ols_outliers.predict(X_plot), color="black", linestyle="--", label="OLS con atípicos")
plt.plot(X_plot, huber.predict(X_plot), color="tab:orange", linestyle=":", linewidth=3, label="Regresión robusta")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Sensibilidad frente a valores atípicos")
plt.legend()
plt.grid(True)

plt.subplot(1, 2, 2)
plt.imshow(correlation, vmin=-1, vmax=1, cmap="coolwarm")
plt.colorbar(label="Correlación")
plt.xticks([0, 1, 2], ["X1", "X2", "X3"])
plt.yticks([0, 1, 2], ["X1", "X2", "X3"])
for i in range(3):
    for j in range(3):
        plt.text(j, i, f"{correlation[i, j]:.2f}", ha="center", va="center")
plt.title("Detección visual de colinealidad")

plt.tight_layout()
plt.show()
```

## Métodos de regularización

La regularización se usa cuando se desea controlar la complejidad del modelo. En regresión lineal, la complejidad aparece principalmente en el tamaño de los coeficientes.

Un modelo lineal con coeficientes muy grandes puede ser sensible a pequeñas variaciones en los datos. La regularización evita esto agregando una penalización a la función de costo.

El problema general se puede escribir como:

$$
\min_{\boldsymbol{\theta}}
\left[
\sum_{i=1}^{M}(y_i-\hat{y}_i)^2
+
\alpha\,\Omega(\boldsymbol{\theta})
\right]
$$

Donde alpha controla la fuerza de la regularización. Si alpha es cero, se recupera OLS. Si alpha aumenta, los coeficientes quedan más restringidos.

---

### Ridge

Ridge usa penalización l2:

$$
\Omega(\boldsymbol{\theta})=\sum_{j=1}^{s}\theta_j^2
$$

Por tanto:

$$
J(\boldsymbol{\theta})=
\sum_{i=1}^{M}(y_i-\hat{y}_i)^2
+
\alpha\sum_{j=1}^{s}\theta_j^2
$$

La solución cerrada de Ridge es:

$$
\boldsymbol{\theta}_{ridge}=
(\mathbf{X}^T\mathbf{X}+\alpha\mathbf{I})^{-1}\mathbf{X}^T\mathbf{y}
$$

El término alpha I mejora la estabilidad numérica cuando X^T X está mal condicionada. Por eso Ridge es útil frente a colinealidad.

---

### Lasso

Lasso usa penalización l1:

$$
\Omega(\boldsymbol{\theta})=\sum_{j=1}^{s}|\theta_j|
$$

Entonces:

$$
J(\boldsymbol{\theta})=
\sum_{i=1}^{M}(y_i-\hat{y}_i)^2
+
\alpha\sum_{j=1}^{s}|\theta_j|
$$

La penalización l1 puede hacer que algunos coeficientes sean exactamente cero. Por eso Lasso se usa también como método de selección de variables.

---

### ElasticNet

ElasticNet combina l1 y l2:

$$
J(\boldsymbol{\theta})=
\sum_{i=1}^{M}(y_i-\hat{y}_i)^2
+
\alpha
\left[
\lambda\sum_{j=1}^{s}|\theta_j|
+(1-\lambda)\sum_{j=1}^{s}\theta_j^2
\right]
$$

El parámetro lambda controla la mezcla entre Lasso y Ridge. En sklearn, este parámetro se llama l1_ratio.

ElasticNet es útil cuando hay muchas variables correlacionadas y se desea una combinación entre estabilidad y selección de predictores.

```python
# Comparación visual entre OLS, Ridge, Lasso y ElasticNet

import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline

np.random.seed(42)

X, y = make_regression(
    n_samples=200,
    n_features=12,
    n_informative=4,
    noise=20,
    random_state=42
)

X[:, 4] = X[:, 0] + 0.05 * np.random.randn(X.shape[0])
X[:, 5] = X[:, 1] + 0.05 * np.random.randn(X.shape[0])

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.25,
    random_state=42
)

models = {
    "OLS": LinearRegression(),
    "Ridge": Ridge(alpha=20),
    "Lasso": Lasso(alpha=2, max_iter=10000),
    "ElasticNet": ElasticNet(alpha=1, l1_ratio=0.5, max_iter=10000),
}

summary = []
coefs = []

for name, estimator in models.items():
    pipe = make_pipeline(StandardScaler(), estimator)
    pipe.fit(X_train, y_train)
    y_pred = pipe.predict(X_test)

    reg = pipe.named_steps[list(pipe.named_steps.keys())[-1]]
    coef = reg.coef_
    coefs.append(coef)

    summary.append((
        name,
        np.sqrt(mean_squared_error(y_test, y_pred)),
        r2_score(y_test, y_pred),
        int(np.sum(np.abs(coef) > 1e-6)),
        np.linalg.norm(coef, ord=2),
        np.linalg.norm(coef, ord=1),
    ))

print("Modelo       RMSE      R²     coef no cero   ||coef||2   ||coef||1")
for item in summary:
    print(f"{item[0]:10s} {item[1]:8.3f} {item[2]:7.3f} {item[3]:13d} {item[4]:10.3f} {item[5]:10.3f}")

plt.figure(figsize=(13, 5))

plt.subplot(1, 2, 1)
for coef, name in zip(coefs, models.keys()):
    plt.plot(coef, marker="o", label=name)
plt.axhline(0, color="black", linewidth=1)
plt.xlabel("Índice del predictor")
plt.ylabel("Coeficiente estimado")
plt.title("Efecto de la regularización en los coeficientes")
plt.legend()
plt.grid(True)

plt.subplot(1, 2, 2)
model_names = [item[0] for item in summary]
nonzero = [item[3] for item in summary]
plt.bar(model_names, nonzero, color=["gray", "tab:blue", "tab:red", "tab:purple"])
plt.ylabel("Número de coeficientes no cero")
plt.title("Lasso y ElasticNet pueden seleccionar variables")
plt.grid(axis="y")

plt.tight_layout()
plt.show()
```

## Regresión simple no lineal

La regresión lineal simple supone que la salida puede aproximarse con una recta:

$$
\hat{y}=\theta_0+\theta_1x
$$

Sin embargo, muchas relaciones reales no son rectas. Pueden ser exponenciales, logarítmicas, polinomiales, periódicas o tener cambios locales.

En general, una relación no lineal se escribe como:

$$
y=f(x)+\varepsilon
$$

La pregunta es cómo aproximar f. Hay tres caminos frecuentes:

1. Transformar la variable para volver lineal el problema.
2. Crear nuevas características, por ejemplo x^2, x^3 o productos entre variables.
3. Usar modelos que ya son no lineales, como árboles, kernels, SVR o procesos Gaussianos.

---

### Linealización por transformación

Si la relación tiene forma exponencial:

$$
y = b e^{ax}
$$

al aplicar logaritmo natural:

$$
\ln(y)=\ln(b)+ax
$$

Si se define:

$$
z=\ln(y), \qquad \beta_0=\ln(b), \qquad \beta_1=a
$$

queda un modelo lineal:

$$
z=\beta_0+\beta_1x
$$

Esto permite ajustar una recta en la escala transformada y luego regresar a la escala original usando:

$$
\hat{y}=e^{\hat{z}}
$$

Este enfoque se usa cuando la transformación tiene sentido físico o estadístico. No se debe aplicar solo porque mejora una gráfica: también hay que interpretar qué significa transformar la variable de salida.

```python
# Linealización de un modelo exponencial mediante logaritmo natural

import numpy as np
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

np.random.seed(42)

X = np.linspace(0, 3, 120).reshape(-1, 1)
y_true = 2.0 * np.exp(0.9 * X.ravel())
y = y_true * np.exp(np.random.normal(0, 0.12, size=X.shape[0]))

y_log = np.log(y)

model_log = LinearRegression()
model_log.fit(X, y_log)

y_log_pred = model_log.predict(X)
y_pred = np.exp(y_log_pred)

print(f"Modelo linealizado: ln(y) = {model_log.intercept_:.4f} + {model_log.coef_[0]:.4f} x")
print(f"Parámetro b estimado: {np.exp(model_log.intercept_):.4f}")
print(f"Parámetro a estimado: {model_log.coef_[0]:.4f}")
print(f"R² en escala original: {r2_score(y, y_pred):.4f}")

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.scatter(X, y, color="tab:red", s=20, label="Datos observados")
plt.plot(X, y_true, color="green", label="Curva real")
plt.plot(X, y_pred, color="blue", linestyle="--", label="Ajuste linealizado")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Escala original")
plt.legend()
plt.grid(True)

plt.subplot(1, 2, 2)
plt.scatter(X, y_log, color="tab:purple", s=20, label="ln(y)")
plt.plot(X, y_log_pred, color="black", label="Recta ajustada")
plt.xlabel("x")
plt.ylabel("ln(y)")
plt.title("Escala transformada")
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

## Linealización por expansión polinomial

Otra forma de modelar relaciones no lineales consiste en transformar la entrada original x en varias características:

$$
\phi(x)=
\begin{bmatrix}
x & x^2 & x^3 & \cdots & x^d
\end{bmatrix}
$$

Aunque el modelo sigue siendo lineal en los parámetros, ahora puede representar curvas:

$$
\hat{y}=\theta_0+\theta_1x+\theta_2x^2+\cdots+\theta_dx^d
$$

La palabra lineal aquí se refiere a los parámetros, no a la forma geométrica de la curva. El modelo es lineal en theta, pero no necesariamente lineal en x.

El grado d controla la flexibilidad:

- Si d es muy bajo, el modelo puede subajustar.
- Si d es adecuado, captura la tendencia principal.
- Si d es muy alto, puede sobreajustar y seguir el ruido.

Por esta razón, las expansiones polinomiales se suelen combinar con regularización.

```python
# Regresión polinomial: subajuste, ajuste razonable y sobreajuste

import numpy as np
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression, Ridge
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error

np.random.seed(42)

X = np.linspace(-3, 3, 80).reshape(-1, 1)
y_true = 0.5 * X.ravel()**3 - 2 * X.ravel() + 1
y = y_true + np.random.normal(0, 2.0, size=X.shape[0])

X_plot = np.linspace(-3.2, 3.2, 300).reshape(-1, 1)
degrees = [1, 3, 12]

plt.figure(figsize=(12, 4))

for k, degree in enumerate(degrees, 1):
    if degree == 12:
        model = make_pipeline(PolynomialFeatures(degree), Ridge(alpha=2.0))
        title = "Grado 12 con Ridge"
    else:
        model = make_pipeline(PolynomialFeatures(degree), LinearRegression())
        title = f"Grado {degree}"

    model.fit(X, y)
    y_plot = model.predict(X_plot)
    y_fit = model.predict(X)
    rmse = np.sqrt(mean_squared_error(y, y_fit))

    plt.subplot(1, 3, k)
    plt.scatter(X, y, color="tab:red", s=18, label="Datos")
    plt.plot(X_plot, y_plot, color="blue", label="Modelo")
    plt.plot(X_plot, 0.5 * X_plot.ravel()**3 - 2 * X_plot.ravel() + 1, color="green", linestyle="--", label="Función real")
    plt.title(f"{title}\nRMSE entrenamiento = {rmse:.2f}")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.grid(True)

plt.tight_layout()
plt.show()
```

## Linealización por métodos heurísticos: árboles y bosques aleatorios

Los árboles de decisión no ajustan una ecuación global como una recta o un polinomio. En cambio, dividen el espacio de entrada en regiones y asignan una predicción a cada región.

Para regresión, un árbol busca particiones del tipo:

$$
x_j \leq c
$$

donde x_j es una variable de entrada y c es un umbral. Cada partición divide los datos en dos grupos.

La pregunta del árbol es: ¿qué variable y qué umbral reducen más el error dentro de los grupos?

Si un nodo contiene un conjunto de índices R, la predicción constante en ese nodo es el promedio:

$$
\hat{y}_R=\frac{1}{|R|}\sum_{i\in R}y_i
$$

El error dentro del nodo se puede medir como:

$$
\sum_{i\in R}(y_i-\hat{y}_R)^2
$$

Un árbol escoge divisiones que reduzcan esta suma. Si el árbol crece demasiado, puede crear regiones muy pequeñas y memorizar ruido.

---

### Bosques aleatorios

Un bosque aleatorio entrena muchos árboles. Cada árbol ve una muestra aleatoria de los datos y, en cada división, puede considerar solo un subconjunto de variables.

La predicción final es el promedio de las predicciones individuales:

$$
\hat{y}=\frac{1}{B}\sum_{b=1}^{B}T_b(x)
$$

Donde T_b es el árbol b y B es el número de árboles.

El bosque reduce la varianza porque promedia muchos modelos inestables. Un árbol individual puede cambiar mucho si cambian los datos; el promedio de muchos árboles suele ser más estable.

```python
# Árboles de decisión y bosques aleatorios para regresión no lineal

import numpy as np
import matplotlib.pyplot as plt

from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split

np.random.seed(42)

X = np.linspace(0, 10, 180).reshape(-1, 1)
y_true = np.sin(X.ravel()) + 0.25 * X.ravel()
y = y_true + np.random.normal(0, 0.25, size=X.shape[0])

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.3,
    random_state=42
)

models = {
    "Árbol poco profundo": DecisionTreeRegressor(max_depth=3, random_state=42),
    "Árbol profundo": DecisionTreeRegressor(max_depth=None, random_state=42),
    "Bosque aleatorio": RandomForestRegressor(n_estimators=120, max_depth=6, random_state=42),
}

X_plot = np.linspace(0, 10, 400).reshape(-1, 1)

plt.figure(figsize=(12, 4))
for k, (name, model) in enumerate(models.items(), 1):
    model.fit(X_train, y_train)
    y_plot = model.predict(X_plot)
    y_test_pred = model.predict(X_test)
    rmse = np.sqrt(mean_squared_error(y_test, y_test_pred))

    plt.subplot(1, 3, k)
    plt.scatter(X_train, y_train, color="tab:blue", s=16, alpha=0.7, label="Entrenamiento")
    plt.scatter(X_test, y_test, color="tab:red", s=18, alpha=0.8, label="Prueba")
    plt.plot(X_plot, y_plot, color="black", linewidth=2, label="Predicción")
    plt.plot(X_plot, np.sin(X_plot.ravel()) + 0.25 * X_plot.ravel(), color="green", linestyle="--", label="Función real")
    plt.title(f"{name}\nRMSE prueba = {rmse:.3f}")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.grid(True)

plt.tight_layout()
plt.show()
```

## Métodos avanzados: kernels

Algunas relaciones son difíciles de representar directamente en el espacio original. Una estrategia consiste en transformar los datos mediante una función phi:

$$
\phi(x)
$$

y luego aplicar un modelo lineal en ese nuevo espacio.

Por ejemplo, si:

$$
\phi(x)=\begin{bmatrix}x & x^2\end{bmatrix}
$$

un modelo lineal en phi(x) equivale a un modelo cuadrático en x.

El problema es que en espacios de muchas dimensiones calcular phi(x) explícitamente puede ser costoso. El truco kernel permite trabajar con productos internos en el espacio transformado sin construir la transformación completa:

$$
K(x_i,x_j)=\langle \phi(x_i),\phi(x_j)\rangle
$$

La función K mide similitud. Si dos puntos son parecidos, su kernel es grande; si son distintos, su kernel es pequeño.

---

### Kernel polinomial

$$
K(x_i,x_j)=(x_i^Tx_j+c)^d
$$

Permite representar interacciones polinomiales de grado d.

---

### Kernel RBF

$$
K(x_i,x_j)=\exp(-\gamma\|x_i-x_j\|^2)
$$

Este kernel mide cercanía local. Si dos puntos están cerca, el valor del kernel es alto. Si están lejos, el valor se acerca a cero.

Los kernels se usan en modelos como Kernel Ridge, SVM y SVR.

```python
# Kernel Ridge con kernel polinomial y RBF

import numpy as np
import matplotlib.pyplot as plt

from sklearn.kernel_ridge import KernelRidge
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline

np.random.seed(42)

X = np.linspace(-3, 3, 160).reshape(-1, 1)
y_true = np.cos(2 * X.ravel()) + 0.2 * X.ravel()**2
y = y_true + np.random.normal(0, 0.18, size=X.shape[0])

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.3,
    random_state=42
)

models = {
    "Kernel polinomial": make_pipeline(
        StandardScaler(),
        KernelRidge(alpha=0.1, kernel="polynomial", degree=4, coef0=1)
    ),
    "Kernel RBF": make_pipeline(
        StandardScaler(),
        KernelRidge(alpha=0.08, kernel="rbf", gamma=1.5)
    ),
}

X_plot = np.linspace(-3.2, 3.2, 400).reshape(-1, 1)

plt.figure(figsize=(11, 4))
for k, (name, model) in enumerate(models.items(), 1):
    model.fit(X_train, y_train)
    y_plot = model.predict(X_plot)
    y_test_pred = model.predict(X_test)
    rmse = np.sqrt(mean_squared_error(y_test, y_test_pred))

    plt.subplot(1, 2, k)
    plt.scatter(X_train, y_train, color="tab:blue", s=16, alpha=0.7, label="Entrenamiento")
    plt.scatter(X_test, y_test, color="tab:red", s=18, alpha=0.8, label="Prueba")
    plt.plot(X_plot, y_plot, color="black", linewidth=2, label="Predicción")
    plt.plot(X_plot, np.cos(2 * X_plot.ravel()) + 0.2 * X_plot.ravel()**2, color="green", linestyle="--", label="Función real")
    plt.title(f"{name}\nRMSE prueba = {rmse:.3f}")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.legend()
    plt.grid(True)

plt.tight_layout()
plt.show()
```

## Regresión mediante vectores representativos: SVM y SVR

En regresión, la versión de máquinas de soporte vectorial se llama SVR: Support Vector Regression.

La idea no es minimizar todos los errores cuadrados como en OLS. SVR define un tubo de tolerancia alrededor de la función estimada. Los errores pequeños dentro del tubo no se penalizan.

Para cada observación:

$$
|y_i-\hat{y}_i|\leq \varepsilon
$$

Si el punto queda dentro de ese margen, su error se considera aceptable. Si queda por fuera, aparece una penalización.

El problema conceptual puede escribirse como:

$$
\min \frac{1}{2}\|\mathbf{w}\|^2 + C\sum_{i=1}^{M}(\xi_i+\xi_i^*)
$$

sujeto a:

$$
y_i-(\mathbf{w}^T\phi(x_i)+b)\leq \varepsilon+\xi_i
$$

$$
(\mathbf{w}^T\phi(x_i)+b)-y_i\leq \varepsilon+\xi_i^*
$$

$$
\xi_i,\xi_i^*\geq 0
$$

Donde:

- epsilon define el ancho del tubo.
- C controla cuánto se castigan los errores fuera del tubo.
- phi permite trabajar en espacios transformados mediante kernels.
- Los puntos que quedan fuera o sobre el borde del tubo son los que más influyen en el modelo. Por eso se llaman vectores de soporte.

SVR se usa cuando se quiere una regresión flexible, controlada y compatible con kernels.

```python
# Comparación entre SVR lineal y SVR no lineal

import numpy as np
import matplotlib.pyplot as plt

from sklearn.svm import LinearSVR, SVR
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split

np.random.seed(42)

X = np.linspace(0, 8, 160).reshape(-1, 1)
y_true = np.sin(1.5 * X.ravel()) + 0.15 * X.ravel()
y = y_true + np.random.normal(0, 0.18, size=X.shape[0])

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.3,
    random_state=42
)

models = {
    "LinearSVR": make_pipeline(StandardScaler(), LinearSVR(C=1.0, epsilon=0.1, random_state=42, max_iter=20000)),
    "SVR RBF": make_pipeline(StandardScaler(), SVR(kernel="rbf", C=10, epsilon=0.08, gamma=0.7)),
    "SVR polinomial": make_pipeline(StandardScaler(), SVR(kernel="poly", degree=3, C=8, epsilon=0.08, coef0=1)),
}

X_plot = np.linspace(0, 8, 400).reshape(-1, 1)

plt.figure(figsize=(12, 4))
for k, (name, model) in enumerate(models.items(), 1):
    model.fit(X_train, y_train)
    y_plot = model.predict(X_plot)
    y_test_pred = model.predict(X_test)
    rmse = np.sqrt(mean_squared_error(y_test, y_test_pred))

    plt.subplot(1, 3, k)
    plt.scatter(X_train, y_train, color="tab:blue", s=16, alpha=0.7)
    plt.scatter(X_test, y_test, color="tab:red", s=18, alpha=0.8)
    plt.plot(X_plot, y_plot, color="black", linewidth=2)
    plt.plot(X_plot, np.sin(1.5 * X_plot.ravel()) + 0.15 * X_plot.ravel(), color="green", linestyle="--")
    plt.title(f"{name}\nRMSE prueba = {rmse:.3f}")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.grid(True)

plt.tight_layout()
plt.show()
```

## Procesos Gaussianos para regresión

Un proceso Gaussiano es un modelo probabilístico para funciones. A diferencia de modelos que entregan solamente una predicción puntual, un proceso Gaussiano también estima incertidumbre.

La idea es considerar que los valores de la función en distintos puntos son variables aleatorias. Para cualquier conjunto de entradas:

$$
x_1,x_2,\dots,x_M
$$

los valores de la función:

$$
f(x_1),f(x_2),\dots,f(x_M)
$$

se modelan como una distribución normal multivariada:

$$
\mathbf{f}\sim \mathcal{N}(\boldsymbol{\mu},\mathbf{K})
$$

La matriz K es una matriz de covarianza construida con un kernel:

$$
K_{ij}=k(x_i,x_j)
$$

El kernel define qué tan parecidas se consideran dos entradas. Si dos puntos son cercanos, se espera que sus valores de salida también sean parecidos.

Para una nueva entrada x, el modelo produce una distribución predictiva:

$$
\hat{y}(x)\sim \mathcal{N}(\mu(x),\sigma^2(x))
$$

Donde:

- mu(x) es la predicción promedio.
- sigma(x) representa la incertidumbre.

Esto permite visualizar una curva central y una banda de confianza. La incertidumbre suele aumentar en regiones donde hay pocos datos observados.

```python
# Regresión con Gaussian Process: predicción e incertidumbre

import numpy as np
import matplotlib.pyplot as plt

from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import RBF, WhiteKernel, ConstantKernel
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

np.random.seed(42)

X = np.linspace(0, 10, 70).reshape(-1, 1)
y_true = np.sin(X.ravel()) + 0.1 * X.ravel()
y = y_true + np.random.normal(0, 0.18, size=X.shape[0])

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.3,
    random_state=42
)

kernel = ConstantKernel(1.0) * RBF(length_scale=1.2) + WhiteKernel(noise_level=0.05)
gp = GaussianProcessRegressor(kernel=kernel, normalize_y=True, random_state=42)
gp.fit(X_train, y_train)

X_plot = np.linspace(0, 10, 400).reshape(-1, 1)
y_mean, y_std = gp.predict(X_plot, return_std=True)
y_test_pred = gp.predict(X_test)
rmse = np.sqrt(mean_squared_error(y_test, y_test_pred))

print(f"Kernel aprendido: {gp.kernel_}")
print(f"RMSE prueba: {rmse:.4f}")

plt.figure(figsize=(10, 5))
plt.scatter(X_train, y_train, color="tab:blue", s=25, label="Entrenamiento")
plt.scatter(X_test, y_test, color="tab:red", s=30, label="Prueba")
plt.plot(X_plot, np.sin(X_plot.ravel()) + 0.1 * X_plot.ravel(), color="green", linestyle="--", label="Función real")
plt.plot(X_plot, y_mean, color="black", linewidth=2, label="Media predictiva")
plt.fill_between(
    X_plot.ravel(),
    y_mean - 2 * y_std,
    y_mean + 2 * y_std,
    color="gray",
    alpha=0.25,
    label="Incertidumbre ±2σ"
)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Gaussian Process Regression")
plt.legend()
plt.grid(True)
plt.show()
```

## Ejercicio con una base de datos tipo vivienda

Para cerrar el recorrido, se trabaja con un problema tipo Housing. El objetivo es predecir el precio de una vivienda a partir de variables numéricas y categóricas.

En un problema real, el modelo no empieza en la línea donde se llama a fit. Antes hay un flujo completo:

1. **Exploración**: revisar variables, tipos, rangos, distribución y valores faltantes.
2. **Limpieza**: decidir cómo tratar datos incompletos.
3. **Codificación**: convertir variables categóricas en información numérica.
4. **Escalamiento**: poner variables numéricas en escalas comparables cuando el modelo lo requiere.
5. **Partición**: separar entrenamiento y prueba.
6. **Validación cruzada**: comparar modelos e hiperparámetros de manera más estable.

Matricialmente, se desea construir:

$$
\hat{\mathbf{y}}=f(\mathbf{X})
$$

pero ahora X contiene columnas de distinta naturaleza. Algunas son continuas, como área o edad. Otras son categóricas, como zona.

Una variable categórica no se debe pasar directamente como texto al modelo. Primero se transforma. Por ejemplo, si zona puede tomar cuatro valores:

$$
\text{zona}\in\{norte,sur,centro,occidente\}
$$

con codificación one-hot se construyen columnas indicadoras. Una observación de zona norte puede representarse como:

$$
[1,0,0,0]
$$

De esta manera, el modelo recibe una matriz completamente numérica.

Como el cuaderno debe correr en local sin depender de descargas externas, se construye una base sintética inspirada en variables típicas de vivienda.

```python
# Construcción, exploración, limpieza y codificación de una base tipo Housing

import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.pipeline import Pipeline

np.random.seed(42)

n = 320
housing = pd.DataFrame({
    "area_m2": np.random.normal(95, 28, n).clip(30, 220),
    "habitaciones": np.random.randint(1, 6, n),
    "banos": np.random.randint(1, 4, n),
    "edad": np.random.exponential(18, n).clip(0, 90),
    "distancia_centro_km": np.random.gamma(2.5, 2.0, n).clip(0.5, 25),
    "estrato": np.random.choice([2, 3, 4, 5, 6], size=n, p=[0.18, 0.28, 0.25, 0.18, 0.11]),
    "zona": np.random.choice(["norte", "sur", "centro", "occidente"], size=n),
})

zona_efecto = housing["zona"].map({
    "norte": 35,
    "sur": -18,
    "centro": 20,
    "occidente": 5,
}).to_numpy()

precio = (
    55
    + 2.4 * housing["area_m2"].to_numpy()
    + 18 * housing["habitaciones"].to_numpy()
    + 25 * housing["banos"].to_numpy()
    - 1.3 * housing["edad"].to_numpy()
    - 4.5 * housing["distancia_centro_km"].to_numpy()
    + 28 * housing["estrato"].to_numpy()
    + zona_efecto
    + np.random.normal(0, 35, n)
)

housing["precio_millones"] = precio

for column in ["area_m2", "edad", "zona"]:
    missing_idx = np.random.choice(housing.index, size=12, replace=False)
    housing.loc[missing_idx, column] = np.nan

print("Primeras observaciones:")
display(housing.head())
print("\nValores faltantes por columna:")
print(housing.isna().sum())
print("\nResumen estadístico de variables numéricas:")
display(housing.describe())

X = housing.drop(columns="precio_millones")
y = housing["precio_millones"]

numeric_features = ["area_m2", "habitaciones", "banos", "edad", "distancia_centro_km", "estrato"]
categorical_features = ["zona"]

numeric_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", StandardScaler()),
])

categorical_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("encoder", OneHotEncoder(handle_unknown="ignore")),
])

preprocessor = ColumnTransformer(
    transformers=[
        ("num", numeric_transformer, numeric_features),
        ("cat", categorical_transformer, categorical_features),
    ]
)

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.25,
    random_state=42
)

print(f"\nTamaño entrenamiento: {X_train.shape}")
print(f"Tamaño prueba:        {X_test.shape}")
```

## Búsqueda de parámetros con validación cruzada

Muchos modelos tienen hiperparámetros: valores que no se aprenden directamente como coeficientes, sino que se escogen antes del entrenamiento.

Ejemplos:

- En Ridge, alpha controla la regularización.
- En SVR, C y epsilon controlan el margen y la penalización.
- En Random Forest, la profundidad y el número de árboles controlan la complejidad.

Si se escogen hiperparámetros mirando solo el conjunto de prueba, se contamina la evaluación. Por eso se usa validación cruzada dentro del conjunto de entrenamiento.

En validación cruzada con K particiones, el conjunto de entrenamiento se divide en K bloques. En cada iteración:

- K-1 bloques se usan para entrenar.
- 1 bloque se usa para validar.

El rendimiento promedio sobre los K bloques estima qué tan bien funciona esa configuración.

GridSearchCV prueba varias combinaciones de hiperparámetros y escoge la que obtiene mejor rendimiento promedio.

El flujo queda así:

1. Separar entrenamiento y prueba.
2. Hacer validación cruzada solo dentro del entrenamiento.
3. Escoger mejores hiperparámetros.
4. Evaluar una sola vez en prueba.

Esto protege la estimación final de desempeño.

```python
# Validación cruzada para comparar modelos de regresión

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.linear_model import ElasticNet
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor
from sklearn.kernel_ridge import KernelRidge
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

models_and_grids = {
    "ElasticNet": (
        ElasticNet(max_iter=10000, random_state=42),
        {
            "model__alpha": [0.01, 0.1, 1.0],
            "model__l1_ratio": [0.2, 0.5, 0.8],
        },
    ),
    "SVR": (
        SVR(),
        {
            "model__C": [1, 10, 50],
            "model__epsilon": [0.1, 0.5],
            "model__kernel": ["rbf"],
            "model__gamma": ["scale"],
        },
    ),
    "RandomForest": (
        RandomForestRegressor(random_state=42),
        {
            "model__n_estimators": [80, 140],
            "model__max_depth": [4, 8, None],
        },
    ),
    "KernelRidge": (
        KernelRidge(kernel="rbf"),
        {
            "model__alpha": [0.1, 1.0],
            "model__gamma": [0.1, 1.0],
        },
    ),
}

summary = []
best_estimators = {}

for name, (estimator, grid) in models_and_grids.items():
    pipe = Pipeline(steps=[
        ("preprocess", preprocessor),
        ("model", estimator),
    ])

    search = GridSearchCV(
        pipe,
        param_grid=grid,
        scoring="neg_root_mean_squared_error",
        cv=4,
        n_jobs=None,
    )
    search.fit(X_train, y_train)

    best_model = search.best_estimator_
    best_estimators[name] = best_model
    y_pred = best_model.predict(X_test)

    summary.append({
        "modelo": name,
        "mejor_RMSE_CV": -search.best_score_,
        "RMSE_prueba": np.sqrt(mean_squared_error(y_test, y_pred)),
        "MAE_prueba": mean_absolute_error(y_test, y_pred),
        "R2_prueba": r2_score(y_test, y_pred),
        "mejores_parametros": search.best_params_,
    })

results_df = pd.DataFrame(summary).sort_values("RMSE_prueba")
display(results_df[["modelo", "mejor_RMSE_CV", "RMSE_prueba", "MAE_prueba", "R2_prueba"]])

print("Mejores parámetros:")
for _, row in results_df.iterrows():
    print(f"\n{row['modelo']}")
    print(row["mejores_parametros"])

best_name = results_df.iloc[0]["modelo"]
best_model = best_estimators[best_name]
y_best_pred = best_model.predict(X_test)

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.bar(results_df["modelo"], results_df["RMSE_prueba"], color="tab:blue")
plt.ylabel("RMSE en prueba")
plt.title("Comparación de modelos")
plt.xticks(rotation=20)
plt.grid(axis="y")

plt.subplot(1, 2, 2)
plt.scatter(y_test, y_best_pred, color="tab:purple", alpha=0.75)
lims = [min(y_test.min(), y_best_pred.min()), max(y_test.max(), y_best_pred.max())]
plt.plot(lims, lims, color="black", linestyle="--")
plt.xlabel("Precio real")
plt.ylabel("Precio predicho")
plt.title(f"Mejor modelo: {best_name}")
plt.grid(True)

plt.tight_layout()
plt.show()
```

El cuaderno completo muestra una ruta progresiva para estudiar regresión.

Primero se entiende la regresión como una dependencia estadística entre variables aleatorias. Luego se estudia el caso lineal, donde el modelo se expresa como una recta y los parámetros se estiman minimizando errores cuadráticos.

Después se revisa cómo evaluar el modelo: no basta con ajustar una función, también hay que medir residuales, errores y desempeño en datos de prueba.

A partir de ahí aparecen los problemas prácticos:

- El ruido aumenta la incertidumbre.
- Los valores atípicos pueden distorsionar OLS.
- La colinealidad vuelve inestables los coeficientes.
- La regularización controla la magnitud de los parámetros.
- Las relaciones no lineales requieren transformaciones, expansiones o modelos más flexibles.

Finalmente, el flujo tipo Housing muestra que un problema de regresión aplicado incluye todo un pipeline: exploración, limpieza, codificación, partición, entrenamiento, validación cruzada y evaluación final.

La idea central es que un regresor no es solamente una fórmula. Es una cadena completa de decisiones matemáticas y computacionales para aproximar una relación de dependencia y verificar si esa aproximación funciona con datos nuevos.
